```python
'''RECTILINEAR AND RADIAL AXIS..

-BOTH COLUMN NUMERIC:---SCATTER PLOT:--- POINT GRAPH/LINE GRAPH.
```

```python
import matplotlib.pyplot as mb
hours =[1,2,3,4,5,6]
marks=[40,50,60,70,90,99]

mb.scatter(hours,marks)   # rule:-BOTH COLUMN NUMERIC,,point graph(you can also join its points)
mb.title('study hours vs marks')
mb.xlabel('hours studied')
mb.ylabel('marks')
mb.show()
```

```python
'''BAR CHART:-MB.BAR(X,Y,COLOR,WIDTH) 
-ONE CATOGERY AND OTHER NUMERIC
-DONT SUPPORT VALUES ON X-AXIS AND NO CATOGERY ON Y-AXIS'''
```

```python
'''HISTOGRAM:-MB.HIST(ONLY_X,BINS,COLOR,EDGECOLOR)
DEFAULT BIN=10
MORE BIN = MORE DETAILED DISTRIBUTION
LESS BINS = MORE GENERALIZED
-ONLY ONE COLUMN THAT ALSO NUMERIC.
-ALWAYS MAKE CONTINUOUS GRAPH.
-ITS FOR FRQUENCY  IN ITS Y-AXIS BY DEFAULT.'''
```

```python
'''MB.PLOT:-MB.PLOT(X,Y,COLOR,LINE-WIDTH,LINE-STYLE,MARKER)
-YOU CAN MAKE LINE.
-BOTH MUST BE NUMERIC.'''
```

```python
# USING EVERYTHING TOGETHER:-

x=[1,2,3,4]
y=[10,44,29,74]

mb.plot(x,y)
mb.scatter(x,y)
mb.bar(x,y)

mb.title('using figure and axis')
mb.xlabel('x values')
mb.ylabel('y values')

mb.show()
```
