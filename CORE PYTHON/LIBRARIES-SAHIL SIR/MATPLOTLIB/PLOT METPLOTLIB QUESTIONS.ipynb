```python
'''Create a simple line plot using: 
x = [1,2,3,4,5] 
y = [10,20,30,40,50] 
Tasks: 
Plot the data
Display the graph'''
import matplotlib.pyplot as mp

x = [1,2,3,4,5]
y = [10,20,30,40,50]
mp.plot(x,y)
mp.show()
```

```python
'''Using the same data: 
Add title "Simple Line Plot" 
Label x-axis as "X Values" 
Label y-axis as "Y Values" '''
mp.plot(x,y)
mp.title('simple line plot')
mp.xlabel('x-values')
mp.ylabel('y-values')
mp.show()
```

```python
'''Create a line plot: 
y = [5, 15, 25, 35, 45] 
Tasks: 
Do not pass x values 
Observe what Matplotlib uses as default x'''
import matplotlib.pyplot as mp

y = [5, 15, 25, 35, 45]   # When you don't provide x values, Matplotlib automatically uses: x = [0, 1, 2, 3, 4]

mp.plot(y)
mp.show()
```

```python
'''Plot the following: 
X = = [1,2,3,4] 
y = [100,200,150,300] 
Tasks: 
• Change line color to red 
• Increase linewidth to 3'''
x = [1,2,3,4]
y = [100,200,150,300]
mp.plot(x,y,color='red',linewidth='3')
mp.show()
```

```python
'''Using same data:
• Change linestyle to dashed 
• Add marker 'o' '''

x = [1,2,3,4]
y = [100,200,150,300]
mp.plot(x,y,linestyle='--',marker='o',color='grey')
mp.show()
```

```python
'''Plot: 
X = [1,2,3,4,5] 
y = [10,30,20,40,25] 
Tasks: 
• Use green color 
• Use dotted line 
• Add square markers'''

x = [1,2,3,4,5]
y = [10,30,20,40,25]
mp.plot(x,y,color='green',linestyle='dotted',marker='s')
mp.show()
```

```python
'''Create a sales dataset: 
months = ["Jan","Feb","Mar","Apr"] 
sales = [100,200,150,300] 
Tasks:
Plot the data 
Add proper title and axis labels'''
months = ["Jan","Feb","Mar","Apr"] 
sales = [100,200,150,300]
mp.plot(months,sales,linestyle='--',color='grey',marker='^',markerfacecolor='black')
mp.xlabel('x-months')
mp.ylabel('y-sales')
mp.title('months X sales')
mp.show()
```

```python
'''Plot the same data: 
Tasks: 
• Add label "Sales Data"
• Display legend'''
months = ["Jan","Feb","Mar","Apr"] 
sales = [100,200,150,300]
mp.plot(months,sales,linestyle='--',color='grey',marker='^',markerfacecolor='black',label='sales data')
mp.legend()              # This displays the legend box containing Sales Data.
mp.show()
```

```python
'''Plot two lines in same graph: 
X = [1,2,3,4] 
y1 = [10,20,30,40] 
y2 = [40,30,20,10] 
Tasks:
• Use different colors 
• Add labels for both lines
Show legend'''


x = [1,2,3,4] 
y1 = [10,20,30,40] 
y2 = [40,30,20,10]
mp.plot(x,y1,color='red',label='line 1')
mp.plot(x,y2,color='black',label='line 2')
mp.legend()
mp.show()
```
