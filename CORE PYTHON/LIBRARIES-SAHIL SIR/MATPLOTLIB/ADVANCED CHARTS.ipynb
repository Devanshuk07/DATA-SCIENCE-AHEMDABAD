```python
'''1. Stacked Bar Chart 
Purpose 
Shows contribution of multiple categories in one bar. '''

import matplotlib.pyplot as mp

months=['jan','feb','mar']
product_a=[100,150,200]
product_b=[180,120,210]
mp.bar(months,product_a,label='product a')
mp.bar(months,product_b,bottom=product_a,label='product b')  # Because I used bottom=product_a, the bars of Product B are drawn on top of the bars of Product A.
# it is usual two graphs like stucture but due to bottom in mp.bar with product_b it is stacked bar chart.
mp.show()
```

```python
import matplotlib.pyplot as mp
import numpy as np

months=['jan','feb','mar']
product_a=[100,150,200]
product_b=[180,120,210]
x=np.arange(len(months))
width=0.2

mp.bar(x-width/2,product_a,width,label='product a')     # the third argument width specifies how wide each bar should be.
mp.bar(x+width/2,product_b,width,label='product b')
mp.show()
```

```python
'''AREA CHART:--
Purpose 
Shows trend and magnitude over time.'''

x=[1,2,3,4]
y=[100,200,150,300]

mp.fill_between(x,y,color='red')

mp.title('area chart')
mp.show()
```

```python
''' MP.ANNOTATION:--
PURPOSE
ADD TEXT WITH ADDITIONAL ARROW POINTING TO A DATA POINT.'''

x=[1,2,3,4]
y=[100,200,150,300]

mp.fill_between(x,y,color='red')

mp.annotate('highest sales',xy=(4,300),xytext=(2,250),arrowprops=dict(facecolor='black'))   # xytext=(2,250)-->Position of the text,,,arrowprops stands for arrow properties

mp.show()
```
