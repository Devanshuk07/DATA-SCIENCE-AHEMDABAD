```python
# CREATE TWO ARRAY : A=[5,10,15] B=[1,2,3] PERFORM ELEMENT-WISE ADDITION AND MULTIPLICATION:-
import numpy as np

a=[5,10,15]
b=[1,2,3]
print(np.add(a,b))
print(np.multiply(a,b))
```

```python
# GIVEN ARRAY: ARR=[4,8,12,16] EXTRACT ELEMENTS GREATER THAN 10.
import numpy as np

arr=np.array([4,8,12,16,20])
print(arr[arr>10])   # this arr[arr>10]---> this will add the elements to arr but only those which is correct and condition is[arr>10].
```

```python
# CREATE A 2×3 ARRAY OF ONES AND ADD 5 TO EVERY ELEMENT USING BROADCASTING:-
one=np.ones([2,3])
print(one,'\n')
print('AFTER ADDING 5 TO EACH ELEMENT:-'+'\n',np.add(one,5))
```

```python
# GIVEN ARRAY:ARR = [10, 20, 30, 40] COMPUTE: MEAN,SUM,MAXIMUM:-
arr=[10,20,30,40]
print(np.mean(arr))
print(np.sum(arr))
print(np.max(arr))
print(np.min(arr))
```

```python
# CONVERT 45 DEGREES INTO RADIANS AND COMPUTE ITS SINE VALUE.:-
import math

a=int(input('ENTER THE DEGREE :-'))
rad=math.radians(a)
print('RADIAN IS :-',rad)
print('ITS SIN VALUE IS :-',math.sin(rad))
```

```python
# CREATE TWO ARRAYS:A = [2, 4, 6, 8] B = [1, 2, 3, 4] PERFORM:FLOOR DIVISION,MODULUS OPERATIONS:-
a=np.array([2,4,6,8])
b=np.array([1,2,3,4])
print('FLOOR DIVISION IS :-',a//b)
print('MODULES ARE :-',a%b)
```

```python
# GENERATE A 3×3 RANDOM INTEGER MATRIX BETWEEN 1 AND 20.FIND:STANDARD DEVIATION,VARIANCE:-
mat=np.random.randint(1,21,(3,3))
print(mat)
print('STANDARD DEVIATION :-',np.std(mat))
print('VARIANCE :-',np.var(mat))
```

```python
# CREATE AN ARRAY FROM 1 TO 9 AND RESHAPE IT INTO 3×3.COMPUTE CUMULATIVE SUM (CUMSUM):-
a=np.arange(1,10)
b=a.reshape(3,3)
print(a)
print(b)
print(np.cumsum(b))  # cumsum means add 1st then 2nd then 3rd like this.1+2+3 like fibonacci series.
```

```python
# CREATE ARRAY:ARR = [-3, -1, 0, 1, 3]APPLY:SQUARE,SQUARE ROOT (ONLY VALID VALUES):-
arr=np.array([-3,-1,0,1,3])
print('SQUARE :-',np.square(arr))
print('SQUARE ROOT :-',np.sqrt(arr[arr>=0]))   #TO STORE ONLY THE VALID VALUES.IN ARR.
```

```python
# Create a 3×4 matrix using np.arange().Add 1D array [10, 20, 30, 40] using broadcasting:-
mat=np.arange(0,12).reshape(3,4)
print('MATRIX FORMED :-','\n',mat,'\n')
arr=[10,20,30,40]
res=mat+arr
print('BROADCASTING IS DONE :-','\n',res)

'''broadcasting means :-“Automatically expanding dimensions to make shapes compatible.”
Most of the time broadcasting happens automatically in NumPy.
There is a function called np.broadcast, but it is mainly for advanced/internal use, not normal calculations.'''
```

```python
# Generate 1000 random numbers from normal distribution.Compute:Mean,Standard deviation,95th percentile:-
a=np.random.rand(1,1000)
print(a)
print('MEAN IS :-''\n',np.mean(a))
print('STANDARD DEVIATION MEANS :-',np.std(a))
print('95TH PERCENTILE :-',np.percentile(a,95))
```

```python
#Create two matrices:A (2×3) B (3×2) Transpose A and verify its shape:-
a=np.arange(1,7).reshape(2,3)
print(a)
b=np.arange(1,7).reshape(3,2)
print(b)
```

```python
# Create array:arr = [1, 2, 3, 4]Compute:Exponential of each element,Natural log,Log base 10:-
arr=[1,2,3,4]
print('EXPONENTIAL:-',np.exp(arr))
print('NATURAL LOG:-',np.log(arr))
print('LOG BASE 10 :-',np.log10(arr))
```

```python
# Create array:arr = [1.2, 2.5, 3.7, 4.1]Apply:round()floor()ceil()Explain differences based on output:-
arr=[1.2, 2.5, 3.7, 4.1]
print(np.round(arr))    # round means it will decide just by if 3.5 =3 and 3.6=4 like that auto decide.
print(np.floor(arr))   # floor means after point goes to -1 value.
print(np.ceil(arr))    # ceil means after point add+1 to a value.
```
