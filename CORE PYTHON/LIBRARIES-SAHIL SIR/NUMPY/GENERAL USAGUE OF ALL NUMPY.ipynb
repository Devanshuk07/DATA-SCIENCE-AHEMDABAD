```python
import numpy as np

prices=np.array([100,200,150,300,np.nan,250,900])

quantity=np.array([10,5,8,3,7,6,2])

items=np.array(['RICE','MILK','TOMATOS','PANEER','SOYA','MEAT','GOLD'])

print('PRICES:- ',prices)
print('\nQUANTITY :',quantity)
mask=~np.isnan(prices)

original_price=prices[mask]
original_quantity=quantity[mask]
original_item=items[mask]

print('\nORIGINAL PRICE:-',original_price)
print('\nORIGINAL QUANTITY:-',original_quantity)
print('\nORIGINAL ITEM:-',original_item)
total_sales=original_price*original_quantity
print('\nTOTAL SALES:-',total_sales)
print('\nTOTAL SUM FOR TOTAL SALES :- ',np.sum(total_sales))
print('\nTHIS IS AVERAGE SPENT PER MONTH ON THIS ITEMS:-',np.mean(total_sales))
print('\nTHIS IS THE MAXIMUM RUPEE SPENT ON AN ITEM :-',np.max(total_sales))
print('\nTHIS IS THE MINIMUM RUPEE SPENT ON AN ITEM :-',np.min(total_sales))
print('\nTHIS IS THE INDEX OF MAXIMUM VALUE OF AN ITEM(price x quantity) :-',np.argmax(total_sales))
print('\nTHIS IS THE INDEX OF MINIMUM VALUE OF AN ITEM(price x quantity) :-',np.argmin(total_sales))
print('\nTHE SORTED LIST OF TOTAL SALES :-',np.sort(total_sales))
print('\nTHE UNIQUE LIST OF TOTAL SALES :-',np.unique(total_sales))
print('\nFINDING THE MILK FROM ITEMS:-',np.where(original_item=='MILK','MILK','NOT THERE'))
print('\nCLIPPING THE TOTAL SALES FROM 1000 TO 1300 ONLY :-',np.clip(total_sales,1000,1300))
print('\nFINDING IF I HAD TOMATOS LEFT OR NOT :-',np.any(original_item=='TOMATOS'))
print('\nIF MY TOTAL SALES HAS EVERYTHING TO 1000 RUPEES ONLY :-',np.all(total_sales==1000))

a=np.concatenate((original_price,original_quantity))
print('\nTOTAL PRICE WITH QUANTITY CONCATENATED :-',a)

b=original_item.reshape(3,2)
print('\nORIGINAL ITEMS IN TABLE :-\n',b)
print('FLATTEN THE ITEMS/TABLE:- ',b.flatten())
print('\nDTYPE OF ORIGINAL PRICES:- ',original_price.dtype)
print('\nNUMBER OF DIMENSIONS IN ORIGINAL ITEM TABLE :-',np.ndim(b))
print('\nTHIS IS SHAPE(ROWS X COLUMNS) TABLE :-',np.shape(b))
print('\nSIZE--> 2x3=6 OF TABLE :-',np.size(b))
print('\nCOUNTING NON-ZERO ORIGINAL QUANTITY :-',np.count_nonzero(original_quantity))
print('\nCHECKING IS THERE ANY NAN VALUES:-',np.isnan(prices))
print('\nGETTING MEAN WITHOUT NAN VALUES:-',np.nanmean(prices))
print('\nGETTING MAXVALUE WITHOUT NAN VALUES:-',np.nanmax(prices))
print('\nGETTING MINVALUE WITHOUT NAN VALUES:-',np.nanmin(prices))
print('\nADDING GHEE TO ITEMS :-',np.insert(items,4,'GHEE'))
print('\nDELETING THE ITEMS :-',np.delete(items,4))
print('\nAPPEND THE ITEMS LIST:- ',np.append(items,'BULB'))
print('\nUSING NP.DIFF ON ORIGINAL QUANITY :-',np.diff(original_quantity))
print('\nUSING CUMSUM ON ORIGINAL PRICE:-',np.cumsum(original_price))
print('\nUSING ROUND OF TO ORIGINAL QUANTITY:- ',np.round(original_quantity))
print('\nTAKING SQUARE ROOT OF ORIGINAL PRICES:-',np.sqrt(original_price))
print('\nRAISING POWER TO ORIGINAL PRICES:-',np.power(original_quantity,2))
print('\nSORTING THE TABLE :-',np.sort(b))
print('\nTRANSPOSE ON TABLE :-',np.transpose(b,(1,0)))
print('\nUSING REPETATION FOR TABLE :-',np.repeat(b,2))
print('\nUSING NP.EYE :-',np.eye(6,7))
print('\nUSING NP.ZEROS :-',np.zeros((6,7)))
print('\nUSING NP.ONES :-',np.ones((6,7)))
print('\nUSING NP.linspace :-',np.linspace(1,7,20))   #start,stop,number of values
print('\nTAKING RANDOM RAND :-',np.random.rand(3,1))
print('\nTAKING RANDOM CHOICE FROM ORIGINAL PRICE:-',np.random.choice(original_price))
print('\nTAKING RANDOM RAND :-',np.random.randint(2,4))
#print('',np.random.seed())   when to use seed--> its lock randomness so other element depend on one we are using dont change.
print('\nUNIQUE WITH RETURN_COUNTS:-',np.unique(total_sales,return_counts=True))
print('\nUSING ARGSORT On ORIGINAL QUANTITY :-',np.argsort(original_quantity))  #GIVES INDEX SORTING OF PERTICULAR ARRAY
```
