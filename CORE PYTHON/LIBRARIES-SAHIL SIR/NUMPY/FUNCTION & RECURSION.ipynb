```python
def num(a):
    b=sorted(a)
    return b[-1]

a=[2,4,6,8,9,10,15,2,22,78,90]
print(num(a))
```

```python
def num(a):
    largest=a[0]

    for i in a:
        if i>largest:
            largest=i
    return largest

a=[2,4,6,8,9,10,15,2,22,78,90]
print(num(a))
```

```python
def st(a):
    count=0
    for i in a:
        if i in ['a','e','i','o','u']:
            count=count+1
    return count
a=str(input('ENTER THE STRING :-'))
print(st(a))
```

```python
#BASIC RECURSION PROGRAM:-

def nu(a):    # HERE A = 5 FIRST,THEN A=4....
    if a==0:
        return

    print(a)  # 5 IS NOT EQUAL TO 0 SO IT WILL PRINT
    nu(a-1)   # THEN IT WILL BE A-1 SO ===5-1=4.
nu(5)
```

```python
def nu(a):
    if a==5:
        return 5
        
    return a+nu(a+1)

print(nu(1))
```

```python
def ch(a):
    if a=='':
        return ''
        
    return a[0] + ch(a[1:])
ch('hello')
```

```python
#ALWAYS USE THIS FOR FACT OR ADDITION OF NUMBERS A==1 NOT A==5.
def fac(a):
    if a==1:
        return 1

    return a*fac(a-1)   # fac(5)* 4 then fact(4)*3 like this 5*4*3*2*1=120
print(fac(5))
```

```python
#JUST TO SHOW A==10 BUT IN REVERSE ORDER:-
def fac(a):
    if a==10:
        return 10

    return a*fac(a+1)   
print(fac(1))
```

```python
# TO REVERSE A STRING:-
def re(a):
    if a=='':
        return ''
    return a[-1] + re(a[:-1])
print(re('vijay'))
```

```python
# EXAMPLE OF PALIDROME:-
def re(a):
    if len(a)<=1:       # If string has 1 character or no character, then it is automatically a palindrome.”
        return 'IS A PALIDROME'
    if a[0]!=a[-1]:
        return False
    return re(a[1:-1])    # start from [1:-1] from 1 till -1 means full string.every time it return mm then aa and then last d remains that goes to <=1 so it is palidrome.

print(re('madam'))
```
