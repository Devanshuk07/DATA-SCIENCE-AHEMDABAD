```python
'''PIVOT:---
In pandas, a pivot is used to reshape data.
It converts values from rows into columns,making the data easier to summarize and analyze.
SYNTAX:-

DF.PIVOT(INDEX,COLUMNS,VALUES)
```

```python
import pandas as pd

df2 = pd.DataFrame({
    'City':['Rajkot','Rajkot','Ahmedabad','Ahmedabad'],
    'Product':['Mobile','Laptop','Mobile','Laptop'],
    'Sales':[50,30,70,40]
})
df2.pivot(index='Product',columns='City',values='Sales')
df2
```

```python
import pandas as pd
import numpy as np

data={'month':['feb','feb','jan','jan','hh','jj'],
      'product':['a','b','a','b','b','b'],
      'sales':[100,200,150,250,150,250]
     }
df=pd.DataFrame(data)
df
df2=df.pivot(index='month',columns='product',values='sales')
df2   # PIVOT MAKES A NEW DF .THEN WE CAN APPLY ANYTHING TO IT.
```

```python
import pandas as pd
import numpy as np

data={'month':['feb','feb','jan','jan'],
      'product':['a','b','b','b'],   # GIVING JAN 2 TIMES B.THAT MEANS DUPLICATE.
      'sales':[100,200,150,250]
     }
df=pd.DataFrame(data)
df
df2=df.pivot(index='month',columns='product',values='sales')
df2   # PIVOT cannot handle duplicate value so it will give error.
```

```python
import pandas as pd
import numpy as np

data={'month':['feb','feb','jan','jan'],
      'product':['a','b','b','b'],
      'sales':[100,200,150,250]
     }
df=pd.DataFrame(data)

df3=df.pivot_table(index='month',columns='product',values='sales',aggfunc='sum')  
# aggfunc stands for Aggregation Function. It tells pivot_table() what to do when multiple values exist for the same group.
df3   # pivot_table used to handle extra value it will sum the values.jan has two b so it will add that similar values.
```
