```python
data={
    "Name" : ['rahul','priya','aman'],
    'age':[25,29,24],
    'salary':[50000,65000,48000]
}
df=pd.DataFrame(data)  # A DataFrame in pandas is a 2D table (like Excel sheet or SQL table).It converts your dictionary into a table structure.A DataFrame = collection of Series (columns)
print(df)
print('\n',df[1:2])
```

```python
data={
    "Name" : ['rahul','priya','aman'],
    'age':[25,29,24],
    'salary':[50000,65000,48000]
}
dt=pd.DataFrame(data)
print(dt)
print('\n',dt.head())   # HEAD JUST SHOW FIRST 5 ROWS ONLY.
print('\n',dt.tail())  # tail used to show last 5 values only.
print('\n',dt.info())  # 
print(dt.describe())   # every mathematical(min,mean,count,std,max) it show them all for int column only.
```

```python
import pandas as pd
data={'customer':['aman','priya','rahul','sneha','jeet'],
      'city':['delhi','mumbai','delhi','banglore','goa'],
      'revenue':[12000,8000,15000,7000,20000]
     }
dd=pd.DataFrame(data)
print('\n1st:-',dd)
print('\n2nd:-',dd[['city','customer','revenue']])  # selecting multiple columns.
print('\nJUST LOC WITH 2 :----\n',dd.loc[2])     # just for 2 it will fetch 2nd index from every column
print('\nLOC:--REVENUE',dd.loc[1,'revenue'])  # now we have given specific 'revenue' name then it will fetch that perticular value only.
print('\n5th:-',dd[dd['revenue']>10000])  # return only when true.
print('\n6th:-',dd[(dd['revenue']>10000) & (dd['city']=='delhi')])
print('\n7th:-',dd.iloc[2,2])  # 2-row,2-column.
print('\n8th:-',dd.query("revenue >10000 and city=='delhi'"))          # dd.query() is a pandas method used to filter rows using a string expression instead of writing long bracket conditions.
```
