```python
#SORTING AND REINDEXING:-
import pandas as pd
data={'customer':['aman','priya','rahul','sneha','jeet'],
      'city':['delhi','mumbai','delhi','banglore','goa'],
      'revenue':[12000,8000,15000,7000,20000]
     }
dd=pd.DataFrame(data)
#dd.sort_values(by='customer')
#dd.sort_values(by='city',ascending=False) # ascending by default True.
dd.sort_values(by=['city','revenue']) # revenue sort when two city is common then for that revenue is sort to ascending order.
#dd.sort_index()   # changing to required index after sort_values.
```

```python
sorted_dd=dd.sort_values(by=['city','revenue'])
sorted_dd.reset_index()   # to sort index according to sor_values.
#sorted_dd.reset_index(drop=True) # drop will drop the original index forever so we cannot change it again to original.
```
