```python
'''Create a DataFrame with columns: 
Name, Age, City, Salary 
Tasks:
Display only the Age column 
Display Name and Salary together'''
import pandas as pd
import numpy as np

data={'name':['Devanshu','Viral','Manoj','Veeru','Jay'],
      'age':[29,23,22,39,21],
      'city':['Rajkot','Ahemdabad','delhi','delhi','Mevada'],
      'salary':[22000,23000,34000,24000,34000]
     }
df=pd.DataFrame(data)
df
print(df['age'])
print('\n',df[['name','salary']])
```

```python
'''Using the same dataset:
Select the 3rd row using iloc 
Select the Salary of the 2nd row using loc '''

print(df.iloc[3])
print(df.loc[2,'salary'])
```

```python
'''
|  Method                              | Uses              |
|------------------------------------- | ----------------- |
| loc[row_label, column_label]`        | Names/labels      |
| iloc[row_position, column_position]` | Numeric positions |
'''
```

```python
'''Filter the dataset:
Show rows where Salary > 50000 
Show rows where City is "Delhi"'''
print(df[df['salary']>22000])
print(df[df['city']=='delhi'])
```

```python
'''Apply multiple conditions: 
Salary > 40000 AND City = "Mumbai"'''
print(df[(df['salary']>23000) & (df['city']=='delhi')])
```

```python
'''Sort the dataset: 
Sort by Salary in ascending order 
Sort by Salary in descending order'''
#df.sort_values('salary')
df.sort_values('salary',ascending=False)
```

```python
'''After sorting:
Reset the index 
Remove the old index'''
df.sort_values('salary',ascending=False)
df.reset_index()
df.reset_index(drop=True)
```

```python
'''Create a Sales column: 
[100, 120, 110, 115, 500] 
Tasks:
Detect outliers using IQR method'''
import pandas as pd
import numpy as np
data={'name':['Devanshu','Viral','Jay','Veeru','Jay'],
      'age':[29,23,21,39,21],
      'city':['Rajkot','Ahemdabad','Mevada','delhi','Mevada'],
      'salary':[22000,23000,34000,33000,34000]
     }
df=pd.DataFrame(data)
df
df['sales']=[100, 120, 110, 115, 500]
df

q1=df['sales'].quantile(0.25)
q3=df['sales'].quantile(0.75)

QTR=q3-q1

lower=q1-1.5*QTR
upper=q3+1.5*QTR

print(df[(df['sales']<lower) | (df['sales']>upper)])
```

```python
'''Using the same sales data:
Calculate Z-score for each value
Identify values where Z-score > 2'''
df['sales']=[100, 120, 110, 115,5500]

z_score=(df['sales']-df['sales'].mean())/df['sales'].std()
print(z_score)
```

```python
'''Create a dataset: 
Product, Price, Quantity 
Tasks:
Create new column Revenue = Price * Quantity 
Calculate total revenue'''
import pandas as pd
import numpy as np

data={'product':['pen','eraser','lazer','book','note','ghee'],
      'price':[20,30,90,190,290,3000],
      'qty':[20,22,3,20,21,3]
     }
df=pd.DataFrame(data)
df['revenue']=df['price']*df['qty']
df
a=df['revenue'].sum()
print(f'{df}\n\ntotal revenue is :-{a}')
```

```python
'''Use apply() and lambda:
Increase all prices by 10%
Create a new column Discounted Price = Price - 10%'''
df['price_incresed']=(df['price']*1.10)-df['price']
df['after_increased']=(df['price']+df['price_incresed'])*df['qty']
df['discounted_price_5%']=df['after_increased']-(df['after_increased']*0.05)
df
```

```python
'''Create a dataset: 
Customer, Age, City, Price, Quantity 

Include:
Missing values in Age and City  
Duplicate rows 

Tasks: 
1 Clean missing values 
-Replace Age with mean 
-Replace City with "Unknown" 
2 Remove duplicate rows 
3 Create new column: 
-Revenue = Price * Quantity 
4 Filter rows where Revenue > 500 
5 Sort the dataset by Revenue in descending order 
6 Reset the index 
7 Identify any outliers in Revenue using IQR method'''

import pandas as pd
import numpy as np
data={'customer':['Devanshu','Viral','Jay','Veeru','Jay'],
      'age':[29,np.nan,21,39,21],
      'city':[np.nan,'Ahemdabad','Mevada','delhi','Mevada'],
      'price':[20,30,190,3000,190],
      'qty':[20,12,3,21,3]
     }
df=pd.DataFrame(data)
df['age']=df['age'].fillna(df['age'].mean())
df['city']=df['city'].fillna('unknown')
df=df.drop_duplicates()
df['revenue']=df['price']*df['qty']
print(df[df['revenue']>400])
df=df.sort_values('revenue',ascending=False)
df.reset_index()

q1=df['price'].quantile(0.25)
q3=df['price'].quantile(0.75)

IQR=q3-q1

lower=q1-1.5*IQR
upper=q3+1.5*IQR

outlier=df[(df['price']<lower)|(df['price']>upper)]
print('\n',outlier)
```

```python
# HOW PROGRAM CHECKS EVERY COLOUMN FOR OUTLIER INDEPENDENTLY BY LOOP:-

df=pd.DataFrame(data)
df
numerical_cols=df.select_dtypes(include=np.number).columns  # select_dtypes used to select the column according to its data type(include=np.number)also include='object' for string.

for col in numerical_cols:
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    IQR=q3-q1
    lower=q1-1.5*IQR
    upper=q3+1.5*IQR
    outlier=df[(df[col]<lower)|(df[col]>upper)]
    print(f'\n OUTLIERS IN {col}')
    if outlier.empty:
        print("No outliers found")
    else:
        print(outlier)
```
