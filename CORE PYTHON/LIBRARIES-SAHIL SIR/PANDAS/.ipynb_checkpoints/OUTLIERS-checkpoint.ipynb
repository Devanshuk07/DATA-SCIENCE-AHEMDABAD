```python
'''An outlier is a value that is very different from the rest of the data.
EXAMPLE:-
a=[22, 25, 24, 23, 26, 120]
Here, 120 is an outlier because all the other ages are around 22–26.'''
```

```python
import numpy as np

ages =np.array([22, 25, 24, 23, 26])
print(ages.mean())
```

```python
ages=np.array([22,24,23,22,120])
print(ages.mean())  # this is how outliers affect the mean.
```

```python
# EXAMPLE OF OUTLIER:--
import pandas as pd
import numpy as np

salary = np.array([20000, 22000, 21000, 23000, 24000, 25000, 150000])

df = pd.DataFrame({'SALARY':salary})

print(df)
print('\nMEAN :--',salary.mean())     #The outlier is pulling the average upward.
```

```python
# How to find outliers quickly?
import pandas as pd
salary =list(np.random.randint(20000,80000,491))
salary+=[150000, 180000, 220000, 250000, 300000,350000, 400000, 450000, 500000, 750000]
df = pd.DataFrame({'salary':salary})
print(df.describe())

q1=df['salary'].quantile(0.25)   # Pandas looks for an exact match:'salary'
print('\nvalue of q1:-',q1)
q3=df['salary'].quantile(0.75)   # quantile() is used to find the value below which a certain percentage of data falls.
print('\nvalue of q3:-',q3)

IQR=q3-q1

lower=q1-1.5*IQR       # Any value outside these limits is considered an outlier.
upper=q3+1.5*IQR

print('\nLOWER :-',lower)
print('\nUPPER',upper)

print(df[(df['salary']<lower) | (df['salary']>upper)])  # df['salary']<lower is condition and if true df[...] returns only the rows where the final result is True,,,,  | means OR.
```
