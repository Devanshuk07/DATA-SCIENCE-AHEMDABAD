```python
# USING .DT.YEAR/DT.MONTH.DT.DAY

import pandas as pd
import numpy as np

data={'date':['2024-01-01','2024-02-15','2024-03-10']}
df=pd.DataFrame(data)
df['date']=pd.to_datetime(df['date'])
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
df['day']=df['date'].dt.day
df
```
