```python
'''Create a column:
Product = ["Phone-123", "Laptop-456", "Tablet-789"]
Tasks:
• Remove numbers using str.replace()
• Keep only product names'''
import pandas as pd
import numpy as np

data={'product' :["Phone-123", "Laptop-456", "Tablet-789"]}
df=pd.DataFrame(data)
df['product']=df['product'].str.replace(r'-\d+','',regex=True)   # here r'-\d+' is a regex and r'- here - is bcz phone-123 so -.
df
```

```python
''' SOME COMMON REGEX :---
| Pattern | Meaning                   | Example Match |
| ------- | ------------------------- | ------------- |
| `\d`    | One digit                 | `5`           |
| `\d+`   | One or more digits        | `123`         |
| `\w`    | Letter, digit, underscore | `A`, `x`, `7` |
| `.`     | Any character             | `a`, `#`, `9` |
| `+`     | One or more               | `aaa`         |
| `*`     | Zero or more              | `aaa`, empty  |
| `^`     | Start of string           | `^Phone`      |
| `$`     | End of string             | `123$`        |
```

```python
'''Create a dataset:
Start_Date = ["2023-01-01", "2023-06-01"]
End_Date = ["2024-01-01", "2024-06-01"]
Tasks:
• Convert both columns to datetime
• Calculate tenure in days'''

import pandas as pd

data = {
    'Start_Date': ['2023-01-01', '2023-06-01'],
    'End_Date': ['2024-01-01', '2024-06-01']
}

df = pd.DataFrame(data)

# Convert to datetime
df['Start_Date'] = pd.to_datetime(df['Start_Date'])
df['End_Date'] = pd.to_datetime(df['End_Date'])

# Calculate tenure in days
df['Tenure_Days'] = (df['End_Date'] - df['Start_Date']).dt.days

df
```

```python
'''Create a dataset:
Customer, Category, Start_Date, End_Date
Include:
• Mixed case category values
• Missing values in Category
• Date columns as strings
Tasks:
1 Clean Category column
• Convert to lowercase
• Replace missing values with "unknown"
2 Convert Start_Date and End_Date to datetime
3 Calculate tenure in days
4 Extract Year and Month from Start_Date
5 Filter customers with tenure > 200 days
6 Identify rows where Category contains "electronics"'''
```

```python
import pandas as pd
import numpy as np

data={'customer':['DeVaNSHu','RahUL','viJAy',np.nan],
     'category':['premium',np.nan,'basic','deluxe'],
     'start_date':['2024-09-17','2023-09-22','2022-12-20','2024-08-09'],
     'end_date':['2025-03-01','2028-03-30','2027-10-22','2030-07-19']}

# Convert to lowercase:-
df=pd.DataFrame(data)
df['customer']=df['customer'].str.lower()
df

# Replace missing values with "unknown"
df[['category','customer']]=df[['category','customer']].fillna("Unknown")
df

# Convert Start_Date and End_Date to datetime
df['start_date'] = pd.to_datetime(df['start_date'])   # when told to convert to datetime convert each row seperately.
df['end_date'] = pd.to_datetime(df['end_date'])
df

# Calculate tenure in days
df['tenue']=(df['end_date']-df['start_date']).dt.days
df

# Extract Year and Month from Start_Date
df['year']=df['start_date'].dt.year
df['months']=df['start_date'].dt.month
df

# Extract Year and Month from Start_Date
df['years_in']=((df['end_date']-df['start_date']).dt.days/365).round(2)   # round is used to round_off after decimal.
df['months_in']=((df['end_date']-df['start_date']).dt.days/30).round(1)
df

# Filter customers with tenure > 200 days
df=df[df['tenue']>200]
df

# identify rows where categories contains=='deluxe'
df=df[df['category']=='deluxe']
df
```
