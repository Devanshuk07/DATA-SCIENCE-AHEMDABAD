```python
'''GROUPBY:--
-It is used to group rows having the same value and then perform calculations on each group.

-Think of it as:

Split → Apply → Combine
Split data into groups
Apply a calculation (sum, mean, count, max, etc.)
Combine the results

EXAMPLE:-
A COLUMN HAS 3 DEPARTTMENT
THEN WE CAN GIVE GROUP BY DEPARTMENT(HOSPITALITY)
THEN WE WILL GIVE AGGREGATION FUNCTION LIKE==MIN,SUM,MAX,MEAN,MODE,STD ETC.

SYNTAX:-
df.groupby('City')['Sales'].agg(['sum','mean','max'])
```

```python
'''THIS ALL ARE AGGREGATION FUNCTIONS:--

| Function
| --------|
| `sum()`  
| `mean()` 
| `max()`  
| `min()`  
| `count()`
| 'MEDIAN()'
| 'STD()'
| 'COUNT()':--Counts non-null values.

'''
```

```python
import pandas as pd

df = pd.DataFrame({
    'City':['Rajkot','Rajkot','Ahmedabad','Ahmedabad'],
    'Sales':[100,200,300,400]
})

df.groupby('City')['Sales'].sum()
```

```python
# GROUP BY MULTIPLE COLUMNS:-
df = pd.DataFrame({
    'City':['Rajkot','Rajkot','Rajkot'],
    'Product':['Mobile','Laptop','Mobile'],
    'Sales':[100,200,150]
})
df.groupby(['City','Product'])['Sales'].sum()
```

```python
'''Why is groupby() Useful?

Suppose a company asks:

Total sales by city
Average salary by department
Number of orders by customer
Maximum marks by student

All of these are solved using groupby().'''
```
