```python
'''Region = ["North","South","North","East"] 
Revenue = [100, 200, 150, 300] 
Tasks:
Group by Region
Calculate total revenue '''
import pandas as pd

d1=pd.DataFrame({'region':['north','south','north','east'],
                'revenue':[100,200,150,300]})
print(d1)
a=d1.groupby('region')['revenue'].sum()
print(a)
```

```python
'''Using the same dataset: 
Find average revenue per Region'''
import pandas as pd

d1=pd.DataFrame({'region':['north','south','north','east','south','east','east'],
                'revenue':[100,200,150,300,450,300,290]})
print(d1)
a=d1.groupby('region')['revenue'].mean()
print('\n AVERAGE :-',a)
```

```python
'''Create a dataset: 
Category = ["A","A","B","B", "A"] 
Revenue = [100,200,150,250,300] 
Tasks: 
Group by Category
Find total revenue'''
import pandas as pd
d1=pd.DataFrame({'category':['A','A','B','B','A'],
                 'revenue':[100,200,150,250,300]
})
a=d1.groupby('category')['revenue'].sum()
print(a)
```

```python
'''Using the same dataset: 
• Find maximum revenue per Category 
• Find minimum revenue per Category'''
import pandas as pd
d1=pd.DataFrame({'category':['A','A','B','B','A'],
                 'revenue':[100,200,150,250,300]
})
a=d1.groupby('category')['revenue'].max()
print(a)
a=d1.groupby('category')['revenue'].min()
print(a)
```

```python
'''Create a dataset: 
Region = ["North","South", "North","South","East"] 
Revenue = [100,200,150,250,300] 
Tasks:
Count number of rows per Region'''
d1=pd.DataFrame({'region':['north','south','north','south','east'],
                 'revenue':[100,200,150,250,300]})
a=d1.groupby('region').size()
print(a)
```

```python
'''Create a dataset:
Region ["North","South","North","East"] 
Category = ["A","A","B","B"] 
Revenue = [100,200,150,300] 
Tasks:
Group by Region and Category
Calculate total revenue
'''
d1=pd.DataFrame({'region':["North","South","North","East"],
                 'category':["A","A","B","B"],
                 'Revenue':[100,200,150,300]})
a=d1.groupby(['region','category'])['Revenue'].sum()
print(a)
```

```python
'''Using the same dataset:
Calculate mean revenue for each Region'''
d1=pd.DataFrame({'region':["North","South","North","East"],
                 'category':["A","A","B","B"],
                 'Revenue':[100,200,150,300]})
a=d1.groupby('region')['Revenue'].mean()
print(a)
```

```python
'''Create a dataset: 
Region = ["North","South","North","East"] 
Revenue = [100,200,150,300] 
Tasks:
Use groupby with as_index=False
Compare output with default groupby'''
d1=pd.DataFrame({'Region':["North","South","North","East"],
                 'Revenue':[100,200,150,300]})
a=d1.groupby('Region',as_index=False)['Revenue'].sum()
print(a)
```

```python
'''Create a dataset: 
Region ["North","South","North","East"] 
Revenue = [100,200,150,300] 
Tasks: 
Sort groups in descending order of total revenue 
'''
d1=pd.DataFrame({'Region':["North","South","North","East"],
                 'Revenue':[2000,1300,1500,1200]})
a=d1.groupby('Region',as_index=False)['Revenue'].sum()
a=a.sort_values(by='Revenue',ascending=False)
print(a)
```

```python
'''Create a dataset: 
Region = ["North","South","North","South","East"]
Revenue = [100,200,150,250,300] 
Tasks:
Use agg() to calculate sum and mean'''
d1=pd.DataFrame({'region':["North","South","North","South","East"],
                 'revenue':[100,200,150,250,300]})
a=d1.groupby('region')['revenue'].agg(['sum','mean'])
print(a)
```

```python
'''Using the same dataset: 
Rename aggregated columns using named aggregation'''
d1=pd.DataFrame({'region':["North","South","North","South","East"],
                 'revenue':[100,200,150,250,300]})
a=d1.groupby('region',as_index=False).agg(total_revenue=('revenue','sum'),average_revenue=('revenue','mean'))
print(a)
```

```python
'''Create a dataset: 
Region = ["North","South","North","South","East"] 
Category = ["A","A","B","B","A"] 
Revenue = [100,200,150,250,300]
Group by Region and Category 
Find total revenue
Reset index'''

d1=pd.DataFrame({'region':["North","South",'South','South',"North","South","East"],
                 'category':["A","A",'A','A',"B","B","A"],
                 'revenue':[100,200,130,380,150,250,300]})
a=d1.groupby(['region','category'],as_index=False)['revenue'].sum()
print(a)
```

```python
'''Create a dataset: 
Region = ["North","South","North","East"] 
Revenue = [100,200,150,300] 
Tasks:
Use groupby and find count of Revenue values'''
d1=pd.DataFrame({'region':["North","South","North","South","South","East","North","North"],
                 'revenue':[100,200,150,120,110,300,400,120]})
a=d1.groupby('region')['revenue'].size()
print(a)
```

```python
'''Create a dataset:
Region = ["North", "South", "North","East"] 
Revenue = [100,200,150,300] 
Tasks: 
Find standard deviation of revenue per Region'''
d1=pd.DataFrame({'region':["North", "South", "North","East"],
                'revenue':[100,200,150,300]})
a=d1.groupby('region')['revenue'].std()
print(a)
```
