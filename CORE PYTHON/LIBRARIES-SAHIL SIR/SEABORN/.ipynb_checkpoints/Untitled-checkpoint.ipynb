```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset('tips')
sns.regplot(x='total_bill',y='tip',data=df)  # regplot==> used to find linear regration.

plt.title('regplot example')
plt.show()
```
