```python
'''import seaborn as sp

MEMORY RULE:----

| Plot            | Data Type                             |
| --------------- | ------------------------------------- |
| `catplot()`     | Category + Numeric (or Category only) |
| `countplot()`   | Category only                         |
| `histplot()`    | Numeric only                          |
| `kdeplot()`     | Numeric only                          |
| `scatterplot()` | Numeric + Numeric                     |
| `relplot()`     | Numeric relationships                 |
| `boxplot()`     | Category + Numeric                    |
| `violinplot()`  | Category + Numeric                    |
'''
```

```python
# HISTPLOT IS USED IN SEABORN:--

import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sp

delivry=np.random.randint(15,61,50)
print(delivry)

df=pd.DataFrame({'delivry_time':delivry})
sp.histplot(data=df,x='delivry_time',bins=10)    # histplot is for seaborn syntax :--sp.histplot(data=None, x=None, bins=None)
mp.show()

'''histplot() (Histogram Plot) is used to show the distribution of numerical data.

It answers questions like:

How are delivery times distributed?
What score range occurs most often?
How are ages spread out?
What is the distribution of daily steps?'''
```

```python
'''Load the 'tips' dataset from Seaborn and create a boxplot to visualize the distribution of total bill amounts by day of the week,
similar to how Zomato might analyze spending patterns across weekdays.'''

# WE ARE USING HERE A BOXPLOT THAT IS USED IN SEABORN.

tips=sp.load_dataset('tips')    # SEABORN HAS  A BUILT-IN DATASET CALLED TIPS WHICH IS USED FOR JUST TEACHING PURPOSE.
print(tips)

sp.boxplot(data=tips,x='day',y='total_bill')
mp.show()


'''A boxplot is used to summarize the distribution of numerical data and detect outliers.

It shows 5 important values:

Minimum (smallest value, excluding outliers)
Q1 (25th percentile)
Median (Q2) (50th percentile)
Q3 (75th percentile)
Maximum (largest value, excluding outliers)


HERE AS YOU CAN SEE THE BLUE BOX HAS UPPER,MIDDLE AND LOWER LINE THAT MEANS UPPER IS FOR UPPER LIMIT FOR Q3 AND LOWER FOR LOWER LIMIT MEANS Q1 
AND MIDDLE IS FOR Q2 THAT IS MEDIAN.SO YOU CAN SAY THAT IF Q2 IS TOWARD LOWER SIDE JUST IN FRI(FIGURE)THEN THE THAT MEANS VALUES OF 10.5 IS MORE THAN LESS THAN 10.5.'''
```

```python
'''Simulate IPL match scores for 8 teams and use a violinplot to compare the run distributions per team.
Style your plot using the 'darkgrid' Seaborn theme.'''

# THIS IS EXAMPLE OF VIOLINPLOT:--

teams = ['CSK', 'MI', 'RCB', 'KKR', 'SRH', 'RR', 'DC', 'PBKS']

df = pd.DataFrame({
    'Team': np.repeat(teams, 20),
    'Runs': np.random.randint(120, 220, 160)
})

sp.set_theme(style='darkgrid')
sp.violinplot(data=df, x='Team', y='Runs')
mp.show()

'''violin plot is used to compare the distribution of numerical data across different categories.

You can think of it as:

Boxplot + KDE plot combined

It shows:

Median (middle value)
Spread of the data
Shape of the distribution (where values are concentrated)'''
```

```python
'''Create a countplot that displays the number of songs per genre from a list of 40 Spotify tracks, with at least 4 different genres represented.
<br><br><em><strong>Hint:</strong> Use a Python list or pandas DataFrame to store your data, then plot with Seaborn.</em>'''

# THIS IS EXAMPLE OF COUNTPLOT:--

genres=(['Pop']*12+
        ['Hip-Hop']*10+
        ['Jazz']*10+
        ['Rock']*15)

df=pd.DataFrame({'Genre':genres})
sp.countplot(data=df,x='Genre')
mp.show()

'''countplot() is used to count how many times each category appears in your data and display those counts as bars.'''
```

```python
'''Use Seaborn's kdeplot to visualize the distribution of daily step counts for a week, as if analyzing data from a fitness app.
Apply the 'whitegrid' theme and customize the plot color.'''

# THIS IS EXMAPLE OF KDEPLOT:--

step=np.random.randint(600,1200,7)
weeks=['Monday', 'Tuesday', 'Wednesday', 'Thursday',
     'Friday', 'Saturday', 'Sunday']
print(step)
df=pd.DataFrame({'Steps':step,
                'Week':weeks})
sp.set_theme(style='whitegrid')
sp.kdeplot(data=df,x='Steps',fill=True,color='Grey')   # you're creating a 1-dimensional KDE plot. NO Y HERE ,,,It only needs one numeric variable (Steps) to show its distribution.
mp.show()

# WITHOUT FILL YOU WILL ONLY GET CURVE TO FILL INSIDE THE AREA OF CURVE USE FILL=TRUE.
'''sp.kdeplot()
kde stands for Kernel Density Estimation.
It creates a smooth curve showing how your data is distributed.
Think of it as a smoother version of a histogram.'''
```

```python
'''Using the 'flights' dataset from Seaborn, generate a heatmap showing the correlation between months and years based on the number of passengers.
<br><br><em><strong>Hint:</strong> Use pivot_table to reshape the data before plotting the heatmap.</em>'''

# UISNG HEATMAP() :--

flt=sp.load_dataset('flights')
print(flt)

pivot=flt.pivot_table(
    values='passengers',
    index='month',
    columns='year')
sp.heatmap(pivot)  # (annot=True,fmt='.1%') here annot means in color fig use print numbers.fmt is for format means whioch format 'd' for decimal.1% is for percentage.
mp.show()

'''heatmap() requires a 2-dimensional grid of numbers, but the flights dataset is stored in a long format.
WHAT HEATMAP WANTS:--
| month | 1949 | 1950 | 1951 |
| ----- | ---- | ---- | ---- |
| Jan   | 112  | 115  | 145  |
| Feb   | 118  | 126  | 150  |
| Mar   | 132  | 141  | 178  |

SO WE HAVE TO CONVERT AND TAKE PIVOT_TABLE THEN CONVERT THAT SO THAT HEATMAP CAN USE IT'''
```

```python
'''Create a relplot using the 'fmri' dataset from Seaborn to visualize how the signal changes over time for different event types.'''

# EXAMPLE OF RELPLOT:-IT STANDS FOR RELATIONAL PLOT.

a=sp.load_dataset('fmri')
print(a)

sp.relplot(a)       # syntax:-(data=a,x='col1',y='col2')
mp.show()

'''It is a high-level Seaborn function used to visualize relationships between variables.'''
```

```python
'''Pick any categorical column from the 'titanic' dataset (like 'class' or 'sex') and use catplot to display the survival rate for each group 
with confidence intervals.'''

# EXAMPLE OF CATPLOT:- MEANS CATEGORICAL PLOT

a=sp.load_dataset('titanic')
print(a)

sp.catplot(data=a,x='sex',y='survived',kind='bar',label='titanic survied list')
mp.legend()
mp.show()

'''catplot() stands for Categorical Plot.

It is a high-level Seaborn function used to visualize relationships involving categorical variables.
Yes, for catplot() at least one variable should usually be categorical.

kind='bar'      # average values
kind='count'    # count categories
kind='box'      # quartiles & outliers
kind='violin'   # distribution shape
kind='strip'    # individual points
kind='swarm'    # spread-out points'''
```

```python
'''Use jointplot on the 'penguins' dataset to visualize the relationship between bill_length_mm and flipper_length_mm,
and fit a regression line on the scatter plot.<br><br><em><strong>Hint:</strong> Set kind='reg' in jointplot for regression line.</em>'''

# JOINTPLOT:---

a=sp.load_dataset('penguins')
print(a)

sp.jointplot(data=a,kind='reg',x='bill_length_mm',y='flipper_length_mm')
mp.show()

'''
jointplot() is used to analyze the relationship between two numeric variables while also showing the distribution of each variable.
jointplot() needs to know which two variables you want to compare.'''
```
