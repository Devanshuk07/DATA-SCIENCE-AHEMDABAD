```python
'''Use plt.subplots() to create a 2x2 grid of subplots and 
plot four different types of charts (line, bar, scatter, and pie) using any sample data of your choice.'''

import matplotlib.pyplot as mp
import pandas as pd

fig,ax=mp.subplots(2,2)

x=[1,2,3,4]
y=[10,20,15,25]

ax[0,0].plot(x,y)

ax[0,1].bar(x,y)

ax[1,0].scatter(x,y)

ax[1,1].pie(y)

mp.show()
```

```python
'''Plot a comparison of average delivery times for Zomato, Swiggy, and Domino's using a bar chart, and 
style each bar with a different color, linestyle, and linewidth using Matplotlib plot styling options.'''

x=['zomato','swiggy',"domino's",'pizza_hut','papa_louise']
y=[25,30,15,22,28]

mp.bar(x,y,color=['grey','red','yellow','purple','green'],linestyle='--',linewidth=['10','12','14','12','11'])
mp.show()
```

```python
'''Create a multi-axis chart that shows the number of Instagram followers (left y-axis) and average daily posts (right y-axis) for five influencers.
Use different colors and markers for each axis.'''

x=['123m','223m','213m','332m','413m']
y=[7,10,9,11,8]

mp.bar(x,y,color='grey',label='INSTAGRAM NETWORK')
mp.xlabel('FOLLOWERS')
mp.ylabel('DAILY POSTS')
mp.legend()
mp.show()
```

```python
'''Plot the number of tickets sold for five recent Bollywood movies (categorical) versus their IMDB ratings (numeric) using a scatter plot.
Add annotations to display the movie names above each point.'''

tic_sold=['2231m','623m','913m','232m','413m']
imdb=['9.1','8.1','3.4','6.9','7.1']
nums=[float(i)for i in imdb]   # simpleast way to convert string like 9.1 to float.
movie_name=['3-idiots','veer-zaara','mohabbatein','raaz','harry-potter']

mp.scatter(tic_sold,nums,color='green')
mp.plot(tic_sold,nums,linestyle='--',color='orange')

for i in range(len(tic_sold)):
    mp.annotate(movie_name[i],(tic_sold[i],nums[i]))
mp.show()
```

```python
'''Add a custom text annotation to a Matplotlib chart showing Flipkart's monthly sales, marking the highest sales point with the label 'Big Billion Days'.
<br><br><em><strong>Hint:</strong> Use the ax.annotate() function to place the label at the correct data point.</em>'''

sal=[231,623,913,232,413,2231,223,991]
mon=['jan','feb','mar','apr','may','sep','dec','jun']

mp.scatter(mon,sal,color='grey')
mp.plot(mon,sal,color='pink')
mp.annotate('big billion days',(mon[5],sal[5]))
mp.show()
```
