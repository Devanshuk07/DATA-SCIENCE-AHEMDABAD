```python
'''Given a list of delivery timestamps as strings (e.g., ['2024-06-01 14:30', '2024-06-02 09:15', '2024-06-03 20:45']),
use pandas and pd.to_datetime() to convert them into datetime objects and print the result.'''
import pandas as pd
deli=pd.array(['2024-06-01 14:30', '2024-06-02 09:15', '2024-06-03 20:45'])
deli=pd.to_datetime(deli)
print(deli)
```

```python
'''Load a CSV file containing order dates from a Flipkart-style order history (column: 'order_date', format: 'YYYY-MM-DD HH:MM:SS').
Extract the year, month, and weekday for each order and add them as new columns in the DataFrame.'''
import pandas as pd

df=pd.read_csv('users.csv')
df['order_date']=pd.to_datetime(df['order_date'])
df['year']=df['order_date'].dt.year
df['month']=df['order_date'].dt.month
df['weekday']=df['order_date'].dt.day_name()

print(df)
```

```python
'''Set the 'order_date' column as the index of your DataFrame and use resampling to calculate the total number of orders placed each week.
<br><br><em><strong>Hint:</strong> Use df.resample('W').size() after setting the datetime index.</em>'''

df=pd.read_csv('users.csv')
df['order_date']=pd.to_datetime(df['order_date'])
df=df.set_index('order_date')
print(df)
week_ord=df.resample('W').size()
print(week_ord.to_string())   # to display full csv file.
```

```python
'''Suppose you have a DataFrame of Instagram posts with a 'posted_at' column in UTC.
Convert these timestamps to 'Asia/Kolkata' timezone and display the first 5 converted times.'''
import pandas as pd

data = pd.DataFrame({
    'posted_at': [
        '2024-06-01 11:30:00',
        '2024-06-01 22:20:00',
        '2024-06-01 21:10:00',
        '2024-06-01 12:10:00',
        '2024-06-01 09:10:00'
    ]
})
print(data)
data['posted_at']=pd.to_datetime(data['posted_at']).dt.tz_localize('UTC')
data['posted_at']=data['posted_at'].dt.tz_convert('Asia/kolkata')
print(data)
```

```python
'''Create a new feature called 'is_weekend' in your DataFrame that marks True if an order was placed on Saturday or Sunday, and False otherwise.
<br><br><em><strong>Constraint:</strong> Do not use any external libraries except pandas and numpy.</em>'''

df=pd.read_csv('users.csv')
print(df)
df['order_date']=pd.to_datetime(df['order_date'])
df['weekend']=df['order_date'].dt.day_name()
print(df)
df['is_weekend']=(df['weekend'].isin(['Saturday','Sunday']))
print(df)
```
