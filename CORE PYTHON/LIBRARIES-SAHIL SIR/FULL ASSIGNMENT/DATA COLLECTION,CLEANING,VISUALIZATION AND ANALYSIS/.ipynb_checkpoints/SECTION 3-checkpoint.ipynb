```python
'''Simulate a Spotify-like 'Recommended Songs' feature: Given two 3x3 matrices representing user-song interaction scores (user preferences and song popularity),
use dot() and matmul() to compute the final recommendation matrix and explain the difference between the two results.'''
user_song=np.array([
    [5,4,2],
    [3,4,1],
    [2,2,5]])

song_pop=np.array([
    [3,2,5],
    [4,3,1],
    [4,4,5]])

dot=np.dot(user_song,song_pop)
print(dot)
matmul=np.matmul(user_song,song_pop)
print(matmul)

# FOR 2-D ARRAY THERE IS NO DIFFERENCE BETWEEN DOT AND MATMUL.Both perform standard matrix multiplication and produce the same result.
```

```python
'''Given a 4x4 NumPy matrix representing the pixel brightness of a small Instagram image, use transpose (T) to rotate the image and
then calculate the mean, median, standard deviation, and variance of the pixel values.'''

ins_img=np.array([
    [4,3,2,1],
    [4,5,6,7],
    [5,6,7,8],
    [8,9,9,1]])
trans=np.transpose(ins_img)
print(trans)
print('MEAN:--\n',np.mean(trans))
print('MEDIAN:--\n',np.median(trans))
print('STD:--\n',np.std(trans))
print('VARIANCE:--\n',np.var(trans))
```

```python
'''Take a 3x3 NumPy matrix representing a Zomato restaurant rating correlation grid and
use np.linalg.inv(), np.linalg.det(), and np.linalg.eig() to compute its inverse, determinant, and eigenvalues/eigenvectors.<br><br><em><strong>Hint:</strong>
If the matrix is not invertible, modify one value and try again.</em>'''

rat=np.array([
    [2,3,5],
    [4,3,1],
    [5,5,4]])

deter=np.linalg.det(rat)          # np.linalg stands for NumPy Linear Algebra.
print(deter)

inv=np.linalg.inv(rat)
print(inv)

eig=np.linalg.eig(rat)
print(eig)

'''
| Function          | Purpose                                                 |
| ----------------- | ------------------------------------------------------- |
| `np.linalg.det()` | Calculates the determinant of the matrix                |
| `np.linalg.inv()` | Calculates the inverse matrix (only if determinant ≠ 0) |
| `np.linalg.eig()` | Returns eigenvalues and eigenvectors                    |
'''
```
