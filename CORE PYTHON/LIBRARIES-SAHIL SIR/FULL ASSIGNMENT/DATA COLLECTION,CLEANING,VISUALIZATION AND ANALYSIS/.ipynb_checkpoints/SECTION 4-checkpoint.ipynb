```python
'''Download a large Excel file (at least 10,000 rows) of Flipkart product listings.
Use pd.read_excel() with the chunksize parameter to read the file in chunks of 2000 rows, and print the number of rows in each chunk as you iterate.'''

with open('flipkart.csv','r')as f:
    data=pd.read_csv(f)

for chunk in pd.read_csv('flipkart.csv', chunksize=2000):  #chunk==contains rows 1–2000  Second iteration contains rows 2001–4000.
    print(len(chunk))

'''Chunk 1: 2000 rows
Chunk 2: 2000 rows
Chunk 3: 2000 rows
Chunk 4: 2000 rows
Chunk 5: 2000 rows
total=2000*5=10,000'''
```
