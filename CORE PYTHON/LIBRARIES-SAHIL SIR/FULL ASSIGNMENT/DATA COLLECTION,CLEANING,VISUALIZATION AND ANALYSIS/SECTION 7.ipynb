```python
'''Given a CSV file containing Zomato restaurant ratings,
use pandas to detect outliers in the 'user_rating' column using the IQR method and print the indices of the detected outliers.'''
import pandas as pd

with open('zomato.csv','r')as f:
    data=pd.read_csv(f)

data

q1=data['Rating'].quantile(0.25)
q3=data['Rating'].quantile(0.75)

iqr=q3-q1

lower=q1-1.5*iqr
upper=q3+1.5*iqr

outliers=data[(data['Rating']<lower)|(data['Rating']>upper)]

print(outliers)
```

```python
'''Apply winsorization to the 'transaction_amount' column in a Paytm transactions DataFrame to cap all values above the 95th percentile and 
below the 5th percentile, then display the updated column statistics.'''

import pandas as pd

# Load dataset
df = pd.read_csv('ptm.csv')

# Calculate percentiles
lower = df['transaction_amount'].quantile(0.05)
upper = df['transaction_amount'].quantile(0.95)
print('lower:-',lower)
print('upper:-',upper)
# Apply winsorization
df['transaction_amount'] = df['transaction_amount'].clip(lower=lower, upper=upper)   #Any value less than lower becomes lower.Any value greater than upper becomes upper.
print(df)
```

```python
'''You have a DataFrame of Flipkart product prices stored as strings with currency symbols (e.g., '₹1,299').
Convert this column to numeric type using pandas, ensuring all non-numeric characters are removed.<br><br><em><strong>Hint:</strong>
Use str.replace() and astype().</em>'''
import pandas as pd

df=pd.DataFrame({'products':['iphone','mixer','airpods','imac'],
                 'prices':['₹70,000','₹12,000','₹13,299','₹122,299']})

df['prices']=df['prices'].str.replace('₹','',regex=False)\   # Regex (Regular Expression) is a pattern used to search, match, or replace text.
                         .str.replace(',','',regex=False)\
                         .astype(int)       # astype() is a pandas method used to change the data type of a column or Series.
print(df)
```

```python
'''Fix the following code snippet where the 'is_premium' column in a Spotify user DataFrame is a mix of boolean, string, and integer types.
Convert the entire column to boolean type, treating 'True', 1, and 'yes' as True, and everything else as False.'''

a=pd.read_json('spotify.json')
b=pd.DataFrame(a)
print(b)

b['is_premium']=b['is_premium'].replace({
    True:True,
    False:False,
    1:True,
    0:False,
    'yes':True,
    'no':False
}).astype(bool)
print(b)
```
