```python
'''Using Matplotlib, create a scatter plot of 10 restaurants from your city with their Zomato ratings on the x-axis and average meal price on the y-axis.
Add axis labels and a title to the chart.'''

x=[4.5,3.2,3.1,4.4,4,5,4.9,5,2,1.9]
y=[200,120,300,290,280,220,310,120,290,300]

mp.scatter(x,y,color='red')
mp.plot(x,y)
mp.xlabel('AVG RATING')
mp.ylabel('AVERAGE PRIZE')
```

```python
'''Build a bar chart that shows the number of orders you or your friends placed on Swiggy, Zomato, and Domino’s in the last month.
Use different colors for each bar and add a legend.'''

import matplotlib.pyplot as mp

platforms = ['Swiggy', 'Zomato', "Domino's"]
orders = [12, 8, 5]
colors = ['red', 'orange', 'blue']

mp.bar(platforms, orders, color=colors)

mp.xlabel('Food Delivery Platform')
mp.ylabel('Number of Orders')
mp.title('Orders Placed Last Month')

mp.show()
```

```python
'''Customize a Matplotlib figure by creating a plot with two subplots: one line plot showing your daily Instagram screen time for a week,
and one bar chart showing the number of posts you liked each day.
Add appropriate titles, axis labels, and save the figure as social_media_usage.png.<br><br><em><strong>Hint:</strong>
Use plt.subplots() to create multiple axes in one figure.</em>'''

screen={2:20,3:40,4:20,5:50,7:10,1:40,6:19}   # here it is like 1:40 so day-1 and 40 mins. like that here taken.
likes=[200,300,120,420,455,520,320]

days=list(screen.keys())
screen_time=list(screen.values())
fig,ax=mp.subplots(1,2,figsize=(10,4))

ax[0].plot(days,screen_time,marker='o')
ax[0].set_title('instagram screen time')

ax[1].bar(days,likes)
ax[1].set_title('instagram likes')

mp.savefig('insta_screen_likes.png')
mp.show()
```
