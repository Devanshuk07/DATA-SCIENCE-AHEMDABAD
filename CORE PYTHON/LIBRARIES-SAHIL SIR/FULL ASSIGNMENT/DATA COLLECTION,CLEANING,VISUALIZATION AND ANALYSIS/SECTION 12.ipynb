```python
'''Install Seaborn in your Python environment and use it to create a histplot showing the distribution of delivery times (in minutes) for
50 Zomato food orders. Generate random data if needed.'''
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sp

delivry=np.random.randint(15,61,50)
print(delivry)

df=pd.DataFrame({'delivry_time':delivry})
sp.histplot(data=df,x='delivry_time',bins=10)    # histplot is for seaborn syntax :--sp.histplot(data=None, x=None, bins=None)
mp.show()
```

```python
'''Load the 'tips' dataset from Seaborn and create a boxplot to visualize the distribution of total bill amounts by day of the week,
similar to how Zomato might analyze spending patterns across weekdays.'''

tips=sp.load_dataset('tips')
print(tips)

sp.boxplot(data=tips,x='day',y='total_bill')
mp.show()
```

```python
'''Simulate IPL match scores for 8 teams and use a violinplot to compare the run distributions per team.
Style your plot using the 'darkgrid' Seaborn theme.'''
teams = ['CSK', 'MI', 'RCB', 'KKR', 'SRH', 'RR', 'DC', 'PBKS']

df = pd.DataFrame({
    'Team': np.repeat(teams, 20),
    'Runs': np.random.randint(120, 220, 160)
})

sp.set_theme(style='darkgrid')
sp.violinplot(data=df, x='Team', y='Runs')
mp.show()
```

```python
'''Create a countplot that displays the number of songs per genre from a list of 40 Spotify tracks, with at least 4 different genres represented.
<br><br><em><strong>Hint:</strong> Use a Python list or pandas DataFrame to store your data, then plot with Seaborn.</em>'''

genres=(['Pop']*12+
        ['Hip-Hop']*10+
        ['Jazz']*10+
        ['Rock']*15)

df=pd.DataFrame({'Genre':genres})
sp.countplot(data=df,x='Genre')
mp.show()
```

```python
'''Use Seaborn's kdeplot to visualize the distribution of daily step counts for a week, as if analyzing data from a fitness app.
Apply the 'whitegrid' theme and customize the plot color.'''
step=np.random.randint(600,1200,7)
weeks=['Monday', 'Tuesday', 'Wednesday', 'Thursday',
     'Friday', 'Saturday', 'Sunday']
print(step)
df=pd.DataFrame({'Steps':step,
                'Week':weeks})
sp.set_theme(style='whitegrid')
sp.kdeplot(data=df,x='Steps',fill=True,color='Grey')  # you're creating a 1-dimensional KDE plot. NO Y HERE ,,,It only needs one numeric variable (Steps) to show its distribution.
mp.show()

# WITHOUT FILL YOU WILL ONLY GET CURVE TO FILL INSIDE THE AREA OF CURVE USE FILL=TRUE.
```
