```python
'''Load the 'tips' dataset from Seaborn and create a pairplot to quickly visualize relationships between all numeric variables.'''

import pandas as pd
import numpy as np
import seaborn as sp
import matplotlib.pyplot as mp

load=sp.load_dataset('tips')
print(load)

sp.pairplot(load)
mp.show()
```

```python
'''Using the 'flights' dataset from Seaborn, generate a heatmap showing the correlation between months and years based on the number of passengers.
<br><br><em><strong>Hint:</strong> Use pivot_table to reshape the data before plotting the heatmap.</em>'''

flt=sp.load_dataset('flights')
print(flt)

pivot=flt.pivot_table(
    values='passengers',
    index='month',
    columns='year')
sp.heatmap(pivot)
mp.show()
```

```python
'''Create a relplot using the 'fmri' dataset from Seaborn to visualize how the signal changes over time for different event types.'''

a=sp.load_dataset('fmri')
print(a)

sp.relplot(a)
mp.show()
```

```python
'''Pick any categorical column from the 'titanic' dataset (like 'class' or 'sex') and use catplot to display the survival rate for each group 
with confidence intervals.'''

a=sp.load_dataset('titanic')
print(a)

sp.catplot(data=a,x='sex',y='survived',kind='bar',label='titanic survied list')
mp.legend()
mp.show()
```

```python
'''Use jointplot on the 'penguins' dataset to visualize the relationship between bill_length_mm and flipper_length_mm,
and fit a regression line on the scatter plot.<br><br><em><strong>Hint:</strong> Set kind='reg' in jointplot for regression line.</em>'''

a=sp.load_dataset('penguins')
print(a)

sp.jointplot(data=a,kind='reg',x='bill_length_mm',y='flipper_length_mm')
mp.show()

```
