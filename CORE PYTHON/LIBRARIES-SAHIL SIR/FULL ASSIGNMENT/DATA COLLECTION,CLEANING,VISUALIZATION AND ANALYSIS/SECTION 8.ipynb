```python
'''Create a DataFrame with mock data for Spotify playlists, including playlist names and artist usernames, where some rows are exact duplicates.
Use drop_duplicates() to remove duplicate playlists and print the cleaned DataFrame.'''
spo=pd.read_json('spotify.json')
df=pd.DataFrame(spo)
print(df)
df_clean=df.drop_duplicates(subset=['artist'])
print('cleaned dataframe :---\n\n',df_clean)
```

```python
'''You have a DataFrame column for payment status from a Paytm-like app with mixed values: 'Yes', 'yes', 'Y', 'No', 'no', 'N', and some with extra spaces.
Write code to unify this column so all paid statuses become 1 and all unpaid statuses become 0, trimming whitespace and fixing capitalization where needed.
<br><br><em><strong>Hint:</strong> Use str.strip(), str.lower(), and map/replace methods in pandas.</em>'''

import pandas as pd

df = pd.DataFrame({
    'payment_status': [
        'Yes', ' yes ', 'Y',
        'No', 'no ', ' N ',
        'Yes ', ' y ', 'No'
    ]
})
df['payment_status']=df['payment_status'].str.strip().str.lower().map({    # map():-Yes, map() and replace() are similar, but they are used in slightly different situations.map()Used mainly on a single Series (column) to transform values.
    'yes':1,
    'yes':1,
    'y':1,
    'no':0,
    'no':0,
    'n':0
})
print(df)
```
