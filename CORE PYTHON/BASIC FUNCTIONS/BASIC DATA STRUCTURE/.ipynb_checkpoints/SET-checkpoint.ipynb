```python
'''SET:-(UNIQUE ITEMS/ELEMENTS)-->{}--> ALSO FOR DICTIONARY.
MUTABLE:-CAN BE CHANGED.
NO INDEXING
UNORDERED
NO DUPLICATE VALUES ALLOWED
-COMPREHENSIVE/COMPREHENSION..'''
```

```python
#SET:-{}
#a={}-->by default dictionary.
a={1,2,3,3,2,1,4,5,7,10,9}       #ALWAYS TAKE UNIQUE.
print(type(a))
print(a)
print(len(a))             #LENGTH ALSO UNIQUE.
```

```python
a={'harmi','Dharmistha','dharmistha','harmi'}
print(a)
```

```python
a=[1,22,33,22,44,55,77]
b=set(a)             #CONVERTED A-LIST INTO A-SET.
print(b)
```

```python
#SET UNION:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a.union(b)
print(c)
```

```python
#SET UNION:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a | b                 # | -->UNION.
print(c)
```

```python
#SET INTERSECTION:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a.intersection(b)
print(c)
```

```python
#SET INTERSECTION:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a&b            #INTERSECTION:- &
print(c)
```

```python
#SET DIFFERENCE:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a.difference(b)     #HERE A-B HAPPENS.
d=b.difference(a)     #HERE B-A HAPPENS.
print(c)
print(d)
```

```python
#SET DIFFERENCE:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a-b
d=b-a
print(c)
print(d)
```
