```python
#1. Print half of String in Uppercase:-
a=str(input('ENTER THE STRING :-'))
half= len(a)//2      #LET A=PYTHON SO LEN OF A IS 6 AND 6//2 MEANS=3 SO HALF IS STORED
result= a[:half].upper() + a[half:].lower()   #a[:half]--This means take characters from index 0 up to half(not including half).a[:half]--This means take characters from index half to the end.
print(result)
```

```python
#6. Count Even and Odd Numbers from given range.
a=int(input('ENTER THE LAST NUMBER OF RANGE:-'))
count=0
count1=0
for i in range(1,a+1):
    if i%2==0:
        count=count+1
    else:
        count1=count1+1
print('EVEN NUMBERS ARE :-',count)
print('ODD NUMBERS ARE:-',count1)
```

```python
#7.Convert a month name to a number of days. with if else and match case.
a=str(input('ENTER THE NAME OF MONTH:-'))
match a:
    case 'january'| 'jan':print(31)
    case 'february'|'feb':print(28)
    case 'march'|'mar':print(31)
    case 'april'|'apr':print(30)
    case 'may':print(31)
    case 'june'|'jun':print(30)
    case 'july'|'jul':print(31)
    case 'august'|'aug':print(31)
    case 'september'|'sep':print(30)
    case 'octomber'|'oct':print(31)
    case 'november'|'nov':print(30)
    case 'december'|'dec':print(31)
    case _:print('INVALID ENTRY')
```

```python
#8.Count no of even and odd digits from given number.
a=int(input('ENTER THE NUMBER:-'))
count=0
count1=0
while a>0:     # LET a=1234 , FOR SECOND LOOP a=123, THIRD LOOP a=12, FOURTH LOOP a=1.
    b=a%10     # A%10=1234%10==4
    if b%2==0: #4%2==0
        count=count+1    # increase a counter.
    else:                # 3%2==1
        count1=count1+1  # increase a counter.
    a=a//10       # THIS WILL REMOVE THE LAST DIGIT. 1234//10==123,,, THEN AGAIN a=123//10==12. LIKE THIS.
print('EVEN NUMBERS:-',count)
print('ODD NUMBERS:-',count1)
```

```python
#9.Convert temperature, degrees Fahrenheit to degrees Celsius and vice versa. C = (5/9) * (F - 32) &  F = C * 9/5 + 32.
a=int(input('ENTER THE TEMPERATURE :-'))
b=str(input('ENTER THE C AND F FOR CONVERT TEMPERATURE INTO CESIUS AND FAHRENHEIT:-')).lower()
if b=='c':
    c=(5/9)*(a-32)
    print('TEMPERATURE IN CELSIUS IS :-',c)
elif b=='f':
    f=a*9/5+32
    print('TEMPERATURE IN FAHRENHEIT IS :-',f)
```

```python
'''1. Write a program to sort a list of strings by their length e.g 
['apple', 'bananas', 'kiwi', 'orange'] output : ['kiwi', 'apple', 'orange', 'bananas']'''
a=['apple', 'bananas', 'kiwi', 'orange']
srt=sorted(a, key=len)    #sorted() is a built-in Python function from any iterable (like a list, tuple, string).key==a function that tells Python what value to use for sorting.
print(srt)                #key=len-->“Use the length of each string as the value for sorting.”
```
