```python
'''Take a number n from user
👉 Print the sum of only even numbers from 1 to n
👉 Do it using a for loop'''
a=int(input('ENTER THE NUMBER TILL WHICH YOU WANT EVEN NUMBER SUM :-'))
x=0
for i in range(2,a+1,2):    #range() does NOT include the last number.
    x+=i
    print(i)
print("SUM OF NUMBERS IS :-",x)
```

```python
#TAKE A NUMBER FROM USER AND CHECK WEATHER IT IS PRIME NUMBER OR NOT (USE FOR LOOP):-
a = int(input("ENTER NUMBER :- "))
if a<=1:   # THIS IS FOR 0 , 1 AS BOTH ARE NOT PRIME.
    print("NOT PRIME")
else:
    for i in range(2, a):
        if a % i == 0:
            print("NOT Prime")
            break
    else:
        print("Prime")
```

```python
#OTHER METHOD TO FIND PRIME NUMBER(PROFESSIONAL METHOD)
# BOTH ARE SAME BUT ITS HAS (2, INT(A**0.5)+1
a = int(input("ENTER NUMBER :- "))
if a <= 1:
    print("NOT Prime")
else:
    for i in range(2, int(a**0.5) + 1):  # HERE (2, INT(SQUARE ROOT OF A=0.5)+1)--INT MEANS LET A=29 THEN 0.5 SQUARE ROOT TO 29==5.38 THEN IN INT 5.38==5 THEN +1 =6.
        if a % i == 0:                   # SO THAT IS (2,6) WHICH IS BASIC LOGIC FOR PRIME NUMBER TO CHECK FROM 2 TO 6 EVERY NUMBER.
            print("NOT Prime")
            break
    else:
        print("Prime")
```

```python
#Take a number from user.Print the next prime number after that number:-
a = int(input("ENTER NUMBER :- "))
b=a+1     # START CHECKING AFTER GIVEN NUMBER.
while True:   # INFINITE LOOP UNTILL WE FIND PRIME NUMBER.
    for i in range(2,int(b**0.5)+1):
        if b%i==0:
            break  # NOT PRIME STOPS CHECKING THIS NUMBER.
    else:
        print("NEXT PRIME NUMBER IS :-",b)
        break  # STOP THE WHILE LOOP
    b+=1
    
```

```python
#PRINT EVERY PRIME NUMBER TILL USER ENTER THE NTH NUMBER:-
num= int(input("ENTER NUMBER :- "))
for n in range(2,num+1):
    for i in range(2, int(n**0.5)+1):  # here n=2 then i is (2,2) so range becomes(2,2) that is never executed.
        if n%i==0:     # if 2%2 never runs.So else print(n) will be printed.
            break
    else:
        print(n)
        
```
