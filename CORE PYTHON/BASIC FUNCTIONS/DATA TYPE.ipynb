```python
#ARITHMETIC OPERATORS:-
number1=int(input("ENTER NUMBER 1 :"))
number2=int(input("ENTER NUMBER 2 :"))
ans=number1+number2
print(f"SUBSTRACTION OF {number1} and {number2} is {number1-number2}")
print(f"MULTIPLICATION OF {number1} and {number2} is {number1*number2}")
print(f"DIVISION OF {number1} and {number2} is {number1/number2}")
```

```python
# FLOOR DIVISION:-(no value after point)
number1=int(input("ENTER NUMBER 1 :"))
number2=int(input("ENTER NUMBER 2 :"))
ans=number1+number2
print(f"FLOOR DIVISION OF {number1} and {number2} is {number1//number2}")   # NO VALUES AFTER POINT
print(f"DIVISION OF {number1} and {number2} is {number1/number2}")           # NORMAL DIVISION
```
