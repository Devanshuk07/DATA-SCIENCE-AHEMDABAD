```python
'''IF WE WRITE NAME="TOPS TECHNOLOGIES" -->THIS IS STRING.
IF WE WRITE INT=12345 THIS IS INTEGER.
STRING IS IMMUTABLE(YOU CANNOT CHANGE IT).
YOU CAN USE IN BUILT-->METHOD
INDEXING.
SLICING.'''
```

```python
'''INDEX ALWAYS DISPLAY BY [] BRACKETS.
ALWAYS START FROM [0] AND END UPTO.

SLICING-->PARTITION
2-6 LETTERS
UPTO 6 LETTERS.
LAST 4 LETTERS.

SLICING NAME[START:END:STEP]
```

```python
#PRINT ALTERNATE LETTERS FROM A STRING:-
name=str(input('ENTER THE NAME :-'))
for i in range(0,len(name),2):   #start=0 , end= len(name) , step=2.
    print(name[i])
```

```python
name=str(input('ENTER THE NAME :-'))
part=name[2:7]    #ONLY PRINT 2:7 --> 7-2=5 LETTERS ONLY.
for i in range(0,len(part)):   #start=0 , end= len(name) , step=2.
    print(part[i])
```

```python
#FROM 0 TO UPTO 6:-
name="DEVANSHU"
print(name[:6])
```

```python
#FROM 6TH POSITION TO LAST:-
name="DEVANSHU"
print(name[6:])
```

```python
#PRINT REVERSE:-[::-1] always print reverse string.
name="DEVANSHU"
print(name[::-1])
```

```python
#TOPS--->SOPT
name=str(input('ENTER THE NAME :-'))
first=name[-1]
last=name[0]
mid=name[1:len(name)-1]
ans=first+mid+last
print(ans)
```
