```python
'''CONDITIONAL STATEMENT--WHEN WE HAVE OPTIONS/ALTERNATIVES.
---IF
---IF ELSE
---IF ELIF
---NESTED IF
```

```python
'''IF :- IF CONDITION IS TRUE THEN IT WILL EXECUTE THAT PERTICULAR STATEMENT ONLY'''
a=int(input("ENTER THE VALUE FOR A :-"))
b=int(input("ENTER VALUE FOR B :-"))
if a>b:
    print("A IS GREATER:")
print("B IS GREATER")
```

```python
'''IF ELSE :- IS USED TO GIVE CONDITION THEN CHECK IF IS NOT TRUE THEN ELSE WILL BE EXECUTED'''
a=int(input("ENTER THE VALUE FOR A :-"))
b=int(input("ENTER VALUE FOR B :-"))
if a>b:
    print(f"HERE {a} is BIGGER THAN {b}")
else:                # YOU CANNOT GIVE CONDITION IN ELSE AND IF YOU GIVE YOU HAVE TO USE ELIF.
    print(f"HERE {b} is BIGGER THAN {a}")
```

```python
a=int(input("ENTER YOUR AGE :-"))
if a>=18:
    print("YOU ARE ELEGIBLE")
else:
    print("SORRY ! YOU ARE NOT ELEGIBLE")
```

```python
# FROM 18 TO 50 ONLY CAN DONATE:-
a=int(input("ENTER YOUR AGE :-"))
if 18<=a<=50:
#IF A>=18 AND <=50:
    print("YOU ARE ELEGIBLE")
else: 
    print("SORRY ! YOU ARE NOT ELEGIBLE")
```
