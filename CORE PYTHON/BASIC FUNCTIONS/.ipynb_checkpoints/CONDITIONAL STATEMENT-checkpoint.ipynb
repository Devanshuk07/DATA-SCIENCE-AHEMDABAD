```python
'''CONDITIONAL STATEMENT--WHEN WE HAVE OPTIONS/ALTERNATIVES.
---IF
---IF ELSE (YOU CANNOT GIVE CONDITION IN ELSE)
---IF ELIF (YOU CAN GIVE MANY CONDITIONS IN ELIF STATEMENT)
---NESTED IF(IN IF,IF CONDITION IS GIVEN)
```

```python
'''IF :- IF CONDITION IS TRUE THEN IT WILL EXECUTE THAT PERTICULAR STATEMENT ONLY'''
a=int(input("ENTER THE VALUE FOR A :-"))
b=int(input("ENTER VALUE FOR B :-"))
if a>b:
    print("A IS GREATER:")
print("B IS GREATER")
```

```python
'''IF ELSE :- IS USED TO GIVE CONDITION THEN CHECK IF IS NOT TRUE THEN ELSE WILL BE EXECUTED'''
a=int(input("ENTER THE VALUE FOR A :-"))
b=int(input("ENTER VALUE FOR B :-"))
if a>b:
    print(f"HERE {a} is BIGGER THAN {b}")
else:                # YOU CANNOT GIVE CONDITION IN ELSE AND IF YOU GIVE YOU HAVE TO USE ELIF.
    print(f"HERE {b} is BIGGER THAN {a}")
```

```python
a=int(input("ENTER YOUR AGE :-"))
if a>=18:
    print("YOU ARE ELEGIBLE")
else:
    print("SORRY ! YOU ARE NOT ELEGIBLE")
```

```python
# FROM 18 TO 50 ONLY CAN DONATE:-
a=int(input("ENTER YOUR AGE :-"))
if 18<=a<=50:
#IF A>=18 AND <=50:
    print("YOU ARE ELEGIBLE")
else: 
    print("SORRY ! YOU ARE NOT ELEGIBLE")
```

```python
#IF ELIF:-(YOU CAN GIVE MANY NUMBER OF CONDITION IN ELIF STATEMENTS)
a=int(input("ENTER THE NUMBER AS PER YOUR CHOICE :-"))
if a==1:        # SINGLE = IS FOR ASSIGN VALUE DOUBLE== IS USED FOR COMPARISION.
    print("MONDAY")
elif a==2:
    print("TUESDAY")
elif a==3:
    print("WEDNESDAY")
elif a==4:
    print("THURSDAY")
elif a==5:
    print("FRIDAY")
elif a==6:
    print("SATURDAY")
elif a==7:
    print("SUNDAY")
else:
    print("PLEASE ENTER NUMBER FROM 1 TO 7 ONLY !!")
```

```python
a=int(input("ENTER THE NUMBER AS PER YOUR CHOICE :-"))
match a:
    case 1: print('monday')
    case 2: print('tuesday')
    case 3: print('wednesday')
    case 4: print('tuesday')
    case 5: print('friday')
    case 6: print('saturday')
    case 7: print('sunday')
    case _: print('invalid entry')   # for default _ is used.
```

```python
#AGE BETWEEN 18 TO 50 THEN MESS MUST BE ADULT.IF USER AGE IS 0-2 THEN INFANT IF USER AGE IS 3-17 MINOR.IF 51-70 THEN SENIOR.ABOVE 70 SUPERSINOUR.
a=int(input("ENTER YOUR AGE :-"))
if 18<=a<=50:
    print("ADULT")
elif 0<=a<=2:
    print("INFANT")
elif 3<=a<=17:
    print("MINOR")
elif 51<=a<=70:
    print("SENIOUR")
else:
    print("SUPER SENIOUR")
    
```

```python
a=str(input("ENTER THE MONTH NAME :-"))
if a in ['january','march','may','july','august','octomber','december']:
    print('31 days')
elif a=='febrary':
    print("28 days")
elif a in ['february','april','june','september','november']:
    print('30 days')
else:
    print("INVALID MONTH NAME")
```

```python
#match case with age program.
##AGE BETWEEN 18 TO 50 THEN MESS MUST BE ADULT.IF USER AGE IS 0-2 THEN INFANT IF USER AGE IS 3-17 MINOR.IF 51-70 THEN SENIOR.ABOVE 70 SUPERSINOUR.
a=int(input("ENTER YOUR AGE :-"))
match a:
    case a if 18<=a<=50:
        print("ADULT")
    case a if 0<=a<=2:
        print("INFANT")
    case a if 3<=a<=17:
        print("MINOR")
    case a if 51<=a<=70:
        print("SENIOR")
    case a if a>70:
        print("SUPER SENIOUR")
    case _ : print("INVALID AGE NUMBER")

```

```python
#NESTED IF:- IF WITHIN IF.
'''if condition 1 true then only then check condition 2.'''
a=int(input("ENTER THE AGE FOR LEARNING LISENCE:-"))
if a>=18:
    b=(input("DO YOU HAVE LEARINING LISENCE"))
    if b in ['i have','yes','i do have','yes yes']:
        print("YOU CAN APPLY FOR LISENCE")
    else:
        print("FIRST APPLY FOR LEARNING LISENCE")
else:
    print("NOT ELIGIBLE FOR LISENCE")
```
