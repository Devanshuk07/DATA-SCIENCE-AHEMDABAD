```python
'''MODULS:-(%-REMINDER)'''
a=int(input("ENTER THE NUMBER 1 :-"))
b=int(input("ENTER THE NUMBER 2 :-"))
print(f"MODULES OF {a} And {b} is {a%b}")
```

```python
'''EXPONENT:-(2 RAISE TO 3= 2**3)'''
a=int(input("ENTER THE NUMBER 1 :-"))
b=int(input("ENTER THE NUMBER 2 :-"))
print(f"MODULES OF {a} And {b} is {a**b}")
```

```python
'''ASSIGNMENT OPERATORS:-(TO GIVE SOME VALUE/ASSIGN)'''
a=10
#then increase it by 5
a+=5
a*=5
b=a//3
print(a)
print(f"{a+20}")  # YOU CANNOT ASSIGN VALUE IN PRINT MEANS YOU CANNOT USE = IN PRINT
print(b)
```

```python
'''RELATIONAL OPERATORS:-(<,>,<=,>=,==,!=)total=6'''
a=int(input("ENTER VALUE OF 1 :- "))
b=int(input("ENTER VALUE OF 2 :- "))
c=a>b
print(f"{a}>{b}is{c}")
```
