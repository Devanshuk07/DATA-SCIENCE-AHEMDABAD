```python
#1. Print half of String in Uppercase:-
a=str(input('ENTER THE STRING :-'))
half= len(a)//2      #LET A=PYTHON SO LEN OF A IS 6 AND 6//2 MEANS=3 SO HALF IS STORED
result= a[:half].upper() + a[half:]   #a[:half]--This means take characters from index 0 up to half(not including half).a[:half]--This means take characters from index half to the end.
print(result)
```

```python
#2. Find frequency of letter from given string.(IT MEANS A SPECIFIC LETTER COMES HOW MANY TIMES IN A STRING)
a=str(input('ENTER THE STRING :-'))
b=str(input('ENTER THE LETTER :-'))
count=0
for i in a:
    if i==b:
        count=count+1
print(f'THE FREQUENCY OF {b} IN {a} IS',count)
```

```python
#3. Write a program to print length of string without len function.
a=str(input('ENTER THE STRING :-'))
count=0
for i in a:
    count=count+1
print(count)
```

```python
#4. Accept a string from the user and print it in uppercase if the length of the string is greater than 5, else print it in lowercase using a function.
a=str(input('ENTER THE STRING :-'))
if len(a)>5:
    print((a).upper())
else:
    print((a).lower())
```

```python
#5. Write a python program to convert hours into days.
a=int(input('ENTER THE HOURS TO CONVERT IT TO DAYS:-'))
days=a//24
print(f'ACCORDING TO {a} HOURS, THE DAYS ARE:-',days)
```

```python
#6. Count Even and Odd Numbers from given range.
a=int(input('ENTER THE LAST NUMBER OF RANGE:-'))
count=0
count1=0
for i in range(1,a+1):
    if i%2==0:
        count=count+1
    else:
        count1=count1+1
print('EVEN NUMBERS ARE :-',count)
print('ODD NUMBERS ARE:-',count1)
```

```python
#7.Convert a month name to a number of days. with if else and match case.
a=str(input('ENTER THE NAME OF MONTH:-'))
match a:
    case 'january'| 'jan':print(31)
    case 'february'|'feb':print(28)
    case 'march'|'mar':print(31)
    case 'april'|'apr':print(30)
    case 'may':print(31)
    case 'june'|'jun':print(30)
    case 'july'|'jul':print(31)
    case 'august'|'aug':print(31)
    case 'september'|'sep':print(30)
    case 'octomber'|'oct':print(31)
    case 'november'|'nov':print(30)
    case 'december'|'dec':print(31)
    case _:print('INVALID ENTRY')
```

```python
#7.Convert a month name to a number of days. with if else and match case.(SHORTEST WAY)
a = input("ENTER MONTH: ").lower()

if a in ["april","june","september","november"]:
    print(30)
elif a == "february":
    print(28)
elif a in ["january","march","may","july","august","october","december"]:
    print(31)
else:
    print("INVALID")
```

```python
#8.Count no of even and odd digits from given number.
a=int(input('ENTER THE NUMBER:-'))
count=0
count1=0
while a>0:     # LET a=1234 , FOR SECOND LOOP a=123, THIRD LOOP a=12, FOURTH LOOP a=1.
    b=a%10     # A%10=1234%10==4
    if b%2==0: #4%2==0
        count=count+1    # increase a counter.
    else:                # 3%2==1
        count1=count1+1  # increase a counter.
    a=a//10       # THIS WILL REMOVE THE LAST DIGIT. 1234//10==123,,, THEN AGAIN a=123//10==12. LIKE THIS.
print(count)
print(count1)
```

```python
#9.Convert temperature, degrees Fahrenheit to degrees Celsius and vice versa. C = (5/9) * (F - 32) &  F = C * 9/5 + 32.
a=int(input('ENTER THE TEMPERATURE :-'))
b=str(input('ENTER THE C AND F FOR CONVERT TEMPERATURE INTO CESIUS AND FAHRENHEIT:-')).lower()
if b=='c':
    c=(5/9)*(a-32)
    print('TEMPERATURE IN CELSIUS IS :-',c)
elif b=='f':
    f=a*9/5+32
    print('TEMPERATURE IN FAHRENHEIT IS :-',f)
```
