```python
#METHOD/FUNCTION -- STRING INBUILT:-
'''len -- to find length of string.
type -- which type of data type.
input -- accept input from user.
print -- to print something in output.
function have a specific task.
inbuilt -- already there and we have to use those function.'''

#upper , lower , title , capitalize , count , replace , split , find etc....
```

```python
name=input('enter name :-')
print(name.capitalize())
print(name.title())
print(name.upper())
print(name.lower())
print(name.count('is',5,10))    #(WHAT YOU WANT TO COUNT, START, STOP)(IT IS CASE SENSITIVE)
print(len(name))
print(name.find('morning'))      # RETURNS THE PLACE/POSITION OF THE GIVEN FIND STR.
for i in range(len(name)):
    print(i,'----->',name[i])
print(name.index('is'))
print(name.split())       # will split every word
print(name.split('is'))   # will remove the perticular word.
print(name.split(maxsplit=2))   # count=0,1,2 and max split only:-3
```

```python
a=str(input('ENTER YOUR STRING :-'))
print('a:',a.find('a'))
print('e:',a.find('e'))
print('i:',a.find('i'))
print('o:',a.find('o'))
print('u:',a.find('u'))
```
