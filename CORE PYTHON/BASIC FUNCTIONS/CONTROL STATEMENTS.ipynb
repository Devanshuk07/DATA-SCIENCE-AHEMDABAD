```python
#THIS ARE CONTROL STATEMENTS:-
'''BREAK , CONTINUE AND PASS.
BREAK:- BASED ON CONDITION BREAK THE LOOP.
CONTINUE:-BASED ON CONDITION SKIP THE CURRENT ITERATION AND RUN NEXT ITERATION.
PASS:-PLACEHOLDER , NEXT TIME IMPLEMENT.'''
```

```python
#BREAK THE LOOP WHEN REACES AT 7TH POSITION:-
i=1
while 1<=10:
    if i==8:
        break
    print(i)
    i+=1
```

```python
#CONTINUE THE LOOP WHEN REACHES AT 7TH POSITION:-
i=0
while i<=9:
    i+=1
    if i==3:
        continue
    print(i)
```

```python
#USING PASS IN IF STATEMENTS:-(PASS IS A NULL OPERATION WHEN IT IS EXECUTED NOTHING HAPPENS.)
#Python does not allow empty blocks, so when you don’t want to write code yet, you use pass.
i=1
while i<=10:
    if i==2:
        pass    #IF IN IF CONDITION NO PRINT OR STATEMENT IS GIVEN THEN IT WILL SHOW ERROR BUT PASS WORKS AS PLACEHOLDER TO CONTINUE PROGRAM.
    print(i)
    i+=1    
```
