```python
'''LOOPS MEANS REPEATATION.
1)FOR LOOP:-
2)WHILE LOOP:-'''
```

```python
'''SYNTAX:-
FOR LOOP:-
FOR I IN RANGE(START,END,STEP)(BYDEFAULT STEP=1 AND BYDEFAULT START=0 SO END IS COMPULSORY).
    BLOCK OF CODE

EXAMPLE-
PRINT 1-10 NUMBERS
FOR I IN RANGE(1,11):
    PRINT(I)'''
```

```python
'''WHEN TO USE FOR/WHILE LOOP:-
FOR---NO. OF ITERATION KNOWN IN ADVANCE..
WHILE---NOT KNOWN IN ADVANCE...'''
```

```python
"Run 10 times" → for

"Run until something happens" → while
```

```python
#FOR LOOP:-
#print 1-10:-
for i in range(1,11):
    print(i)
```

```python
for i in range(10):    #BYDEFAULT START IS 0.
    print(i)
```

```python
#FOR ALTERNATE NUMBERS FROM 1-100:-
for i in range(1,100,2):
    print(i)
```

```python
#ASK USER FOR START AND END NUMBERS:-
a=int(input('ENTER THE STARTING OF LOOP:-'))
b=int(input('ENTER THE ENDING OF LOOP:-'))
for i in range(a,b+1):   #range gives start and stop for a loop.+1 is for to take till 20 or else it will be till -1 only.
    print(i)
```

```python
#NOW REVERSE:-
for i in range(10,-1,-1):   # HERE STEP=-1 IS NECESSARY FOR -PRINTING OF VALUES.
    print(i)
```

```python
a=int(input('ENTER THE STARTING OF LOOP:-'))
b=int(input('ENTER THE ENDING OF LOOP:-'))
if a<b:
    for i in range(a,b+1):
        print(i)
else:
    for i in range(a,b,-1):
        print(i)
```

```python
#ADDITION OF 1 TO 10 NUMBERS:-
sum=0
for i in range(1,11):
    sum+=i

print(sum)
'''here i=1 and sum=0
sum=sum+i
then:   sum=0+1=1
        sum=1+2=3
        sum=3+3=6
        sum=6+4=10
        like this it will continueee.....'''
```

```python
sum=0
for i in range(1,11):
    sum+=i
    print(sum)
```

```python
#FIND A FACTORIAL NUMBER:-
#FACTORIAL OF 5 IS --1*2*3*4*5=120
a=int(input('ENTER THE NUMBER YOU WANT FACTORIAL:-'))
fact=1
for i in range(1,a+1):
    fact*=i                  # * multi will be factorial...

print(fact)
```

```python
#SUM OF EVEN NUMBERS FROM 1 TO 1000:-
a=int(input('ENTER THE SUM OF EVEN NUMBER TILL YOU WANT :-'))
x=0
for i in range(2,a+1,2):     # 0+2=2,2+2, 4+2 like this.... 
    x+=i
    print(i)
print(x)
```

```python
#SUM OF THOSE NUMBERS WHICH ARE DIVIDED BY 5 AND 7 BETWEEN 1-1000.
a=int(input('ENTER TILL YOU WANT :-'))
x=0
for i in range(1,a):
    #x+=i   # THIS IS OUTSIDE THE IF SO You are adding every number from 1 to a-1.Not only numbers divisible by 5 and 7.
    if i%5==0 and i%7==0:
        x+=i     #If you want to sum only specific numbers:The addition must be inside the if block.
        print(i)
print(x)
```

```python
#Multiplication table:- user enter number.
a=int(input('ENTER THE NUMBER WHICH YOU WANT TABLE :-'))
x=0
for i in range(1,11):
    x=i*a
    print(F"{a} x {i} = ",x)
```

```python
#WHILE LOOPS:-
#SYNTAX:-
'''WHILE CONDITION:
    BLOCK_STMT'''
# IN WHILE LOOP YOU HAVE TO DO i+=1 TO INCREASE THE i MANUALLY AS IT IS DONE AUTOMATICALLY IN IF LOOP.
```

```python
#PRINT 1-10 NUMBERS WITH WHILE LOOP:-
i=0
while i<=10:
    print(i)
    i+=1
```

```python
#TABLE USING WHILE LOOP:-
a=int(input('ENTER THE NUMBER YOU WANT TABLE :-'))
i=1
while i<=10:
    print(F"{a}x{i}={a*i}")
    i+=1
```
