```python
'''A class in Python is like a blueprint (template) used to create objects.

Think of it like this:

Class = design of a car 🚗
Object = actual car created from that design

A class is used to group data (variables) and functions (methods) together.

👉 It is the core concept of Object-Oriented Programming (OOP) in Python.

👉 Class is written once, objects are created many times
```

```python
'''SYNTAX:-

class ClassName:
    def __init__(self, parameters):
        # constructor
        self.variable = parameters

    def method_name(self):
        # method logic

⚠️ Important Points
Must write exactly: __init__ (double underscore)
First parameter is always self
Automatically called (you don’t call it manually).

__init__:- is used to initialize (give values to) an object.IT IS CONSTRUCTOR.

self:-👉 It is used to access variables and methods of the same object. 👉 self makes sure each object has its own data.
Without self, all objects would mix data ❌.
⚠️ Important Rules
self must be first parameter in methods
You don’t pass it manually
Python passes it automatically.'''
```

```python
''' | Situation                  | Need `__init__`? |
    | -------------------------- | ---------------- |
    | Just create object         | ❌ No             |
    | Pass value during creation | ✅ Yes            |
    | Store value later manually | ❌ No             |'''
```

```python
# EXAMPLE USING 1 DEF METHOD:-(JUST DISPLAY)
class friends():              #class is just a blueprint.
    def display(self,name):   #this display is just for print you cannot store value in it like __init__ can store value.
        self.name=name
        print('name:-',self.name)
f=friends()       #you have to create object f=friends() is object. object is the real thing.f is object here. if you want to friends('dev') you must have __init__ to store data.
f.display('dev')
```

```python
# USING PROPER AND BEST PRACTISE:-
class friends():
    def __init__(self, name):  # stores the value.
        self.name = name

    def display(self):
        print(self.name)

f = friends('dev')   # ✔ pass her. OBJECT -->f
f.display()          #WITHOUT USING __INIT__ YOU CAN STORE VALUE BY----> f.display('dev')manually.
```

```python
'''A class can have zero, one, or many methods depending on its purpose.'''

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")

    def calculate_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        else:
            return "C"
# a1=Student('sahil',92)
# a2=Student('vishal',39)
# a3=Student('sekhar',88)
# a1.display()
# print("GRADE:-",a1.calculate_grade())
# a2.display()
# print('GRADE:-',a2.calculate_grade())
# a3.display()
# print('GRADE:-',a3.calculate_grade())

# BETTER APPROUCH USING LOOP:-
students = [
    Student('sahil', 92),
    Student('vishal', 39),
    Student('sekhar', 88)
]

for student in students:
    student.display()
    print("GRADE:-", student.calculate_grade())
    print("-" * 30)
```

```python
#Think of a class like a blueprint of a mobile phone 📱:-
# CLASS:-
class phone:
    def __init__(self,name,brand,RAM,battery,price):        #METHOD CONSTRUCTOR. __init__ is a constructor in python.__init__ is used to initialize (give values to) an object.
        self.name=name                                # this is attributes. they belong to each object. it helps to store values inside that specific object.
        self.brand=brand
        self.RAM=RAM
        self.battery=battery
        self.price=price

    def greet(self):                      # METHOD INSIDE CLASS.A method is used to perform some action on an object
        print(self.name)
        print(self.brand)
        print(self.RAM)
        print(self.battery)
        print(self.price,'\n')

# OBJECTS:-
p1 = phone("iPhone 17", "Apple", "8GB", "4000 mAh",820000) #now you can give as many data as here to print. no need to write new codes.  self.name → "iPhone"
p2 = phone("Galaxy S25", "Samsung", "12GB", "5000 mAh",125000)
p3 = phone("OnePlus 13", "OnePlus", "16GB", "4800 mAh", 65000)

p1.greet()
p2.greet()
p3.greet()


'''p1 = Phone("iPhone 17", "Apple")
👉 Python automatically calls:
Phone.__init__(p1, "iPhone 17", "Apple")


Think like:

Class = Mobile design 📱
__init__ = Setting specifications (RAM, battery, etc.)
Object = Actual phone

👉 When a phone is created, specs are set immediately.'''
```

```python
class person:
    def __init__(self,name):
        self.name=name
        
    def greet(self):           #when we create class always write self in its def parameters.
        print('GOOD MORNING',self.name)
        print('good bye',self.name)

p1=person('dev')
p2=person('kd')
p1.greet()
p2.greet()
```
