```python
'''inheritance:-👉 is a way to create a new class using an existing class.

Existing class → Parent class / Base class / Superclass
New class → Child class / Derived class / Subclass.
-Child inherits all the properties (variables) and behaviors (methods) of the parent.
It can also add new features or override existing ones.'''
```

```python
# INHERITANCE:-
class vehicle:                     # PARENT CLASS.
    def __init__(self,brand,wheels):
        self.brand=brand
        self.wheels=wheels
    def show(self):
        print('Brand:-',self.brand)
        print('Wheels:-',self.wheels)

class car(vehicle):       # inheritance here.CHILD CLASS
    def __init__(self,brand,wheels,doors):
        super().__init__(brand,wheels)  #call parent class construction
        self.doors=doors
    def show_car(self):
        super().show()    # call parent method
        print('Doors:-',self.doors)

c1=car("MERCEDES",4,4)
c1.show_car()

'''🧠 Important Points
Child inherits parent properties → No need to rewrite code
Child can add new variables/methods → doors and show_car()
Child can override parent methods → give new behavior'''
```

```python
#INHERITANCE:-
class person():                    # PARENT CLASS
    def __init__(self,name,c_no,address):
        self.name=name
        self.c_no=c_no
        self.address=address

    def display(self):
        print(f'{self.name}--{self.c_no}--{self.address}')

class employee(person):            # CHILD CLASS.
    pass                 #The pass statement acts as a placeholder, indicating that no additional code is added yet.
                        # Without pass, Python would raise an IndentationError because a class body cannot be empty.
p1=employee('vijay',7866,'vijay-vada')
p2=employee('dhanoor',9090,'mumbai')
p3=employee('shiva',9999,'paris')
p1.display()
p2.display()
p3.display()
```

```python
#INHERITANCE:-(SINGLE:- 1 PARENT 1 CHILD.)
class person():
    def __init__(self,name,c_no,address):
        self.name=name
        self.c_no=c_no
        self.address=address

    def display(self):
        print(f'{self.name}--{self.c_no}--{self.address}')

class employee(person):
    def __init__(self,name,c_no,address,prof,dept):  #NEW __INIT__ TO TAKE OLD THEN TAKE NEW WITH IT.
        super().__init__(name,c_no,address)
        self.prof=prof
        self.dept=dept

    def display_emp(self):
        super().display()
        print(f'{self.prof}--{self.dept}\n')  # DON'T NEED TO PRINT AGAIN JUST PRINT WHAT IS REMAINING.

p1=employee('vijay',7866,'vijay-vada','doctor','gyno')
p2=employee('dhanoor',9090,'mumbai','engineer','civil')
p3=employee('shiva',9999,'paris','teacher','science')
p1.display_emp()
print("-"*30)
p2.display_emp()
print("-"*30)
p3.display_emp()
```

```python
# SHAPE HAVING ONE METHOD GETAREA WHICH RETURN 0:-(SINGLE:- 1 PARENT 1 CHILD.)
class shape():                         #We write this to create a base (parent) class.
    def getarea(self):              # All shapes have area, right? def getarea(self)--> means This ensures every shape class will have this method.
        return 0                   # It acts as a default implementation. incase,x If child class does NOT override → it returns 0
        
class rectangle(shape):            # rectangle class inherit from shape
    def __init__(self,length,breath): # this is method always use super() in method only.
        self.length=length
        self.breath=breath
    def getarea(self):                        # Each class has its own getarea(). This is called overriding.
        return self.length*self.breath        #INSTEAD OF PRINT WRITE RETURN MORE USABLE AND PROPER FOR INTERVIEWS.

class circle(shape):                # circle class inherit from shape
    def __init__(self,radius):      #initialization---> declaring
        self.radius=radius
    def getarea(self):                         #This is called overriding
        return 3.14*self.radius**2

# a=rectangle(20,3)           # //  shapes=[rectangle(20,3),circle(20)]
# print(a.getarea())          # for s in shapes:
#                                 #print(s.getarea())
# b=circle(20)
# print(b.getarea())

shapes = [rectangle(20, 3), circle(20)]

for s in shapes:
    print(s.getarea())

'''IMP POINTS:-
2 TIME DEF IS REQUIRED BCZ PARENT AND CHILD HAVE DIFFERENT BEHAVIOUR.
So yes—the parent class is not required for this simple example.

We create a base class like shape to follow Object-Oriented Programming (OOP) principles that make programs scalable, maintainable, and reusable.
-The parent class ensures that all shapes implement a getarea() method.'''
```

```python
'''TYPES OF INHERITANCE:-
SINGLE:- 1 PARENT 1 CHILD.
MULTI LEVEL:-GRAND PARENT , PARENT , CHILD
MULTIPLE:-MOTHER FATHER AND CHILD.
HIERCHICAL:-1 CLASS USED IN OTHER CLASS.(ONE PARENT TWO OR MORE CHILD)
HYBRID:-MIXING OF ANY TWO UPPER INHERITANCE
```

```python
# MULTI LEVEL:-GRAND PARENT , PARENT , CHILD:-
class grandparent():
    def gp(self):
        print('hi i am grand-parent')

class parent(grandparent):
    def p(self):
        print('hi i am parent')

class child(parent):
    def c(self):
        print('hi i am child')
        
# creating object of child class.
c=child()

# accessing the methods of all classes.
c.gp()
c.p()
c.c()
```

```python
# USING SUPER TO INHERIT PARENT HEBITS:-
class grandparent():
    def gp(self):
        print('hi i am grand-parent')

class parent(grandparent):   # class definition
    def parent_method(self): # method
        super().gp()              # You should write super() inside a method, not inside the class body.
        print('hi i am parent')

class child(parent):
    def child_method(self):
        super().parent_method()
        print('hi i am child')

c=child()
c.child_method()
```

```python
# HIERCHICAL:-1 CLASS USED IN OTHER CLASS.(ONE PARENT TWO OR MORE CHILD):-
class parent():
    def property(self):
        print("I am Parent")

class child1(parent):
    def __init__(self,name):
        self.name=name
    def display1(self):
        print(f'hi.. i am child1 my name is {self.name}')

class child2(parent):
    def __init__(self,name):
        self.name=name
    def display2(self):
        print(f'hi.. i am child2 my name is {self.name}')

c1=child1('meetraj')
c2=child2('abhay')
c1.property()
c1.display1()
c2.property()
c2.display2()
```

```python
# hybrid inheritance:--
#This structure is known as Hybrid Inheritance (a combination of hierarchical and multiple inheritance) and is also called the Diamond Structure.
class A():
    def show_a(self):
        print('class a')
class B(A):                         # b(child) depend on parent-->a 
    def show_b(self):
        print('class b')
class C(A):                          # child2(c) depends on parent-->a
    def show_c(self):
        print('class c')
class D(B,C):
    def show_d(self):
        print('class d')         # D depends on B and C directly

d=D()
d.show_a()
d.show_b()
d.show_c()
d.show_d()
```
