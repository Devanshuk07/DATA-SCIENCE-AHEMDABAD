```python
'''ENCAPSULATION:-
-Encapsulation = Data Hiding + Controlled Access.

It ensures that the internal details of an object are hidden from the outside world 
and can only be accessed through controlled interfaces (methods).

"Encapsulation is the process of wrapping data and 
methods into a single unit and restricting direct access to the data to ensure security and integrity."

DATA MUST BE PRIVATE.
----------------------------------------------------
**Real-Life Example**
----------------------------------------------------
Think of an ATM machine:

You cannot directly access the bank's database.
You interact through options like withdraw, deposit, and check balance.
Your PIN secures your account.

This is encapsulation—sensitive data is protected and accessed through defined methods.
```

```python
''' HOW TO USE ENCAPSULATION IN CODE:----

| Access Modifier | Syntax       | Accessibility                                                  |
| --------------- | ------------ | -------------------------------------------------------------- |
| Public          | `variable`   | Accessible from anywhere                                       |
| Protected       | `_variable`  | Accessible within the class and its subclasses (by convention) |
| Private         | `__variable` | Accessible only within the class (name mangling applied)       |'''
```

```python
# BASIC ENCAPSULATION:---
class bank():
    def __init__(self,acc_name,balance):
        self.acc_name=acc_name    # public variable
        self.__balance=balance    # private variable
    def __str__(self):
        return f'YOUR NAME IS {self.acc_name} & YOUR BALANCE IS {self.__balance}'
    def deposit(self,amount):
        if amount>0:
            self.__balance=self.__balance+amount
            print(f'Deposite Amount-->{amount}')
    def withdraw(self,amount):
        if amount<=self.__balance:
            self.__balance=self.__balance-amount
            print(f'Withdrawed--->{amount}')
        else:
            print(f'INSUFFISIANT BALANCE !!')

    def get_bal(self):
        return f'REMAINING BALANCE-->{self.__balance}'

b=bank('devanshu khakhar',50000)
print(b)
b.deposit(20000)
b.withdraw(10000)
print(b.get_bal())

'''PROOF OF ENCAPSULATION:--
Attempting to access the private variable directly will raise an error:
print(b.__balance)  # ❌ AttributeError'''
```

```python
class student:
    def __init__(self,name,cno):
        self.name=name
        self.__cno=cno               # with double _ _ it is private.
    def __str__(self):
        return f'{self.name} - {self.__cno}'
    def getCno(self):                            # GETTER METHOD.--->ACCESS THROUGH GETTER.
        return self.__cno
    def setcno(self,cont_no):                    # SETTER METHOD. without this both you cannot have direct access to private variable.-->MODIFY WITH SETTER.
        self.__cno=cont_no                       # here new variable cont_no is used to store new varible.

s=student('devanshu',777779)
print(s)
#print(s.__cno)
s.setcno(999990)        # used to update private num.
print(f'AFTER UPDATE:-',s)
```

```python
'''GENERATOR:--
IN OBJECT-ORIENTED PROGRAMMING, ESPECIALLY IN PYTHON,A GENERATOR IN ENCAPSULATION USUALLY REFERS TO
A GETTER METHOD—A FUNCTION USED TO RETRIEVE THE VALUE OF A PRIVATE (ENCAPSULATED) ATTRIBUTE.

In encapsulation, you cannot directly access a private variable from outside the class.
You need a generator (getter) method or a property to access it safely.

SYNTAX:-
def get_variable(self):
    return self.__variable     here __varaible is private and we create a def function to retrive that private variable.
```

```python
# EXAMPLE WITHOUT GENERATOR:-(TRYING TO ACCESS PRIVATE VARIABLE)

class Employee:
    def __init__(self, salary):
        self.__salary = salary  # Private variable

emp = Employee(50000)

print(emp.__salary)  # This will raise an error
```

```python
class employee:
    def __init__(self,emp_id,emp_name,emp_sal):
        self.emp_id=emp_id
        self.emp_name=emp_name
        self.__emp_sal=emp_sal
        
    def __str__(self):
        return f'{self.emp_id} - {self.emp_name} - {self.__emp_sal}'
        
    def getsal(self):                # getter method----> called generator.
        return self.__emp_sal
    def setemp_sal(self,employ_sal):    # setter method---> called iterator.
        self.__emp_sal=employ_sal 
        
a=employee(12,'devanshu',2300)
a.setemp_sal(23000)
print(a)
print(a.getsal())        # calling generator function getsal to access private variable.
```

```python
# Instead of traditional getters, Python encourages the use of properties:---(USING PROPERTY LATEST)

class Employee:
    def __init__(self, salary):
        self.__salary = salary

    @property
    def salary(self):         # Getter---> MODERN APPROACH
        return self.__salary

emp = Employee(50000)
print(emp.salary)       # calling def salary function.
```

```python
'''WHAT IS ITERATION:-
By default, iter() does not limit access—it simply provides a way to access the data one element at a time.
```

```python
# WHAT IS ITERATION:-

numbers = [10, 20, 30]

# Convert list into an iterator
it = iter(numbers)          # so iter() will convert any list to iterator then we can access it one by one just by next

print(next(it))  # 10
print(next(it))  # 20
print(next(it))  # 30
print(next(it))  # Raises StopIteration---> After all elements are consumed, Python raises this error.
```

```python
# USE OF ITERATION WITH ENCAPSULATION:-

class Student:
    def __init__(self):       # If you want to assign fixed marks inside the class, then only self is required.
        self.__marks = [85, 90, 78, 92]  # Private variable :- then here you have to write self.__mark=mark.

    # Providing controlled access using an iterator
    def __iter__(self):               # Here, iter() allows you to traverse existing data one item at a time.
        return iter(self.__marks)

# Creating an object
s = Student()

# Iterating through encapsulated data
for mark in s:
    print(mark)
```

```python
'''YEILD VS RETURN :-
The yield keyword is used to create a generator function.
Instead of returning all values at once like return, it produces values one at a time, pausing and resuming execution as needed.

YIELD:- RETURN WHEN YOU NEED ONE VALUE AT A TIME AND YOU HAVE TO WRITE NEXT IN PRINT FOR EVERY SINGLE VALUE.
RETURN:-RETURN WILL GIVE ALL VALUE IN A SINGLE TIME.
BOTH ARE USEFULL AS PER THE REQUIREMENT OR AS PER CONDITIONS.

| Feature          | `return`                  | `yield`                             |
| ---------------- | ------------------------- | ----------------------------------- |
| Output           | Returns a single value    | Produces multiple values one by one |
| Execution        | Function ends immediately | Function pauses and resumes         |
| Memory Usage     | Stores all data in memory | Generates values lazily             |
| Function Type    | Normal function           | Generator function                  |
| Exception Raised | None                      | Raises `StopIteration` when done    |
```

```python
class student():
    def __init__(self):           # The __init__ method is a constructor and must not return a generator=yield.
        self.__name=['rahul','sanjay','mohammad','zara','taki','big show']

    def access(self):            # this def function is to access the private variable.(this is a type of permission).
        for i in self.__name:
            yield i
s=student()
gen=s.access()
print(next(gen))      # for 1 time next only 1 will be taken.
print(next(gen))      # for 2 for second....
```

```python
# Create a Python class named Library that demonstrates encapsulation and generators:-

class series():
    def __init__(self):
        self.__books=["Python Basics", "Data Structures", "Machine Learning","Artificial Intelligence", "Deep Learning", "SQL Fundamentals"]

    def get_books(self):
        for i in self.__books:
            yield i
b=series()
for books in b.get_books():
    print(books)
```

```python
class series():
    def __init__(self):
        self.__books=["Python Basics", "Data Structures", "Machine Learning","Artificial Intelligence", "Deep Learning", "SQL Fundamentals"]

    def get_books(self,limit):
        for i in self.__books[:limit]:      #You only need to write [:limit] when you want to restrict the number of items returned.
            yield i
b=series()
c=b.get_books(4)         # this 4 is a limit that will help to take a limited books name.
for book in c:
    print(book)
#print(list(c))
# for books in b.get_books():
#     print(books)
```
