```python
#NESTED DICTIONARY:-
a={'name':'pratham',
   'age':20,
   'marks':[90,50,70]
  }
marks=a['marks']
total_marks=sum(marks)
print(total_marks)
a['total']=total_marks    #ADDING TO A.
print(a)
```

```python
#We have multi employees with name,dept,salary,age,key employees.
emp={'emp101':                                                  #HERE EMP101 IS -->K
     {'name':'kd','age':21,'salary':20000,'dept':'finance'},
     'emp201':
     {'name':'rutu','age':23,'salary':23000,'dept':'teacher'},
     'emp301':
     {'name':'meet','age':25,'salary':25000,'dept':'software'},
    }
#display emp salary more than 20000:-
#display emp who works in software dept:-
#display emp whose age is more than 25:-
#display emp whos empid is selected/entered.
for k,v in emp.items():
    if emp[k]['salary']>20000:
        print(k,emp[k]['name'],emp[k]['salary'])
```

```python
#display emp who works in software dept:-
emp={'emp101':                                                  #HERE EMP101 IS -->K
     {'name':'kd','age':21,'salary':20000,'dept':'finance'},
     'emp201':
     {'name':'rutu','age':23,'salary':23000,'dept':'teacher'},
     'emp301':
     {'name':'meet','age':25,'salary':25000,'dept':'software'},
    }
for k,v in emp.items():
    if emp[k]['dept']=='software':
        print(k,emp[k]['name'],emp[k]['dept'])
```

```python
#display emp whos empid is selected/entered:-
emp={'emp1':                                                  #HERE EMP101 IS -->K //emp1 is key and name:kd and all is its value.
     {'name':'kd','age':21,'salary':20000,'dept':'finance'},
     'emp2':
     {'name':'rutu','age':23,'salary':23000,'dept':'teacher'},
     'emp3':
     {'name':'meet','age':25,'salary':25000,'dept':'software'},
    }
b=input('ENTER THE ID OF EMP:-')
for k,v in emp.items():      # HERE K IS KEY,V IS VALUE AND EMP[K] IS ALSO VALUES.
    if b==k:
        print(emp[k]['name'])
```
