```python
'''TUPLE:-()--->BRACKETS.
IMMUTABLE:-CANNOT BE CHANGED.(STRING,INT,FLOAT,BOOLEAN,FROZENSET) THIS ALL ARE IMMUTABLE.
INDEXING
ORDERED
ALLOW DUPLICATE'''
```

```python
a=1,2,3,4,5                # NO BRACKETS.
print(type(a))
```

```python
#TUPLE:-(NO COMPREHENSIVE WORKS)
a=1,2,3,4,5
for i in a:
    print(i)
b=(i for i in a)      # COMPREHENSIVE DON'T WORK WITH TUPLE
print(tuple(b))       # YOU HAVE TO WRITE TUPLE WITH PRINT.
```

```python
#ZIP:-
a=['INDIA','USA','FRANCE','MEXICO']
b=['DELHI','WASHINGTON','PARIS','SANGHAI']
c=['LALKILA','LIBERTY','EFFILE','WALL']
ans=zip(a,b,c)
print(list(ans))        #HAVE TO CONVERT TO LIST.BY DEFAULT IT WILL BE <zip object at 0x000002582D056840>.IN LIST IT WILL BE TUPLE BY DEFAULT.
```
