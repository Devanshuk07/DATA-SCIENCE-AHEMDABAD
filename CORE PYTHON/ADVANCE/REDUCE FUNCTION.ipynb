```python
'''REDUCE:-reduce() is used to reduce a list into a single value by applying a function repeatedly.
SYNTAX:-
SAME AS FILTER AND MAP...
REDUCE(FUNCTION,ITERATION)

RETURNS SINGLE VALUE
LIBRARY IMPORT:-from functools import reduce.

Function must take 2 arguments
Returns single value
Works left to right
Needs functools module...

IMP:-Yes, reduce() always returns a single value ✅
Because reduce() keeps combining elements until only one result remains.
👉 Yes, in reduce() the function must take 2 arguments (x, y)
👉 ❌ It cannot work with only one argument like-->>x:then condition.

EXAMPLE:-
reduce(lambda x, y: x + y, a).👍👌
reduce(lambda x: x+1, a)   # ❌'''
```

```python
#FINDING THE SUM OF ELEMENT USING RANGE:-
from functools import reduce
def sum(x,y):
    return x+y
    
ans=reduce(sum,range(1,101))    #here iteration means range.
print(ans)
```

```python
#SUM OF ALL ELEMENT IN A GIVEN LIST:-
from functools import reduce
a=[1,2,3,4,5,6]
res=reduce(lambda x,y:x+y,a)    #adding every element in list.
print(res)
```

```python
#FIND MAX NUMBER FROM LIST:-
a=[1,4,7,90,99,1901]
ma=reduce(lambda x,y:x if x>y else y,a)
print(ma)
```

```python
from functools import reduce

a = [1, 2, 3]                                #STARTS WITH INITIAL VALUE:-10 THEN 10*1 ,10*2 LIKE THIS.OR ELSE BY DEFAULT INITIAL VALUE IS 1
res = reduce(lambda x, y: x*y, a, 10)        #The third argument (10) is called the initial value (initializer).MEANS X=10.THEN Y=[1,2,3] THEN X*Y.
print(res)
```

```python
#👉 Count total even numbers in a list:-
a=[1,2,34,5,7,9,10,15,30]
count=0
evn=reduce(lambda count,x:count+1 if x%2==0 else count,a,0)  #0 initial or else it will start from 1 and give answer 5.YOU CANNOT TAKE O IN FILTER AND MAPS.
print(evn)
```
