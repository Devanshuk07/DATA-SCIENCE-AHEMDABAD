```python
'''MAP:-In Python, the map() function is used to apply a function to every item of an iterable (like a list, tuple, etc.) and return the result.

map() returns a map object, so you often convert it using:

It is faster and cleaner than using loops in many cases.
It is useful when you want to transform data.
```

```python
'''MAP APPLY FUNCTION TO ITERATOR.

--SYNTAX:-
MAP(FUNCTION,ITERABLE)     ITERATION MEANS DECLARE.

EX
MAP(str.upper,lst_city)'''
```

```python
lst=['ahemdabad','rajkot']
lst_2=list(map(str.upper,lst))   #ADDING LIST MEANS CONVERTING MAP INTO LIST.
#lst_2=map(str.upper,lst)          # <map object at 0x000001FB894080C0>
print(lst_2)
```

```python
#MAP EXAMPLE:-
def square(num):
    return num*num

a=[1,2,3,4,5]
b=list(map(square,a))

print(b)
```

```python
# USING LAMBDA AND NOT TAKING NORMAL DEF FUNCTION :-
# def square(x):
#     return x*x

a=[1,2,3,4,5]
b=list(map(lambda x:x*x,a))
print(b)
```

```python
#ADDING A AND B:-
a=[1,2,3,4,5]
b=[111,222,333,444,555]

result=list(map(lambda x,y:x+y,a,b))
print(result)
```

```python
#MULTIPLYING NUMBERS:-
a=[1,2,3,4,5]
result=map(lambda x:x*2,a)
print(list(result))
```

```python
#SHORTEST FORM:-

a = [1, 2, 3]
b = [4, 5]

print(list(map(lambda x, y: x + y, a, b)))
```

```python
#Write code using map() to:👉 Convert a list of strings into their lengths:-
a=['apple','banana','mango','grapes','guava']
length=list(map(lambda x:len(x),a))
length1=sorted(a,key=lambda x:len(x))
print(length)
print(length1)
```

```python
#converting to upper case:-
a=['apple','banana','mango','grapes','guava']
upp=list(map(lambda x:x.upper(),a))
print(upp)
```

```python
#CONVERTING Fahrenheit → Celsius:-(USING MAP AND FILTER WITH LAMBDA)
a=[20.30,90,110,120,150,190]
con=list(map(lambda x:(x-32)*5/9,a))
print(con)
b=con
c=list(filter(lambda x:x>50,b))
print(c)
```

```python
#Use map() on [1,2,3,4] to multiply each element by 3:-(without converting to list)
a=[1,2,3,4]
mul=list(map(lambda x:x*3,a))
print(mul)             #map() returns a map object (iterator)..It does NOT automatically show values
```

```python
#Use map() to extract first character of each word:-
a=["apple", "banana", "cherry"]
b=list(map(lambda x:x[0],a))         #len means length of that x[0] charactor.so don't use len.
print(b)
```
