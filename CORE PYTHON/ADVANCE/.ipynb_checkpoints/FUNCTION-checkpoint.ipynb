```python
#CALCULATING SIMPLE INTEREST USING FUNCTION:-
def si(p,r,t):
        return (p*r*t)/100              #Outside print → need return ...Inside print → no return needed
        # print('SI IS:-',ints)      HERE BOTH OKK.
for i in range(1,3):
    p=int(input('ENTER PRINCIPAL:-'))
    r=int(input('ENTER RATE:-'))
    t=int(input('ENTER TIME:-'))
    print(si(p,r,t))
```

```python
def si(p,r,t):
        ints=(p*r*t)/100
        return ints         #“Send the calculated value (interest) back to where the function was called.”  WITHOUT THIS OUTPUT WILL BE NONE.
for i in range(1,3):
    p=int(input('ENTER PRINCIPAL:-'))
    r=int(input('ENTER RATE:-'))
    t=int(input('ENTER TIME:-'))
    ans=si(p,r,t)
    if ans>=500:
        print(ans)

#WHEN TO USE RETURN:-
#When you need to store the result.
#When you want to use result in conditions.
#When using function inside another function.
#When doing calculations / logic (important).

#WHEN NOT TO USE RETURN:-
#ONLY FOR DISPLAY.
#ONE-TIME OUTPUT.
```

```python
#PRIME AND COMPOSITE NUMBERS:-
def is_prime(p):
    if p <= 1:
        return False
    
    for i in range(2, int(p**0.5) + 1):    #p = 36,,p**0.5 = 6 then range(2,6+1=7) i=2,3,4,5,6,7 and p takes one by one.if any number div by i its not prime.if none divide itd prime 
        if p % i == 0:
            return False
    
    return True


p = int(input('ENTER THE NUMBER:-'))

if is_prime(p):
    print('Prime')
else:
    print('Not Prime')
```

```python
#Return factorial of a number:-
def facto(f):
    a=1
    for i in range(1,f+1):
        a=a*i
        
    return a

f=int(input('ENTER THE NUMBER:-'))
print(facto(f))
```

```python
#fibonacci(n):-(IN THIS ALWAYS A IS INCRESING)
def fibo(f):
    fibo_seq=[]
    a=0
    b=1
    for i in range(0,f):
        fibo_seq.append(a)         #APPEND:--ONLY WORKS IN INTEGER NOT IN STRING
        a,b=b,a+b
    return fibo_seq
        
f=int(input('ENTER THE NUMBER:-'))
print(fibo(f))
```

```python
#Palindrome Check:-
def palid(p):
    for ch in range(len(p)//2):   #making p-->half
        if p[ch]!=p[-(ch+1)]:     # then checking ch from start and ch from end-->if match then execute and if not out of function.
            return False
    return True

p=input('ENTER THE STRING:-').lower()
if palid(p):
    print('PALIDROME')
else:
    print('NOT A PALIDROME')
```

```python
#LARGEST OF TWO NUMBERS:-
def lar(a,b):
    if a>b:
        return a
    else:
        return b

a=int(input('ENTER VALUE OF A:-'))
b=int(input('ENTER VALUE OF B:-'))
print('LARGEST IS :-',lar(a,b))
```

```python
#COUNT VOWELS IN A STRING:-
def vol(a):
    count=0
    for ch in a:
        if ch in ['a','e','i','o','u']:
            count+=1
    return count

a=input('ENTER THE STRING:-').lower()
print(vol(a))
```

```python
#REVERSE A STRING WITHOUT SLICING:-
def rev(a):
    b=''
    for ch in range(len(a)-1,-1,-1):
        b=b+a[ch]                         #b = '' + 'o' = 'o'
    return b
    
a=input('ENTER THE STRING:-').lower()      
print(rev(a))
```

```python
'''DIFFERENT TYPE OF PARAMETER:-
1)DEFAULT PARAMETER.
2)VARIABLE ARGUMENTS.
3)KEYWORD ARGUMENT.'''
```

```python
#DEFAULT PARAMETER:-
def personaldet(name,city='ahemdabad'):  #IF WE HAVE 35 DATA OF AHEMDABAD.THEN JUST PUT CITY=AHEMDABAD.DEFAULT MUST BE AT THE END ONLY NOT IN BETWIN.
    print(name,city)

personaldet('devanshu','rajkot')
personaldet('vijay')
personaldet('meet')
```

```python
#VARIABLE PARAMETER:-
def personaldet(name,*args):     #VARIABLE ARGUMENT---> *anyname..BYDEFAULT IT WILL BE IN TUPLE.
    print(name)
    for i in args:
        print('\t',i)

personaldet('devanshu',28,'python','rajkot')
personaldet('rutu','msc')
personaldet('vijay','surat','bad')
```

```python
#KEYWORD ARGUMENT:-
def pd(**k):           #IT DISPLAY IN DICTIONARY.
    #print(k)
    for keys,values in k.items():
        print(f'{keys}:{values}')

pd(name='devanshu',age='28',subject='ds')
pd(name='rutvi',age='24',subject='msc')
```
