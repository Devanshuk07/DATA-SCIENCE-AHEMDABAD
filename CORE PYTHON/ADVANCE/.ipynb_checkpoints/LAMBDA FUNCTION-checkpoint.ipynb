```python
#LAMBDA FUNCTION:-WHEN THERE IS NO COMPLEX LOGIC AT THAT TIME USE LAMBDA FUNCTION.
#ALWAYS HAVE SINGLE EXPRESSION AND MAY HAVE MULTIPLE ARHUMENTS.
'''SYNTAX:-
LAMBDA ARGUMENT : EXPRESSION
ADD=lambda X,Y:X+Y
PRINT(add(12,23))

NORMAL FUNCTION:-
def add(a,b):
    return a+b

IN LAMBDA:-
add=lambda a,b:a+b

NOTE:-Same work, but lambda is short and quick.
Only one expression allowed.
Automatically returns value (no return keyword).

❌ Not for complex logic.
✅ Only for short and simple operations.
```

```python
'''USING LIST IN PROGRAM:-
👉 Use list() when:

You want all values at once
You want to print or reuse

👉 Don’t use list() when:
LAMBDA,FILTER,MAP--->USE LIST.
REDUCE---> DON'T USE LIST.
You only need iteration once.'''
```

```python
#LAMBDA SQUARE OF NUMBER:-
a=[1,2,3,4]
for i in a
    square=lambda a:a**2
    print(f'Square {square(i)}')
```

```python
##LAMBDA SQUARE OF NUMBER:-(USING MAP)
a=[1,2,3,4]
square=list(map(lambda x:x**2,a))
print(square)
```

```python
#COMPARE 2 VALUES IN LAMBDA FUNCTION:-(BY USING IF IN LAMBDA)
a=int(input('ENTER THE A:-'))
b=int(input('ENTER THE B:-'))
com=lambda a,b:a if a>b else b   #A lambda function that returns a single integer cannot be turned into a list directly like that.
print(f'GREATER IS :-{com(a,b)}')   #com(a,b) means you are calling a function.
```

```python
#Write a lambda to convert string to uppercase:-
a=input('ENTER THE STRING:-')
upr=lambda a:a.upper()   #HERE UPR IS A FUNCTION.IT EXPECTS ONE INPUT-->a
print(upr(a))       #“Run function upr and give it value of a”
```

```python
#LAMBDA PASS OR FAIL:-
a=int(input('ENTER YOUR MARKS:-'))
marks=lambda a:'pass' if a>=40 else'fail'  #LAMBDA ACCEPT ONE VALUE ONLY.AND WE ARE PASSING ONLY ONE VALUE IN A.
print(marks(a))          #IF WE ARE PASSING SINGLE VALUE THEN (A) IN PRINT.FOR MULTI VALUE JUST PRINT MARKS.
```

```python
#Write a lambda using map() to:👉 Convert a list [1, 2, 3, 4] into squares:-
a=[1,3,5,7,9,12,234,444]
#map--->MAP(FUNCTION,ITERABLE)     here function means lambda and iterable means input-->a.
squares=list(map(lambda x:x*x,a))     #map() takes care of applying function element by element.You write list() when you want to see or store all results at once ✅
print(squares)               #DUE TO MAP WE DON'T WRITE SQUARE(A). MAP APPLY ELEMENT BY ELEMENT.
```

```python
#APPLYING FILTER IN LAMBDA:-
nums = [10, 15, 20, 25,30,35,40,50,69,90,78]
#map and filter BOTH SAME.
#filter-->FILTER(FUNCTION,ITERATION)    here function-->lambda and iteration means variable for input-->nums. 
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)
```

```python
'''sorted() is a built-in Python function used to sort any iterable (list, tuple, string, etc.)

SYNTAX:-
sorted(iterable, key=None, reverse=False)
sorted (variable we take , key--> lambda ,reverse-->if you want you can take reverse the string)
```

```python
#SORT LIST USING LAMBDA AND SORT FUNCTION:-
a=[('brijesh',28),('vijay',21),('meet',20),('devanshu',19),('surya',30)]

sot=sorted(a, key=lambda x:x[1])
zor=sorted(a, key=lambda x:x[1],reverse=True)   #28 is x[1]-->second element of each item sort the x[1] means by age sort the list. 
print(sot)
print(zor)
```

```python
#LAMBDA WITH MULTIPLE ARGUMENTS:-
zoro=lambda x,y,z:x*y*z    #GIVING 3 ARGUMENTS IN LAMBDA.
print(zoro(2,3,4))
```

```python
#LAMBDA INSIDE FUNCTION:-
def square(x):
    return lambda x:x*x

x=int(input('ENTER THE NUMBER:-'))
print(square(x)(x))        #1st x for function & 2nd x for lambda x.so every time when you take lambda in def function you have to give two xx or yy.
```

```python
#FOR SORTING DICTIONARY:-
a={'devanshu':29,'vijay':30,'mehta':21,'pruthvi':33}
sot=sorted(a.items(),key=lambda x:x[1])     #when taking dictionary a.items must be taken.
print(sot)
```
