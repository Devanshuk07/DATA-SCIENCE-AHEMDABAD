```python
'''A class in Python is like a blueprint (template) used to create objects.

Think of it like this:

Class = design of a car 🚗
Object = actual car created from that design

A class is used to group data (variables) and functions (methods) together.

👉 It is the core concept of Object-Oriented Programming (OOP) in Python.

👉 Class is written once, objects are created many times
```

```python
'''SYNTAX:-

class ClassName:
    def __init__(self, parameters):
        # constructor
        self.variable = parameters

    def method_name(self):
        # method logic

⚠️ Important Points
Must write exactly: __init__ (double underscore)
First parameter is always self
Automatically called (you don’t call it manually).

__init__:- is used to initialize (give values to) an object.IT IS CONSTRUCTOR.

self:-👉 It is used to access variables and methods of the same object. 👉 self makes sure each object has its own data.
Without self, all objects would mix data ❌.
⚠️ Important Rules
self must be first parameter in methods
You don’t pass it manually
Python passes it automatically.'''
```

```python
#Think of a class like a blueprint of a mobile phone 📱:-
# CLASS:-
class phone:
    def __init__(self,name,brand,RAM,battery,price):        #METHOD CONSTRUCTOR. __init__ is a constructor in python. __init__ is used to initialize (give values to) an object.
        self.name=name
        self.brand=brand
        self.RAM=RAM
        self.battery=battery
        self.price=price

    def greet(self):                      # FUNCTION INSIDE CLASS.
        print(self.name)
        print(self.brand)
        print(self.RAM)
        print(self.battery)
        print(self.price)

# OBJECTS:-
p1 = phone("iPhone 17", "Apple", "8GB", "4000 mAh",820000) #now you can give as many data as here to print. no need to write new codes
p2 = phone("Galaxy S25", "Samsung", "12GB", "5000 mAh",125000)
p3 = phone("OnePlus 13", "OnePlus", "16GB", "4800 mAh", 65000)

p1.greet()
p2.greet()
p3.greet()


'''p1 = Phone("iPhone 17", "Apple")
👉 Python automatically calls:
Phone.__init__(p1, "iPhone 17", "Apple")


Think like:

Class = Mobile design 📱
__init__ = Setting specifications (RAM, battery, etc.)
Object = Actual phone

👉 When a phone is created, specs are set immediately.'''
```

```python
class person:
    def __init__(self,name):
        self.name=name
        
    def greet(self):           #when we create class always write self in its def parameters.
        print('GOOD MORNING',self.name)
        print('good bye',self.name)

p1=person('dev')
p2=person('kd')
p1.greet()
p2.greet()
```

```python
'''inheritance:-👉 is a way to create a new class using an existing class.

Existing class → Parent class / Base class / Superclass
New class → Child class / Derived class / Subclass.
-Child inherits all the properties (variables) and behaviors (methods) of the parent.
It can also add new features or override existing ones.'''
```

```python
# INHERITANCE:-
class vehicle:
    def __init__(self,brand,wheels):
        self.brand=brand
        self.wheels=wheels
    def show(self):
        print('Brand:-',self.brand)
        print('Wheels:-',self.wheels)

class car(vehicle):       # inheritance here.
    def __init__(self,brand,wheels,doors):
        super().__init__(brand,wheels)  #call parent class construction
        self.doors=doors
    def show_car(self):
        self.show()    # call parent method
        print('Doors:-',self.doors)

c1=car("MERCEDES",4,4)
c1.show_car()

'''🧠 Important Points
Child inherits parent properties → No need to rewrite code
Child can add new variables/methods → doors and show_car()
Child can override parent methods → give new behavior'''
```

```python
#INHERITANCE:-
class person():
    def __init__(self,name,c_no,address):
        self.name=name
        self.c_no=c_no
        self.address=address

    def display(self):
        print(f'{self.name}--{self.c_no}--{self.address}')

class employee(person):
    pass                 #pass is used to just not get error and without writting further necessary code.

p1=employee('vijay',7866,'vijay-vada')
p2=employee('dhanoor',9090,'mumbai')
p3=employee('shiva',9999,'paris')
p1.display()
p2.display()
p3.display()
```

```python
#INHERITANCE:-
class person():
    def __init__(self,name,c_no,address):
        self.name=name
        self.c_no=c_no
        self.address=address

    def display(self):
        print(f'{self.name}--{self.c_no}--{self.address}')

class employee(person):
    def __init__(self,name,c_no,address,prof,dept):  #NEW __INIT__ TO TAKE OLD THEN TAKE NEW WITH IT.
        super().__init__(name,c_no,address)
        self.prof=prof
        self.dept=dept

    def display_emp(self):
        super().display()
        print(f'{self.prof}--{self.dept}\n')  # DON'T NEED TO PRINT AGAIN JUST PRINT WHAT IS REMAINING.

p1=employee('vijay',7866,'vijay-vada','doctor','gyno')
p2=employee('dhanoor',9090,'mumbai','engineer','civil')
p3=employee('shiva',9999,'paris','teacher','science')
p1.display_emp()
p2.display_emp()
p3.display_emp()
```

```python
'''TYPES OF INHERITANCE:-
SINGLE:- 1 PARENT 1 CHILD.
MULTI LEVEL:-GRAND PARENT , PARENT , CHILD
MULTIPLE:-MOTHER FATHER AND CHILD.
HIERCHICAL:-1 CLASS USED IN OTHER CLASS.(ONE PARENT TWO OR MORE CHILD)
HYBRID:-MIXING OF ANY TWO UPPER INHERITANCE
```
