```python
'''MAP APPLY FUNCTION TO ITERATOR.
-SQUARE FUNCTION-- LIST--ALL ELEMENT TO APPLY---THEN USE MAP.
-CITY LIST---WAANT IT IN UPPER CASE--UPPER FUNCTION APPPLY FULL LIST---USE MAP.
--SYNTAX:-
MAP(FUNCTION,ITERABLE)

EX
MAP(str.upper,lst_city)'''
```

```python
lst=['ahemdabad','rajkot']
lst_2=list(map(str.upper,lst))   #ADDING LIST MEANS CONVERTING MAP INTO LIST.
print(lst_2)
```

```python
#MAP EXAMPLE:-
def square(num):
    return num*num

a=[1,2,3,4,5]
b=[]
for i in a:
    b.append(square(i))       #USING SQUARE FUNCTION APPLY NUM*NUM THEN SQUARING EACH A-LIST AND APPEND ADD IT TO B[].

print(b)
```
