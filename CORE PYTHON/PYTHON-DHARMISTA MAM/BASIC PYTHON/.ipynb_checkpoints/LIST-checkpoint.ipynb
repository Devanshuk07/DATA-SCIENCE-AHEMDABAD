```python
#SUM OF LIST INCLUDING STRING:-
lst=[1,2,3,4,55,'meet',77,99,88,1000,'vijay']
count=0
for i in lst:
    if type(i)==int:
        count=count+i
print(count)
```

```python
'''FIND LENGTH OF STRING FROM LIST[1,2,3,'TOPS','MORNING','WINNING']
OUTPUT:-
TOPS-->4
MORNING-->7
WINNING--->7'''
'''lst=[1,2,3,'tops','morning','winning']'''

a=['tops','morning','winning','awesome','devanshukhakhar']
b=0
for i in a:       #Here i takes each element of the list-->i[1]==tops,i[2]==morning,i[3]==winning like this.
    print(f'{i}-->',len(i))
```

```python
#CONVERT NAME INTO UPPER CASE FROM LIST:-(using indexing)
lst=list(input('ENTER THE NAMES:-'))
for i in lst:
    print(i.upper(),end='')
#print(lst)
# print(lst[0])
for i in range(len(lst)):
    lst[i]=lst[i].upper()       #lst[i]=['v','i','v','e','k'] then convert it to .upper()  ...lst[i]---> store it back in the same position.
    #print(i.upper())
print(lst)
```

```python
'''convert half list string into upper case:-
vivek,nikunj,tufel,mihir,pratham
OUTPUT:-VI,NIK,TU,MI,PRA'''

a=['vivek','nikunj','tufel','mihir','pratham']
for i in a:
    half=len(i)//2    #len("vivek")= 5,, then=(vivek)//2=vi. like this,, len('nikunj')//2= 3 that is nik.
    print(i[:half].upper())
```

```python
#SLICING IN STRING:-
a=['ahemdabad','kolkata','delhi','bihar','rajkot']
print(a[:2])
print(a[2:])

#from 2:4
print(a[2:4])

#alternate:-
print(a[0:5:2])   #(start,stop,step)

#negative index:-
print(a[::-1])

print(a[-3::])
```

```python
#APPEND AND EXTEND:-
a=['ahemdabad','kolkata','delhi','bihar','rajkot']
a.append('doha')                      #IT WILL TAKE FULL NAME
print('after append list is',a)
a.extend('jamnagar')                 # IT WILL TAKE EACH LETTER OF NAME.
print('after extend list is',a)
```

```python
#COUNT & POP & CLEAR & INSERT &INDEX &REMOVE(NAME) &reverse &sort  IN STRING:-
a=['ahemdabad','kolkata','delhi','bihar','rajkot','delhi','mehsana','ajmer','delhi']

#1.COUNT:-
print('1:-',a.count('delhi'))    #COUNT THE NUMBER OF TIME ELEMNT COMES IN LIST.

#2.SORT:-
a.sort()                   #SORT THE LIST FROM A-->Z ALPHABETICALLY.
print('2:-',a)
#a=['ahemdabad','kolkata','delhi','bihar','rajkot']
a.sort(reverse=True)         #USED FOR Z-->A PRINTING.
print('2',a)

#3.POP:-
b=['ahemdabad','kolkata','delhi','bihar','rajkot','delhi','mehsana','ajmer','delhi']
print('3:- LAST ELEMENT BY DEFAULT',b.pop())             #POP REMOVES THE ELEMENT FROM LIST.If index is not provided, it removes the last element by default.
b1=b.pop(1)                #pop[index] remove that postion element.
print('3:-',b)

#4.DEL(DELETE):-
c=['ahemdabad','kolkata','delhi','bihar','rajkot','delhi','mehsana','ajmer','delhi']
del c[1:3]                #DEL REMOVE INDEX FROM [1:3]
print('4:-',c)

#5.CLEAR:-
d=['ahemdabad','kolkata','delhi','bihar','rajkot','delhi','mehsana','ajmer','delhi']
d.clear()              #Removes all items from the list.Does not return anything (None is returned).
print('5:-',d)

#6.INSERT:-
e=['ahemdabad','kolkata','delhi','bihar','rajkot','delhi','mehsana','ajmer','delhi']
e.insert(3,'suru')         #insert(index,element)
print('6:-',e)

#7.INDEX:-(.index(element, start, end))
f=['ahemdabad','kolkata','delhi','bihar','rajkot','delhi','mehsana','ajmer','delhi']
print('7:-',f.index('delhi'))            # always show the first occurance of element in list and never see the repeated element.

#8.REMOVE(NAME):-
g=['ahemdabad','kolkata','delhi','bihar','rajkot','delhi','mehsana','ajmer','delhi']
g.remove('delhi')              #ONLY REMOVES THE FIRST ELEMENT OF LIST AND NOT ANY REPETAED OTHER.
print('8:-',g)

#9.REVERSE:-
h=['ahemdabad','kolkata','delhi','bihar','rajkot','delhi','mehsana','ajmer','delhi']
h.reverse()                   #JUST REVERSE THE FULL LIST.
print('9:-',h)
#print('9:-',h[::-1])-->h.reverse() both same.
```

```python
#SECOND HIGHEST ELEMENT FROM LIST:-
a=[12,22,1,5]
b=sorted(a)          # arrange list from 1-->100 so [-2] means 99. 
print(b[-2])        # [-1] means largest element,[-2] means second largest,[-3] means third largest.
```

```python
#LIST(COMPREHENSIVE/COMPREHENSION.):-
a=[11,12,13,55,78]
b=[i for i in a]        #THIS IS CALLED COMPREHENSIVE.
print(b)
```

```python
#LIST SQUARE AND CUBE:-
a=[11,12,13,55,78]
b=[]
c=[]
for i in a:
    b=i**2          #i**2---> MEANS I-SQUARE AND (i**3)-->I-CUBE.
    print(b)
    c=i**3
    print(c)
```

```python
#LIST SQUARE AND CUBE:-(COMPREHENSIVE WAY)
a=[11,12,13,55,78]
b=[i*i for i in a]       #i**2
c=[i*i*i for i in a]     #i**3
print(b)
print(c)
```

```python
#LIST FINDING EVEN NUMBERS ONLY:-(COMPREHENSIVE)
a=[11,12,13,55,78]
b=[i for i in a if i%2==0]                #COMPREHENSIVE
print(b)
```

```python
a=['mihir','harmi','bhavya']
b=[i.upper() for i in a]          #COMPREHENSIVE
print(b)
```

```python
a=['mihir','harmi','bhavya']
b=[i.upper() if len(i)>5 else i.lower() for i in a]     #COMPREHENSIVE
print(b)
```

```python
a=[11,12,13,55,78]
b=[i**2 if i%2==0 else i**3 for i in a]    #COMPREHENSIVE
print(b)
```
