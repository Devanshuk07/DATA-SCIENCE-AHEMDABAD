```python
'''BASIC FUNCTIONS IN PYTHON:-
FUNCTIONS--(ALWAYS USED TO PERFORM SPECIFIC TASK)
2 TYPE:
1)IN-BUILT
2)USER DEFINE
```

```python
'''
1)IN-BUILT:
ALREADY IN LIBRARY.
2)USER DEFINE:
A user-defined function in Python is a function that you create yourself using the def keyword to perform a specific task.
```

```python
'''
1)IN-BUILT FUNCTIONS:-
print() → display output.
input() → take user input.
type() → check data type.
len() → get length of string/list.

2)MATHAMATICAL FUNCTION:-
abs() → absolute value.
round() → round number.
max() → maximum value.
min() → minimum value.
sum() → sum of elements.

3)TYPE CONVERSION FUNCTION:-
int() → convert to integer.
float() → convert to float.
str() → convert to string.

4)SEQUENCE FUNCTION:-
range() → generate sequence.
sorted() → sort data.
list() → convert to list.
tuple() → convert to tuple.
```

```python
print("hello world",end=" ")   # END IS USED TO ADD STR IN ONE LINE PRINT IN ONE LINE.
print("what is your name")
```

```python
print("hello world","what is your name.",sep="-->")  #SEP(seperate) IS USED IF IN A SINGLE PRINT TWO STATEMNET IS GIVEN THEN IT WILL USED LIKE THIS.
#SEP WILL NOT BE USED IN SINGLE PRINT STATEMENT.
```
