```python
'''A company gives a bonus to employees based on their experience:

If experience is 5 years or more → Bonus = 20% of salary

If experience is between 2 and 5 years → Bonus = 10% of salary

Otherwise → Bonus = 5% of salary'''
```

```python
a=int(input("ENTER YOUR EXPERIENCE IN THE COMPANY :-"))
b=int(input("ENTER YOUR SALARY IN THE COMPANY :-"))
if a>=5:
    print(f"YOUR BONUS IS 20% OF YOUR SALARY AND YOUR SALARY IS {b} YOUR TOTAL SALARY WITH BONUS IS",b+(b*20)/100)
elif 2<=a<5:
    print(f"YOUR BONUS IS 10% OF YOUR SALARY AND YOUR SALARY IS {b} YOUR TOTAL SALARY WITH BONUS IS",b+(b*10)/100)
else:
    print(f"YOUR BONUS IS 5% OF YOUR SALARY AND YOUR SALARY IS {b} YOUR TOTAL SALARY WITH BONUS IS",b+(b*5)/100)
```

```python
# GIVEN BY CHAT GPT.
experience = int(input("Enter your experience: "))
salary = float(input("Enter your salary: "))

if experience >= 5:
    bonus_percent = 20
elif 2 <= experience < 5:
    bonus_percent = 10
else:
    bonus_percent = 5

bonus = salary * bonus_percent / 100
total_salary = salary + bonus

print(f"Bonus Percentage: {bonus_percent}%")
print(f"Bonus Amount: {bonus}")
print(f"Total Salary with Bonus: {total_salary}")
```

```python
'''A bank approves a loan if:

Salary ≥ 30,000

And credit score ≥ 750 → "Loan Approved"

Else → "Low Credit Score"

Else → "Salary Not Eligible"'''
```

```python
a=int(input("ENTER YOUR SALARY :-"))
b=int(input("ENTER YOUR CREDIT CARD SCORE :-"))
if a>=30000:
    if b>=750:
        print("YOU ARE ELIGIBLE...")
    else:
        print("YOU ARE NOT ELIGIBLE...")
else:
    print("YOUR SALARY IS LOW TO APPROVE LOAN...")
```

```python
'''Employee must have:
Experience ≥ 3 years
AND
Attendance ≥ 85%
If not → "Not Eligible for Appraisal"

If eligible, then rating decides increment:
Rating ≥ 9 → 20% increment
Rating between 7 and 8 → 10% increment
Rating between 5 and 6 → 5% increment
Rating < 5 → No increment (Poor Performance)
```

```python
i=int(input("ENTER YOUR SALARY :-"))
a=int(input("ENTER YOUR EXPERIENCE IN THE COMPANY :-"))
b=int(input("ENTER YOUR ATTENDENCE :-"))
if a>=3 and b>=85:
    c=int(input("ENTER THE RATING :-"))
    if c>=9:
        print("20% INCREMENT.TOTAL SALARY WITH INCREMENT :-",i+(i*20)/100)
    elif 7<=c<=8:
        print("10% INCREMENT.TOTAL SALARY WITH INCREMENT :-",i+(i*10)/100)
    elif 5<=c<=6:
        print("5% INCREMENT.TOTAL SALARY WITH INCREMENT :-",i+(i*5)/100)
    else:
        print("NO INCREMENT")
else:
    print("NOT ELEGIBLE FOR APPRAISAL...")
```
