```python
'''Create a parent class named Shape with a method area().
Create two child classes:
Circle
Rectangle
Override the area() method in each subclass.
Create objects of both classes and display their areas using the same method name.'''

class shape():
    def area():
        return 0              #   This is the original method. It serves as a default or placeholder implementation.

class circle(shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.14*(self.radius)**2       # These area() methods override the one defined in the shape class.
class rectangle(shape):
    def __init__(self,length,breath):
        self.length=length
        self.breath=breath
    def area(self):
        return (self.length)*(self.breath)       # These area() methods override the one defined in the shape class.

c=circle(44)
r=rectangle(44,15)
print(c.area())
print(r.area())
```

```python
'''Create a class named Vector.
Initialize it with two attributes: x and y.
Overload the + operator using the __add__() method.
The addition of two vectors should return a new Vector object.
Implement the __str__() method to display the vector neatly.
Print the result of adding two vector objects.'''

class vector():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __add__(self,other):
        return vector(self.a+other.a,self.b+other.b)
    def __str__(self):
        return f'({self.a} , {self.b})'
a1=vector(11,22)
b1=vector(33,44)
result=(a1+b1)
print(result)
```

```python
'''Use the len() function on different data types:
A string
A list
A tuple
A dictionary
Print the length of each.
Show how the same function behaves differently for different objects.'''

class string():
    def __init__(self,name):
        self.name=name
    def __len__(self):         #  len, str takes only self as paramenter /// other maths function et,gt,add takes (self,other)both.
        return len(self.name)

class list1():
    def __init__(self,name):
        self.name=name
    def __len__(self):
        return len(self.name)

class tuple1():
    def __init__(self,name):
        self.name=name
    def __len__(self):
        return len(self.name)

class dictionary():
    def __init__(self,name):
        self.name=name
    def __len__(self):
        return len(self.name)
        
a=string('lion')
f=list1(['apple','banana'])
b=tuple1((90,98,99,100))
d=dictionary({'name':'kd','age':21})

print('length of string name:-',len(a))
print('length of list is:-',len(f))
print('length of tuple is:-',len(b))
print('length of the dictionary is:-',len(d))
```

```python
'''Create an abstract base class named Payment.
Add an abstract method named pay(self, amount).
Create the following subclasses:
CreditCard
UPI
PayPal
Each subclass should implement the pay() method and print a message indicating the payment method and amount.
Create objects of each class.
Use a loop to call the pay() method polymorphically.'''

from abc import ABC,abstractmethod

class payment(ABC):        # define the need that will satisfy all subclass in here amount is there.
    def __init__(self,amount):
        self.amount=amount

    @abstractmethod           # this is an abstract method....
    def pay(self):
        pass
        
class creditcard(payment):
    def pay(self):
        print(f'YOU HAVE DONE PAYMENT BY CREDITCARD OF AMOUNT=={self.amount}')

class UPI(payment):
    def pay(self):
        print(f'YOU HAVE DONE PAYMENT BY UPI OF AMOUNT=={self.amount}')

class paypal(payment):
    def pay(self):
        print(f'YOU HAVE DONE PAYMENT BY PAYPAL OF AMOUNT=={self.amount}')

a=creditcard(5000)
b=UPI(9000)
c=paypal(22000)
a.pay()
b.pay()
c.pay()
```
