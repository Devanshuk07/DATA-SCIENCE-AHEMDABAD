```python
'''Polymorphism in Python is a core concept of Object-Oriented Programming (OOP) that allows the same method, function, or operator to behave 
differently depending on the object or data it is working with.

The word polymorphism comes from Greek:

Poly = Many
Morph = Forms
➡️ Meaning: "One name, many forms."
-------------------------------------------------------------------
Real-Life Example:-----

A person can play different roles:

As a teacher in school
As a parent at home
As a friend among peers

Similarly, in Python, a single interface can perform different tasks based on context.'''
```

```python
'''Types of Polymorphism in Python
1️⃣ Function Polymorphism (Built-in Functions)

Python functions can work with different data types.

print(len("Python"))        # String → 6
print(len([1, 2, 3, 4]))   # List → 4
print(len({1, 2, 3}))      # Set → 3

Here, the len() function behaves differently for each data type.'''
```

```python
'''2️⃣ Operator Polymorphism:-(+,-,/,*)
Operators behave differently depending on the operands.

print(5 + 3)          # Addition → 8
print("Hello " + "KD") # Concatenation → Hello KD
print([1, 2] + [3, 4]) # List merging → [1, 2, 3, 4]

-This is also known as operator overloading, which you have already practiced using __add__, __gt__, and __mod__.'''
```

```python
'''3️⃣ Method Polymorphism (Method Overriding):-
Different classes can define methods with the same name but different behaviors.'''

class Dog:                      # different classes--> dog,cat,cow.
    def sound(self):            # same name--> sound.
        print("Dog-->barks")

class Cat:
    def sound(self):
        print("Cat-->meows")

class Cow:
    def sound(self):
        print("Cow-->moos")

animals = [Dog(), Cat(), Cow()]  # as this is function so we have to write it as dog().

for animal in animals:
    animal.sound()
```

```python
'''4️⃣ Polymorphism with Inheritance:-
A child class can override a method of its parent class.'''

class Shape:             #parent class.
    def area(self):
        return 0

class Circle(Shape):    # child class
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.14*(self.radius)**2

class Rectangle(Shape):        #child class.
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def area(self):
        return (self.length)*(self.breadth)

# shapes = [Circle(22), Rectangle(22,5)]

# for shape in shapes:
#     print(shape.area())

s=Circle(22)
r=Rectangle(22,5)

print(s.area())
print(r.area())
```

```python
'''5️⃣ Duck Typing in Python:-
Python follows duck typing:

Duck typing means that Python determines an object's suitability based on the methods it has, rather than its class or type.
“If it looks like a duck and quacks like a duck, it’s a duck.”'''

class Bird:
    def fly(self):
        print("Bird can fly")

class Airplane:
    def fly(self):
        print("Airplane can fly")
        
# def make_it_fly(obj): obj.fly() is a duck-typing example where any object with a fly() method can be passed to the function, regardless of its class.

def make_it_fly(obj):    #obj is a parameter that represents any object passed to the function.
    obj.fly()            #This line calls the fly() method of the object obj.

class car():             # If the object does not have a fly() method, Python raises an error.
    # def fly(self):              # i have to give it attribute of fly to avoid error or else it will give me error.i cannot give bcz car don't fly.
    #     print('car can fly')
    pass

make_it_fly(Bird())
make_it_fly(Airplane())
make_it_fly(car())

#Even though Bird and Airplane are completely different classes, both work because they implement a fly() method.
```

```python
'''🦆 What Is the Need for Duck Typing in Real Life?

Duck typing focuses on behavior rather than type.
If an object performs the required action, it can be used—regardless of what it is.'''
```

```python
'''TYPES OF POLYMORPHISM:--

static binding (also called Early Binding) is the process where the method or function call is resolved at compile time, rather than at runtime.

“Static binding occurs when the method to be executed is known beforehand,
while dynamic binding occurs when the method is determined at runtime based on the object being used.”
This means the program already knows which method to execute before it runs.

Simple Rule to Remember:-
--------------------------------------------------------------------------------
| Scenario                                 | Binding Type                  |
| ---------------------------------------- | ----------------------------- |
| Method resolved without ambiguity        | Static (conceptual in Python) |
| Method chosen based on object at runtime | Dynamic                       |
| Duck typing (`obj.fly()`)                | Dynamic                       |
| Method overriding in inheritance         | Dynamic                       |
| Operator overloading                     | Dynamic                       |
----------------------------------------------------------------------------------
| Concept         | Analogy                                         |
| --------------- | ----------------------------------------------- |
| Static Binding  | A printed train ticket with a fixed seat number |
| Dynamic Binding | A cab booked on demand, assigned at runtime     |
```

```python
'''STATIC BINDING:-'''

class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
print(calc.add(5, 3))    # add called here.

'''Only one class and one method are involved.
The method add() is directly called.
There is no inheritance or polymorphism(overloading,operator,duck typing).
The call is straight forward and predictable.

This is considered a conceptual example of static binding in Python.'''
```

```python
'''DYNAMIC BINDING:-'''

class Dog:
    def sound(self):
        print("Dog barks")

class Cat:
    def sound(self):
        print("Cat meows")
class hasttr():                 # as just pass and not giving it sound attribute will create an error.
    pass
    
def make_sound(animal):  # WORKS WITH BOTH. CAT AND DOG.
    animal.sound()

make_sound(Dog())
make_sound(Cat())
make_sound(hasttr())

'''The function make_sound() works with different objects.
Python decides at runtime which sound() method to execute.
This is an example of polymorphism and duck typing.
It demonstrates dynamic binding.'''
```

```python
'''WHAT IS ABSTRACT CLASS?
Abstract means hiding the implementation details and showing only the essential features.

| Concept          | Purpose                                    | Main Difference in Code                  |
| ---------------- | ------------------------------------------ | ---------------------------------------- |
| **Abstraction**  | Defines a blueprint and enforces rules     | Uses `ABC` and `@abstractmethod`         |
| **Polymorphism** | Allows one interface to behave differently | Uses the same method name across classes |

✅ In simple words:
Abstract = Blueprint or Skeleton.

An abstract class is a class that cannot be instantiated (you cannot create its object).
It serves as a blueprint for other classes and contains one or more abstract methods that must be implemented by its subclasses.

SYNTAX:--
from abc import ABC, abstractmethod

class ClassName(ABC):  # Abstract class
    @abstractmethod
    def method_name(self):
        pass
```

```python
# EXAMPLE OF ABSTRACT METHOD:-
'''IMP:-----If a class is declared abstract, you cannot create its object (instantiate it).everytime you have to define def sound(self)
it will not take from upper like polymorphism(concrete class/normal class).
However, you can still use it as a blueprint for other classes.'''

from abc import ABC, abstractmethod

class animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class dog(animal):
    def sound(self):
        print('bark')
class cat(animal):
    def sound(self):
        print('meow')
class dinosaur(animal):
    def sound(self):
        print('squeaks')

# THE MAIN THING IS HERE DO NOT USE OBJECT FOR CALLING:-( do not use d=animal() you cannot store values like object)

d=dog()
c=cat()
d1=dinosaur()

d.sound()
c.sound()
d1.sound()
```

```python
# USE OF __INIT__ IN abstract class:-
'''Abstract means hiding the implementation details and showing only the essential features.
IT ENFORCE EVERY SUBCLASS TO IMPLEMENT A METHOD(RULE).

✅ In simple words:
Abstract = Blueprint or Skeleton.'''

from abc import ABC,abstractmethod
class animal(ABC):
    def __init__(self,name,voice):       # if you want to add here @abstractmethod then every class(dog,cat) should have __init__.
        self.name=name
        self.voice=voice
    @abstractmethod
    def sound(self):             # sound is abstarct method here.
        pass

class dog(animal):              # every subclass(dog,cat,dinosaur) must implement sound.means you cannot write pass in any subclasses.
    def sound(self):
        print(f'{self.name} has sound {self.voice}')

class cat(animal):
    def sound(self):
        print(f'{self.name} has sound {self.voice}')

class dinosaur(animal):
    def sound(self):
        print(f'{self.name} has sound {self.voice}')

d=dog('tommy','bark')
c=cat('kitty','meow')
d1=dinosaur('Trex','sqearks')

d.sound()
c.sound()
d1.sound()

'''In this example, we are hiding the implementation of the sound() method.

The abstract class Animal only specifies that every animal must have a sound() method, but it does not define how the sound is produced.

✔ Hidden Details
How a dog barks
How a cat meows
How a dinosaur squeaks

These details are left to the subclasses.'''
```

```python
'''DIFFERENCE BETWEEN POLYMERPHISM AND ABSTRACT:---

-Abstraction: Animal ensures every subclass implements sound().
IT ENFORCE EVERY SUBCLASS TO IMPLEMENT A METHOD(RULE).

-Polymorphism: Each subclass behaves differently when sound() is called.
IT CHECK HOW THE RULE IS OBEYED DONE DIFFERENTLY'''
```
