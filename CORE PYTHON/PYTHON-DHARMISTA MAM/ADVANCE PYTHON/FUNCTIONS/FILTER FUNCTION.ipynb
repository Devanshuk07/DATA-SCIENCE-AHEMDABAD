```python
'''FILTER() IS USED TO SELECT (FILTER) ELEMENTS FROM AN ITERABLE BASED ON A CONDITION.
-LIKE MAP FILTER ALSO TAKES ELEMENT BY ELEMENT FROM LIST.
SYNTAX SAME LIKE MAP...
MAP-->'APPLY TO ALL'
FILTER-->'ADD SOME'

SYNTAX:-
FILTER(FUNCTION,ITERATION)   
-->FILTER AND MAP HAVE SAME SYNTAX BUT DIFFERENT IN BEHAVIOUR.'''
```

```python
def checkeven(x):
    if x%2==0:
        return 'even'

x=[245,65,334,563,789,7890,778,77,55,688,43433]
b=filter(checkeven,x)
print(list(b))
    
```

```python
#CHECKING EVEN NUMBER-FILTER AND ALSO SQUARING EACH NUMBERS BY MAP:-
def chkev(x):
    return x%2==0
        
def square(x):
    return x*x

x=[1,2,3,4,5]
even_a=list(filter(chkev,x))
sqr_a=list(map(square,x))
print(even_a)
print(sqr_a)
```

```python
#DOING SAME BUT BY LAMBDA:-
a=[1,2,3,4,5]
even_a=list(filter(lambda x:x%2==0,a))      # lambda works in place of normal(chkev) place.
sqr_a=list(map(lambda x:x*x,a))
print(even_a)
print(sqr_a)
```

```python
#ALWAYS USE LIST WITH VARIABLE DECLARATION AND NOT IN PRINT:-
def con(c):
    return (c-32)*5/9

def chk(b):
    return b>20
        
f=[12,23,11,34,56,120,111,220]
a=map(con,f)
b=a
c=filter(chk,b)
print(list(a))
print(list(c))
```
