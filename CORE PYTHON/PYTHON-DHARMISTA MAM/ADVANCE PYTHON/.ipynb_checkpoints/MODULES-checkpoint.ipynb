```python
'''IMPORT OS:-
os is a built-in Python module
It helps you interact with the operating system
(files, folders, paths, etc.)

👉 After importing, you can use functions like:

os.listdir():- It returns a list of files and folders inside a directory

os.getcwd():-  MEANS “Where am I right now?”  . returns the current working directory of the Python program.

os.mkdir():-  Make Directory .It is used to create a new folder (directory) in your system.

os.chdir():-  MEANS CHANGE DIRECTORY. IN() GIVE PATH YOU WANT TO GO. To work with files in another folder.

os.walk(path):-  is used to go through all folders and subfolders automatically.
```

```python
import os
print(os.listdir('.'))   #'.' means current directory (current folder)
```

```python
'''LABTASK:- PRINT FILENAME AND CHECK IF IT IS SUBFOLDER THEN DIPLAY THE FILE OF THAT FOLDER ALSO:-'''
import os
a='F:\\DATA SCIENCE-AHEMDABAD\\CORE PYTHON\\ADVANCE'
for roots,sub,files in os.walk(a):               #It returns 3 things in each loop:root(main-folder),sub(subfolders ins main-folder),files(files inside sub-folder).
    print('FOLDER:-',roots)
    for file in files:                 #“Take each file from the current folder (root) one by one”
        print('FILES:-',file)
```
