```python
with open('binary_data.bin','wb')as file:
    data=bytes([65,66,67,68])       # 65=A,66=B,67=C,68=D ascii keys values.
    file.write(data)
    print('DATA PRINTED')
```

```python
with open('binary_data.bin','wb')as file:
    strdata='Today is Good Day'
    data=strdata.encode()     #encode used to convert it to bytes.
    file.write(data)
    print('DATA PRINTED')
```

```python
with open('binary_data.bin', 'rb') as file:
    data = file.read()        # read bytes
    text = data.decode()      # convert bytes → string
    print(data)              #shows bytes.
    print(text)              #shows string.
```
