```python
'''CONCEPTS:-
CLASS/OBJECT.
INHERITANCE.
POLYMERPHYCISM.
ENCAPSULATION.
ABSTRACTION.

1)CLASS/OBJECT:- CLASS IS A BLUE PRINT AND OBJECT IS REAL ENTITY.
CLASS IS COLLECTION OF OBJECTS.
CLASS HAS DATA MEMBERS AND MEMBER FUNCTIONS.

LET TAKE HUMAN AS CLASS:-
-HANDS,LEGS,HEAD,NOSE--->DATA
-SPEAKING,READING,LISTNING---->FUNCTIONS

OBJECT ----> REAL ENTITY.
OBJECT IS TYPE OF CLASS
OBJECT HAS ITS OWN DATA.
```

```python
class person:
    def __init__(self,name):
        self.name=name
        
    def greet(self):           #when we create class always write self in its def parameters.
        print('GOOD MORNING',self.name)
    def goodbye(self):
        print('good bye',self.name)

p=person('dev')
p.greet()
```
