```python
class acc():
    def __init__(self,name,pin,open_bal):
        self.name=name
        self.pin=pin
        self.__open_bal=open_bal
        self.trans=[]
        self.login=False
    def __str__(self):
        return f'YOUR NAME IS :- {self.name} & YOUR PIN IS :- {self.pin} & YOUR BALANCE IS :- {self.__open_bal}'
    def log_in(self):
        b=int(input('ENTER YOUR ACCOUNT NUMBER :-'))
        c=int(input('ENTER YOUR PIN :-'))
        if b==1 and c==self.pin:
            self.login=True
            print('LOGIN CORRECTLY')
        else:
            self.login=False
            print('LOGIN FAILED')
            
    def deposit(self,amount):
        if not self.login:
            return f'LOGIN FIRST'
        if amount>0:
            self.__open_bal=self.__open_bal+amount
            self.trans.append(f'DEPOSITED AMOUNT :- {amount}')
            return f'YOUR NEW BALANCE IS :- {self.__open_bal}'
        else:
            return f'YOU HAVE TO ADD MONEY MORE THAN 0'
            
    def withdraw(self,amount):
        if not self.login:
            return f'LOGIN FIRST'
        if amount<self.__open_bal:
            self.__open_bal=self.__open_bal-amount
            self.trans.append(f'WITHDRAWAL AMOUNT :- {amount}')
            return f'AFTER WITHDRAWAL YOUR BALANCE IS LEFT WITH :- {self.__open_bal}'
        else:
            return f"YOU DON'T HAVE ENOUGH BALANCE IN ACCOUNT"
            
    def history(self):
        with open('bank.txt','a')as file:
            for i in self.trans:
                file.write(i +'\n')
            
        return self.trans
            
a=acc('devanshu',1997,50000)
a.log_in()
print(a.deposit(10000))
print(a.withdraw(20000))
print(a.deposit(90000))
print(a.withdraw(120000))
print(a.history())
```
