```python
'''WHAT IS MEANS(AVERAGE) AND MEDIAN:---
MEAN :-- TOTAL ELEMENT ADDED / TOTAL ELEMENTS
MEDIAN:-- MEANS MIDDLE VALUE IF 1,2,3,4,5 THEN MEDIAN==3 AND IF 1,2,3,4,5,6 THEN 3+4=7/2==3.5 IS MEDIAN

BUT WHEN A PLAYER HAS :--
| Player | Scores     |
| ------ | ---------- |
| A      | 49, 50, 51 |
| B      | 10, 50, 90 |

HERE PLAYER-A HAS MEAN:-50  AND MEDIAN :- 50
HERE PLAYER-B HAS MEAN:-50 AND MEDIAN :- 50

BUT PLAYER-A IS CONCISTANT AND while Player B's scores change a lot. SO:-

WE NEED VARIANCE:--
Variance tells us how spread out the data is.
Formula (Population):

Variance=  ∑(x−μ)2 / N

HOW TO FIND VARIANCE JUST REMEMBER THIS :--
AD-SAD:-- ADDSAD MEANS
ADSAD:-- AVERAGE-DIFFERNECE-SQUARE-ADDSQUARE-DIVISION

PLAYER-A VARIANCE:--AVERAGE==50
DIFFERENCE==SUBSTRACT THE MEAN FROM EACH NUMBER-->  49-50=-1
                                                    50-50= 0
                                                    51-50=1 
SQUARE== SQUARE OF ALL :--(-1)square
                          (0)square
                          (1)square
ADDSQUARE== 1+0+1=2
DIVISION BY TOTAL ELEMENTS:-- 2/3=0.67
SO THIS IS POPULATION VARIANCE :-- 0.67
WHEN WE ARE TOLD TO FIND SAMPLE VARIANCE:--2/3-1=1(SAMPLE VARIANCE) SO ALWAYS FOR SAMPLE 1- FROM TOTAL ELEMENTS.

NOW STANDARD DEVIATION IS JUST SQUARE ROOT OF VARIANCE.

Standard Deviation=square-root(Variance)
- SO STD DEVIANCE OF PLAYER-A IS square-root(0.67)==0.82.
```

```python
''' THIS IS THE SMALL PROJECT OR TASK GIVEN BY SIR ON SATURDAY TO FIND OUT WHAT SHOULD BE THE GREAT TO ORDER THE FOOD ONLINE 
MEANS SWIGGY OR ZOMATO WE HAVE 2 CSV FILE OF SWIGGY AND ZOMATO WITH DELIVERY TIME AND RATINGS OF CUSTOMER. NOW FROM THAT WE HAVE TO
DECIDE WHICH SHOULD WE CHOOSE?'''

import numpy as np
import pandas as pd

zom=pd.read_csv('zomato.csv')
swg=pd.read_csv('swiggy.csv')
#print(f'{zom},\n\n\n\n\n{swg}')

zom['ZOMATO_AVERAGE_TIME1']=zom['delivery_time_mins'].mean()
zom['ZOMATO_AVERAGE_TIME2']=zom['delivery_time_mins'].median()
zom['ZOMATO_STANDARD_DEV']=zom['delivery_time_mins'].std()
print(zom)
swg['SWIGGY_AVERAGE_TIME1']=swg['delivery_time_mins'].mean()
swg['SWIGGY_AVERAGE_TIME2']=swg['delivery_time_mins'].median()
swg['SWIGGY_STANDARD_DEV']=swg['delivery_time_mins'].std()
print(swg)

'''THEN IF WE SEE ACCORDING TO DELIVERY TIME -- SWIGGY IS BETTER 0.1 minute = 6 seconds
this is such a tiny difference that we cannot confidently say Swiggy is better based only on the mean....'''

'''Standard deviation tells you how much the delivery times vary from the average.
zomato=26 min and swiggy=31min.

Smaller standard deviation → Delivery times are more consistent.
Larger standard deviation → Delivery times are less consistent and fluctuate more.'''

'''My conclusion

For this dataset, Zomato appears better overall because:

It has nearly the same average delivery time as Swiggy.
It has lower variation, making its delivery times more reliable.'''
```
