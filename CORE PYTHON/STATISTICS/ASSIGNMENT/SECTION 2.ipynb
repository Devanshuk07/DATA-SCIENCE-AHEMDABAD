```python
'''Create a Python script called playlist_stats.py that takes a list of daily Spotify song play counts 
(e.g., [120, 135, 150, 200, 120, 90, 200]) and calculates the mean (average) number of plays.'''
import pandas as pd

playlist=[120, 135, 150, 200, 120, 90, 200]
mean=sum(playlist)/len(playlist)
print(mean)
```

```python
'''Given the following array of delivery times (in minutes) for Zomato orders: 
[30, 25, 40, 35, 30, 45, 30], write code to find the median delivery time and print it.'''

zom=[30, 25, 40, 35, 30, 45, 30]
med=sum(zom)/len(zom)
print(med)
```

```python
'''Write a function most_common_rating(ratings) that takes a list of Flipkart product ratings 
(e.g., [5, 4, 4, 3, 5, 4, 2, 4]) and returns the mode (most frequent rating).'''
from statistics import mode

def most_common_ratings(rat):    # here rat is parameter and is used to store the list that you give in rat=[5, 4, 4, 3, 5, 4, 2, 4]
    return mode(rat)   # mode(rat) is to calculate the mode
    
rat=[5, 4, 4, 3, 5, 4, 2, 4]
print(most_common_ratings(rat))   # when function most_common_ratings is called it will always calculate the mode and give it.
```

```python
'''Given three lists representing YouTube video views for three different channels, compare the mean, median, and mode for each channel, and 
decide which channel's data is most affected by outliers.<br><br><em><strong>Hint:</strong> Try using one list with a very high value
(e.g., [100, 120, 110, 105, 5000]) to see the effect on mean vs median.</em>'''

import statistics
vid1=[100, 120, 110, 100, 5000]
vid2=[190,130,200,190,9000]
vid3=[320,620,900,620,5600]

# mean:-
print('MEANS')
print('VIDEO 1:- ',statistics.mean(vid1))
print('VIDEO 2:- ',statistics.mean(vid2))
print('VIDEO 3:- ',statistics.mean(vid3))

# median:-
print('\n MEDIAN')
print('VIDEO 1:- ',statistics.median(vid1))
print('VIDEO 2:- ',statistics.median(vid2))
print('VIDEO 3:- ',statistics.median(vid3))

# mode:-
print('\n MODE')
print('VIDEO 1:- ',statistics.mode(vid1))
print('VIDEO 2:- ',statistics.mode(vid2))
print('VIDEO 3:- ',statistics.mode(vid3))

'''Which channel is most affected by outliers?

Compare the difference between the mean and median:

Video 1: 1086 − 110 = 976
Video 2: 1942 − 190 = 1752 ✅ Largest difference
Video 3: 1612 − 620 = 992'''
```
