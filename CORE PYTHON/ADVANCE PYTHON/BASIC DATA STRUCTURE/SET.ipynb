```python
'''SET:-(UNIQUE ITEMS/ELEMENTS)-->{}--> ALSO FOR DICTIONARY.
REMEMBER ALWAYS TAKE UNIQUE.
MUTABLE:-CAN BE CHANGED.
NO INDEXING---> IN SETS.
UNORDERED
NO DUPLICATE VALUES ALLOWED
-COMPREHENSIVE/COMPREHENSION..'''
```

```python
#SET:-{}
#a={}-->by default dictionary.
a={1,2,3,3,2,1,4,5,7,10,9}       #ALWAYS TAKE UNIQUE.
print(type(a))
print(a)
print(len(a))             #LENGTH ALSO UNIQUE.
```

```python
a={'harmi','Dharmistha','dharmistha','harmi'}
print(a)
```

```python
a=[1,22,33,22,44,55,77]
b=set(a)             #CONVERTED A-LIST INTO A-SET.
print(b)
print(len(b))   # no duplicate is counted.
```

```python
#SET UNION:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a.union(b)
print(c)
```

```python
#SET UNION:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a | b                 # | -->UNION.
print(c)
```

```python
#SET INTERSECTION:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a.intersection(b)
print(c)
```

```python
#SET INTERSECTION:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a&b            #INTERSECTION:- &
print(c)
```

```python
#SET DIFFERENCE:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a.difference(b)     #HERE A-B HAPPENS.
d=b.difference(a)     #HERE B-A HAPPENS.
print(c)
print(d)
```

```python
#SET DIFFERENCE:-
a={1,22,33,2,3}
b={2,3,55,66,77}
c=a-b
d=b-a
print(c)
print(d)
```

```python
s = {1, 2, 3}
#s.remove(5)        # gives error
s.discard(5)       # wroks as same like remove but dont give error if not there.
print(s)
s.add([6,7])   # you cannot use list to add in set as list is mutable and set only allows immutables.
```

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}

# 1. Use union and print result
# 2. Use update and print result
# 3. Explain difference

s1.union(s2)       # will return new s1 and will not modify original one.
s1.update(s2)     # will update s1 with the elements inculded with s2.
print(s1)
```

```python
s1={1,2,3,4,5}
s1.pop()       # pop removes element but not fixed as it can remove any elemnet so output is not fixed.
print(s1)
```
