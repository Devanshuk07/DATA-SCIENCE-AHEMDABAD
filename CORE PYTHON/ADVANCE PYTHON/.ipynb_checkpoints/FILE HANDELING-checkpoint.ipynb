```python
'''MODES FOR FILE HANDLEING:-
| Mode| Meaning         |
---------------------------------------------------
with open('names.txt', 'r') as file:
---------------------------------------------------
👉 Just gives you a file object (like a handle).
👉 It does NOT read anything until you ask.

---------------------------------------------------------------------------
| "r" | Read (default)  |  It is used to read data from a file, NOT write.
---------------------------------------------------------------------------
| Condition             | What happens                |
| File exists ✔         | Opens and reads             |
| File does NOT exist ❌ | ❗ ERROR (FileNotFoundError) |

👉 Unlike 'w', 'r' will NOT create a file.

You must choose manually:-file.read() / file.readline() // file.readlines().
-----------------------------------------------------
| "w" | Write (overwrite) |When you use 'w' mode:
-----------------------------------------------------
Python checks: "Does file exist?"
❌ No → create new file
✔ Yes → delete old content and reuse file

------------------------------------------------------
| "a" | Append            |👉 It means: add data at the END of file
------------------------------------------------------
| Condition         | What happens         |
| File exists ✔     | Data is added at end |
| File NOT exists ❌ | File is CREATED      |
'a' is safe (no data loss).

------------------------------------------------------
| "x" | Create new file   |'x' = create mode (exclusive creation)
------------------------------------------------------
| Condition         | What happens              |
| File exists ✔     | ❌ ERROR (FileExistsError) |
| File NOT exists ❌ | ✔ File is CREATED         |

---------------------------------------------------------
| "b" | Binary mode       |
---------------------------------------------------------
It is used when working with:

images 🖼️
videos 🎥
audio 🎵
PDF files 📄

👉 Basically: non-text files
👉 'b' is NOT used alone:-
You combine it with other modes:

| Mode   | Meaning       |
| ------ | ------------- |
| 'rb' | read binary   |
| 'wb' | write binary  |
| 'ab' | append binary |
-----------------------------------------------------------
| "t" | Text mode         |
-----------------------------------------------------------
```

```python
# FILE READ:-(file.read)
file=open("names.txt","r")   # r is mode where r means read only.
data=file.read()
print(data)
file.close()
```

```python
# WRITE:-(SIMPLE WAY)
file=open('names.txt','w')
file.write('')        #IF FILE EXIST, IT WILL ERASE ALL DATA AND OVERWRITE NEW ONE.
file.close()
```

```python
# WRITE:-(USING WITH)
#use with-keyword for opening and closing.(DON'T NEED TO CLOSE FILE FOR SAFTY CLOSING IT WILL AUTO CLOSE)
with open('names.txt','w') as file:        #w means over-write
    file.write('d=AHEMDABAD\n')
    file.write('e=indrajeet\n')
```

```python
'''with open(...)
Opens the file names.txt
If file does not exist → it creates it         👉 Python creates the file in the current working directory (CWD)
import os print(os.getcwd())-->to check exact path where file was created.

If file already exists → it deletes old content (w-->overwrite)

👉 with--->automatically:

opens file ✔
closes file ✔ (no need to call file.close())'''
```

```python
#APPEND(ADD AT LAST):-
with open('names.txt','a') as file:        #a means add.
    file.write('a=devanshu\n')
    file.write('b=rutu\n')
    file.write('f=forever')
    file.close()                #file close it.
```

```python
# CREATE A NEW FILE( x ):-
with open('names2.txt','x') as file:
    file .write('Hi my name is devanshu\n')
    file.write('Whats ! yours ??')
```

```python
# READING THE FILE CREATED BY X:-
with open('names2.txt','r')as file:
    print(file.read())
```

```python
'''Normal 'r' / 'w' works with text (strings)
But files like images are stored as bytes (0s and 1s)

👉 So Python needs binary mode to handle raw data'''
```

```python
from PIL import Image
from IPython.display import display
import io            #input/output mode in python-->io.BytesIO(data)

with open('ANI-20251211090953.jpg','rb')as file:
    data=file.read()                                  #to read binary file.

img=Image.open(io.BytesIO(data))            #io.BytesIO(data) 👉 Makes Python think: “This is like a file”  . bytes convert--> file-like object
display(img)                           #display in jupyter.
```

```python
# (WB-->BINARY)==CREATING EXACT COPY OF A IMAGE USING BINARY MODE:-
with open('ANI-20251211090953.jpg','rb')as file1:
    data=file1.read()

with open('image_copy.jpg','wb')as file2:
    file2.write(data)
```

```python
# ab--> append binary THIS WILL CORREPT THE FILE:-
with open('image_copy.jpg', 'ab') as f:
    f.write(b'HELLO')             #'wb' needs bytes so you have to write f.write(b'hello')
```

```python
#enter user data into file until user enters ''end'':-(while,input,write method used):-
with open('data.txt','w')as file:
    while True:
        data=input('ENTER WHAT YOU WANT TO ADD:-' )
        if data=='end':
            break
        file.write(data + '\n')
```

```python
# READING THE FILE THAT WE CREATED IN UPPER CODE:-
with open('data.txt','r')as file:
    print(file.read())
```

```python
# read from one file and then write it in another (file.copy):-
with open('data.txt','r')as file:
    with open('data_copy.txt','w')as file2:
        data=file.read()            #This line reads ALL content of data.txt
        print(data)
        file2.write(data)           #It writes whatever is stored in data into data_copy.txt.Writes that same content into data_copy.txt
```
