```python
#PRIME AND COMPOSITE NUMBERS:-
def is_prime(p):
    if p <= 1:
        return False
    
    for i in range(2, int(p**0.5) + 1):    #p = 36,,p**0.5 = 6 then range(2,6+1=7) i=2,3,4,5,6,7 and p takes one by one.if any number div by i its not prime.if none divide itS prime 
        if p % i == 0:
            return False
    
    return True


p = int(input('ENTER THE NUMBER:-'))

if is_prime(p):
    print('Prime')
else:
    print('Not Prime')
```

```python
#Return factorial of a number:-
def facto(f):
    a=1
    for i in range(1,f+1):
        a=a*i
        
    return a

f=int(input('ENTER THE NUMBER:-'))
print(facto(f))
```

```python
#fibonacci(n):-(IN THIS ALWAYS A IS INCRESING)
def fibo(f):
    fibo_seq=[]
    a=0
    b=1
    for i in range(0,f):
        fibo_seq.append(a)         #APPEND:--ONLY WORKS IN INTEGER NOT IN STRING
        a,b=b,a+b
    return fibo_seq
        
f=int(input('ENTER THE NUMBER:-'))
print(fibo(f))
```

```python
#Palindrome Check:-
def palid(p):
    if p!=p[::-1]:        # here [::-1] is for counting from last.
        return False
    else:
        return True

p=input('ENTER THE STRING:-').lower()
if palid(p):
    print('PALIDROME')
else:
    print('NOT A PALIDROME')
```

```python
#COUNT VOWELS IN A STRING:-
def vol(a):
    count=0
    for ch in a:
        if ch in ['a','e','i','o','u']:
            count+=1
    return count

a=input('ENTER THE STRING:-').lower()
print(vol(a))
```

```python
#REVERSE A STRING WITHOUT SLICING:-
def rev(a):
    b=''
    for ch in range(len(a)-1,-1,-1):
        b=b+a[ch]                         #b = '' + 'o' = 'o'
    return b
    
a=input('ENTER THE STRING:-').lower()      
print(rev(a))
```

```python
#KEYWORD ARGUMENT:-
def pd(**k):           #is used to receive multiple keyword arguments and store them as a dictionary.
    #print(k)
    for keys,values in k.items():
        print(f'{keys}:{values}')

pd(name='devanshu',age='28',subject='ds')
pd(name='rutvi',age='24',subject='msc')
```

```python
#Write a function that accepts a list of numbers and returns a new list with only the numbers that are divisible by 3:-
def nl(a):
    b=[]
    for i in a:
        if i%3==0:
            b.append(i)
    return b

a=[1,3,2,3,4,7,9,10,22,28,29,39,30,48,57,18,290,19,10,19,181,177,188,171,100]
print(nl(a))
```

```python
# Write a function that accepts a number and checks if it is an Armstrong number:-
def ams(n):
    num=int(n)        #input() always gives string, not number so to convert n into integer//or you have to write n=int(input())
    power=len(n)      #number of digits.
    count=0
    
    for i in n:
        count=count+int(i)** power   # adding 1*3+5*3+3*3=153 to count.
    
    if count==num:
        return 'ARMSTRONG NUMBER'
    else:
        return 'NOT AN ARMSTRONG NUMBER'
        
n=input('ENTER THE NUMBER:-')
print(ams(n))
```

```python
#Write a function that accepts a list of strings and returns a new list with only the strings that have an odd length:-
def odd(n):
    b=[]
    for i in n:
        if len(i) % 2!=0:
            b.append(i)
    return b
n=['dev', 'meet', 'vijay', 'jahang', 'appoloo', 'shahjahan']
print(odd(n))
```

```python
#Write a function that accepts a list of numbers and returns a new list with all the numbers that are divisible by both 2 and 3:-
def div(a):
    b=[]
    for i in a:
        if i%2==0 and i%3==0:
            b.append(i)
    return b

a=[1,3,2,3,4,7,9,10,22,28,29,39,30,48,57,18,290,19,10,19,181,177,188,171,100]
print(div(a))
```

```python
#Write a function that accepts a list of integers and returns the second largest number in the list:-
def sl(a):
    b=list(set(a))    #A set in Python is a collection of unique elements, meaning it automatically removes duplicates.we are converting a to set by-->set(a)
    b.sort()    #after removing duplicates by set we have disorder list then to make it ordered we used sort.
    return b[-2]     #b[-2] means second last that is second highest after sort.

a=[1,3,2,3,4,7,9,10,22,28,29,39,30,48,57,18,290,19,10,19,181,177,188,171,100]
print(sl(a))
```

```python
# Write a function that accepts a list of numbers and returns the average of the numbers, excluding any zero values:-
def avg(a):
    b=[x for x in a if x!=0]   #simplest method. or else take for i in a then if x!=0 then b.append(i) like this.
    return sum(b)/len(b)

a=[1,3,2,3,4,7,9,10,22,28,29,39,30,48,0,0,0,0,57,18,290,19,10,19,181,177,188,171,100]
print(avg(a))
```

```python
#Write a function that accepts two strings and returns the common characters between them:-
def com(a,b):
    c=[]
    for i in a:
        if i in b:  #use i in b not i==b.
            c.append(i)
    return c

a=[1,2,3,4,5,6,7]
b=[5,6,7,8,9,10]
print(com(a,b))
```

```python
def merge_dicts(dict1, dict2):
    merged = dict1.copy()  # make a copy of the first dictionary
    merged.update(dict2)   # add key-value pairs from the second dictionary. update() adds all key-value pairs from dict2 to merged.
    return merged

# Example
d1 = {'a': 1, 'b': 2,'c':9,'d':8}
d2 = {'b': 3, 'c': 21,'e':17}

print(merge_dicts(d1, d2))
```

```python
#Write a function longest_word(sentence) that accepts a sentence and returns the longest word in the sentence:-
def longest_word(sen):
    words=sen.split()           # split sentence into list of words, sen.split()  # ['HI', 'MY', 'NAME', 'IS', 'DEVANSHU']
    return max(words,key=len)   

sen='HI MY NAME IS DEVANSHU'
print(longest_word(sen))
```

```python
#Write a function create_frequency_dict(lst) that accepts a list and returns a dictionary where the keys are the elements of the list, and the values are the count of how often each element appears:-
def create_feq_dict(lst):
    b={}            # here we want in dict format so {}
    for i in lst:
        if i in b:
            b[i]=b[i]+1         # as you have to write b[i] bcz i is here key b is dict and b[i]==value store in that key.+1 for increment of value.
        else:
            b[i]=1
    return b
lst = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
print(create_feq_dict(lst))
```
