```python
'''REGULAR EXPRESSION :-
Regular Expression (or regex) in Python is a way to search, match, and manipulate text patterns.

Think of it like a powerful search tool where instead of matching exact text, you define a pattern.

DEFINING PATTERNS:-
REGEX
METACHARACTOR
SPECIAL SEQUENCES
CHARACTOR SETS

re.match()----Matches pattern at the beginning
re.search()----Finds pattern anywhere
re.findall()----Returns all matches
re.sub()----Replace text
re.split()----Split string

| Symbol | Meaning                 |
| ------ | ----------------------- |
| `.`    | Any character           |
| `^`    | Start of string         |
| `$`    | End of string           |
| `\d`   | Digit                   |
| `\w`   | Word (a-z, A-Z, 0-9, _) |
| `\s`   | WhiteSpace              |
| `+`    | One or more             |
| `*`    | Zero or more            |
| `{n}`  | Exactly n times         |

^ → start of string
$ → end of string

+ → 1 or more
* → 0 or more (use for password)
? → 0 or 1 (use for password)

{n} → exactly n times
{n,} → at least n
{n,m} → between n and m

SYNTAX:-
IMPORT RE
re.match/findall/search then...
re.match(r'^[]+$') here ^ is to search at start and + to add middle search then $ is to search for last.all in ''.
```

```python
import re
msg=input('enter string:-')
ans=re.findall('python',msg)     # finds how many times given element come in input/string.
ans2=re.search('python',msg)     # will only return the 1st index of asked letter.
print(ans)
print(ans2)
```

```python
# USE OF FINDLL:-

import re
msg=input('enter digits:-')
ans=re.findall(r'\d',msg)     # for single fetching of digits from a given input.yes but single digits are find
print(ans)
ans2=re.findall(r'\d+',msg)   # returns the digits but in group as they are written in input
print(ans2)
ans3=re.findall(r'\d{10}',msg)  # it is condition that takes upto 10numbers that too all in one line.
print(ans3)
```

```python
# USE OF FINDALL:-

import re
msg=input('enter the string:-')
ans=re.findall(r'[A-Z]+',msg)    # return all uppercase-charactor in upper case and also in group.
print(ans)
ans2=re.findall(r'[A-Z]{5}',msg)     # {5} is for 5 letters.
print(ans2)
and3=re.findall(r'[A-Z]{5,9}',msg)
print(and3)
```

```python
# example of phone number validation:-
import re
a=('91-9090909090','91-1234567890','2187271212')
for num in a:
    if re.match(r'^\d{2}-\d{10}$',num):      # matches exactly.here{2 for contry code and {10} for number. $ is for last at last.
        print('yes valid')
    else:
        print('not valid')
```

```python
# PROGRAM TO CHECK VALID EMAIL ADDRESS:-
eml=('devukk7@gmail.com','vijay77@yahoo.com','mohit99@hotmail.com','vijay@99.me')
for mail in eml:
    if re.match(r'^[a-zA-Z0-9._%+-]+@[a-z]+.[com]{3}$',mail):   # no need to write, between classification just put a-zA-Z0-9._%+- like this. $ means at last.{3 means only 3 letters that is for com}
        print('YES THE EMAIL IS VALID !!')
    else:
        print('NOPES EMAIL IS NOT VALID')

'''
^[a-zA-Z0-9._%+-]+   → username (letters, digits, ., _, %, +, -)
@                    → must contain @
[a-zA-Z0-9.-]+       → domain name
\.                   → dot (.)
[a-zA-Z]{2,}         → domain extension (like com, in, org)
$                    → end of string'''
```

```python
# check only aphabets:-

alp=input('ENTER THE ONLY ALPHABETS:-')
if re.match(r'^[a-zA-Z]+$',alp):
    print('YES THE PRESENT CONTAINS ONLY ALPHABETS')
else:
    print('NO IT CONTAINS OTHER THINGS')
```

```python
#CHECK AXACTLY NUMBERS FROM THE STRING:-

num=input('ENTER THE STRING THAT CONTAINS ONLY NUMBERS:-')
if re.match(r'^[0-9 ]+$',num):          # here r'^[0-9 ] then after 9 space is given so it will consider space in input from user.
    print('YES ITS ONLY DIGIT')
else:
    print('NO IT CONTAINS OTHER THINGS')
```

```python
# PASSWORD VALIDATION:-

ps=('devukk7@@','meet@99718','hero22','99099','Devanu@97')
for it in ps:
    if re.match(r'^(?=.*[A-Z])(?=.*[a-z])(?=.*[0-9])(?=.*[@#$%&*])[A-Za-z0-9@#$%&*]{6,}$',it):     # $ means end of string,,Nothing can come after $.(?=...) → Lookahead (checks condition, doesn’t consume). .* → any characters. {6,} → minimum 6 characters required.
        print('PASSWORD IS VALID')
    else:
        print('PASSWORD INVALID')

'''
(?=.*[A-Z])----> "Check that somewhere in the string there is at least one uppercase letter (A–Z)”.
It does NOT consume or match characters
👉 It only checks a condition.(?=.*[]) this is condition matching and in [] we have to give condition.
[A-Za-z0-9@#$%&*]----> “Only these characters are allowed in the password”.
{6,}---> It is called a quantifier. means passwrod must be of minimum 6 charactor. max it can be anything.
{6,10} → between 6 and 10 characters.
```

```python
# REMOVE SPACES FROM STRING:-
txt='hello     world   this                  is     KD'
res=re.sub(r'\s+',' ',txt)
print(res)
```

```python
# VALIDATE DATE FORMATE:-(DD-MM-YYYY)
date='1997-05-09'
if re.match(r'^[0-9]{2}-[0-9]{2}-[0-9]{4}$',date):
    print('DATED')
else:
    print('WRONG FORMAT')
```
