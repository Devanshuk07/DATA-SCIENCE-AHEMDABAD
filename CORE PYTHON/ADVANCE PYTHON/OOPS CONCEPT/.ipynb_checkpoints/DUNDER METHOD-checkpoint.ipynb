```python
'''DUNDER METHODS(SPECIAL METHODS):-

Dunder Methods (Special Methods)—short for “Double UNDERscore methods”—are special functions in Python that begin and end with double underscores,
such as __init__ and __str__. These methods enable objects to interact seamlessly with Python’s built-in operations and syntax.

They are also known as magic methods because they are automatically invoked by Python in specific situations.

-A dunder method is a predefined method that allows you to define how objects behave with operators, built-in functions, and standard operations.

EXAMPLES:-
__init__:- __init__() is a special method used to initialize an object when it is created.
__str__:- is a special (dunder) method that returns a user-friendly description of an object.
__add__:- method tells Python what to do when the + operator is used between them.
__len__:-DISPLAY THE NUMBER OF ITEMS WE GAVE TO OBJECT.
__eq__:-stands for “equal to.” It allows you to customize how Python determines whether two objects are equal.
__gt__:-stands for "greater than".two object comparision who is greater than other.
```

```python
'''operators over loading in python:-

Common Operator Overloading Methods methods inn 

| Operator | Magic Method   | Description           |
| -------- | -------------- | --------------------- |
| `+`      | `__add__`      | Addition              |
| `-`      | `__sub__`      | Subtraction           |
| `*`      | `__mul__`      | Multiplication        |
| `/`      | `__truediv__`  | Division              |
| `//`     | `__floordiv__` | Floor division        |
| `%`      | `__mod__`      | Modulus               |
| `**`     | `__pow__`      | Power                 |
| `<`      | `__lt__`       | Less than             |
| `>`      | `__gt__`       | Greater than          |
| `<=`     | `__le__`       | Less than or equal    |
| `>=`     | `__ge__`       | Greater than or equal |
| `==`     | `__eq__`       | Equality              |
| `!=`     | `__ne__`       | Not equal             |
| `str()`  | `__str__`      | String representation |
| `len()`  | `__len__`      | Length of object      |

this all are overloading in python.
```

```python
''' ARITHMENTIC MATHEMATICAL EQUATIONS:-

| Operator | Dunder Method  | Can Return `less`? |
| -------- | -------------- | ------------------ |
| `+`      | `__add__`      | ✅ Yes              |
| `-`      | `__sub__`      | ✅ Yes              |
| `*`      | `__mul__`      | ✅ Yes              |
| `/`      | `__truediv__`  | ✅ Yes              |
| `//`     | `__floordiv__` | ✅ Yes              |
| `%`      | `__mod__`      | ✅ Yes              |
| `**`     | `__pow__`      | ✅ Yes              |'''
```

```python
# EXAMPLES USING __INIT__ AND __STR__:-
class student():
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
        
    def __str__(self):
        return f'student(name: {self.name}, marks: {self.marks})'

a1=student('mihir',99)
print(a1)             #Python automatically calls the __str__() method behind the scenes. 
                      # In other words, the above line is internally equivalent to print(a1.__str__).
```

```python
# USING __STR__ IN PROGRAM:-
'''A class can have zero, one, or many methods depending on its purpose.'''

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # def display(self):                # YOU CAN WRITE DIRECT __STR__ THEN JUST PRINT IT IN LOWER CODE.
        # print(f"Name: {self.name}, Marks: {self.marks}") 

    def calculate_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        else:
            return "C"
    def __str__(self):
        grade=self.calculate_grade()       # PRINTING CALCULATE_GARDE HERE SO NO NEED TO PRINT AT LOWER CODE.
        return f"Student(Name: {self.name}, Marks: {self.marks}, Grade: {grade})"

a1=Student('mihir',99)
# a1.display()
print(a1)        # Here, print(a1) automatically calls __str__(), eliminating the need for display().
```

```python
# USING __ADD__:-
class rectangle():
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def area(self):
        return self.length* self.breadth
    def __str__(self):                # just by printing if you wannt to display you have to take __str__.
        return f"{self.length}--{self.breadth}"
    def __add__(self,other):
        return rectangle(self.length+ other.length,self.breadth+other.breadth)
        
r1=rectangle(20,5)
print(r1.area())
r2=rectangle(30,10)
print(r2.area())
r3=rectangle(40,20)
print(r3.area())
r4=r1.area()+r2.area()+r3.area()
print(r4)

# # IF WE USE SUM METHOD THEN NO NEED OF USE OF __ADD__ TO WRITE IN CODE.
# rectangles=[r1,r2,r3]
# total_area=sum(r.area() for r in rectangles)
# print('TOTAL AREA:-',total_area)
```

```python
class distance:
    def __init__(self,feet,inch):
        self.feet=feet
        self.inch=inch
    def __str__(self):
        return f'{self.feet}--{self.inch}'
    def __mul__(self,other):
        return distance(self.feet*other.feet,self.inch*other.inch)

d1=distance(12,34)
d2=distance(24,50)
d3=d1*d2
print(d3)
```

```python
# USING __LEN__:-

class Students:
    def __init__(self, names):
        self.names = names

    def __len__(self):
        return len(self.names)

s = Students(["Mihir", "Dev", "Krish", "Riya"])   # __LEN__ DISPLAY THE NUMBER OF ITEMS WE GAVE TO OBJECT.

print(len(s))          
print(s.__len__())  
```

```python
# USING __EQ__:-

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __eq__(self, other):
        return (self.name == other.name and self.marks == other.marks)

s1 = Student("Dev", 90)
s2 = Student("Dev", 90)

print(s1 == s2)
```

```python
# USING __LT__:-(less than)
class less():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __str__(self):
        return f'{self.a}--{self.b}'
    def __lt__(self,other):
        return (self.a+self.b)<(other.a+other.b)   # FOR ANY COMPARISION WE DONT HAVE TO WRITE return less. ONLY WRITE IT IN ADD,SUB,MUL,DIV ONLY.
a=less(5,4)
b=less(6,1)
print(a)
print(b)
c=a<b
print(c)

'''
a=5+4=9
b=6+1=7
so a<b = false'''
```

```python
#__gt__:-(greater than)
class great():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __str__(self):
        return f"{self.a}---{self.b}"
    def __gt__(self,other):
        return (self.a+self.b)>(other.a+other.b)

a=great(24,56)
b=great(45,22)
print(a>b)

'''here (self.a+self.b)>(other.a+other.b)
========>24+56=80  >   45+22=67 so true'''
```

```python
#__truediv__:-(true division)
class great():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __str__(self):
        return f"{self.a}---{self.b}"
    def __truediv__(self,other):
        return (self.a+self.b)/(other.a+other.b)

a=great(24,56)
b=great(45,22)
print(a/b)

'''24+56=80 ,,,45+22=67
so 80/67=1.1940....'''
```

```python
# __floordiv__(FLOOR DIVISION):-
class great():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __str__(self):
        return f"{self.a}---{self.b}"
    def __floordiv__(self,other):
        return (self.a+self.b)//(other.a+other.b)

a=great(24,56)
b=great(45,22)
print(a//b)

'''floor division is like division only but just take only before decimal values.'''
```

```python
# __mod__(modulus):-
class great():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __str__(self):
        return f"{self.a}---{self.b}"
    def __mod__(self,other):
        return (self.a+self.b)%(other.a+other.b)

a=great(24,56)
b=great(45,22)
print(a%b)

'''80%67==13(reminder)'''
```

```python
# __pow__(power):-

class great():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __str__(self):
        return f"{self.a}---{self.b}"
    def __pow__(self,other):
        return (self.a+self.b)**(other.a+other.b)

a=great(2,10)
b=great(4,1)
print(a**b)

'''2+10=12 raise to 4+1=5 so answer==248832'''
```
