```python
class rectangle():
    def __init__(self,length,breath):
        self.length=length
        self.breath=breath
    def area(self):
        a=self.length*self.breath
        print(f'AREA OF RECTANGLE IS : {a}')
    def perimeter(self):
        b=2*(self.length+self.breath)
        print(f'PERIMETER OF RECTANGLE IS : {b}')
a1=rectangle(20,5)
a1.area()
a1.perimeter()
```

```python
class calculator():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        ad=self.a+self.b
        return ad
    def sub(self):
        sb=self.a-self.b
        return sb
a1=calculator(20,40)
print(a1.add())
print(a1.sub())
```

```python
class bankacc():
    def __init__(self,acc_no,balance):
        self.acc_no=acc_no
        self.__balance=balance
    def deposit(self,amount):
        if amount>=500:
            self.__balance=self.__balance+amount
            print(f'YOU HAVE DEPOSITED AMOUNT : {amount}')
            print(f'NOW BALANCE IS : {self.__balance}')
    def withdraw(self,amount):
        if amount<self.__balance:
            self.__balance=self.__balance-amount
            print(f'YOU HAVE REQUESTED FOR WITHDRAWAL OF : {amount}')
            print(f'NOW THE BALANCE IS : {self.__balance}')
        else:
            print('INSUFFICIANT AMOUNT')
a=bankacc('dev',50000)
a.deposit(20000)
a.withdraw(10000)
```

```python
class atm():
    def __init__(self,pin,balance):
        self.__pin=pin
        self.__balance=balance
        
    def set_pin(self,new_pin):
        self.__pin=new_pin
        print('YOUR PIN HAS BEEN SET !!')
    
    def check_pin(self,st_pin):
        return self.__pin == st_pin
        
    def withdraw(self,st_pin,amount):
        if not self.check_pin(st_pin):
            print('invalid pin')
            return 
            
        self.__balance=self.__balance-amount
        print(f'NOW YOUR BALANCE LEFT WITH : {self.__balance}')
            
        if amount>self.__balance:
            print('INSUFFICIENT BALANCE')
            
a=atm(1997,50000)
a.set_pin(1997)
print(a.check_pin(2997))
a.withdraw(1997,20000)
```

```python
class bankacc():
    def __init__(self,bal):
        self.__bal=bal
    def deposit(self,amount):
        self.__bal=amount+self.__bal
        print(F'AFTER DEPOSIT YOUR BALANCE IS : {self.__bal}')
    def get_bal(self):
        return self.__bal

class saving_acc(bankacc):
    def __init__(self,bal):
        super().__init__(bal)
        
    def add_int(self):
        self.__bal=self.get_bal()
        intrst=self.__bal*0.05
        new_bal=self.__bal+intrst
        print(f'AFTER INTEREST YOUR BALANCE IS : {new_bal}')
a=saving_acc(20000)
a.deposit(10000)
a.add_int()
```

```python
class shape():
    def area(self):
        return 0
class rectangle(shape):
    def __init__(self,length,breath):
        self.length=length
        self.breath=breath
    def area(self):
        return self.length*self.breath
class circle(shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.14*self.radius**2
a=rectangle(20,5)
print(a.area())
b=circle(22)
print(b.area())
```
