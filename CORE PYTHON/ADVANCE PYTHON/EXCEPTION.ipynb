```python
'''EXCEPTION MEANS BREAK THE NORMAL EXECUTION OF FLOW:-
-WHEN GIVING INCORRECT INPUT.
-ACCESSING INVALID KEY.
-NUMBER IS DIVIDE BY 0.

EXCEPTION IS HANDLE BY:-
TRY , EXCEPT , FINALLY , ELSE BLOCKS.

TRY-
EXCEPT:- YOU CAN HAVE MULTIPLE TYPE OF EXCEPTION BLOCK.(DUE TO MULTI ERRORS)
```

```python
try:
    dict1={'name':'devanshu'}
    print(dict1['age'])             #HERE WE DON'T HAVE KEY==AGE.
    print(12/0)
except:
    print('printed exception')
```

```python
#FINALLY:-(WILL ALWAYS EXECUTED WHEATHER ERROR OCCURE OR NOT)
try:
    dict1={'name':'devanshu'}
    print(dict1['age'])             #HERE WE DON'T HAVE KEY==AGE.
    print(12/0)
except:
    print('printed exception')    #IF ERROR IN TRY EXCEPT WILL BE EXECUTED.
finally:
    print('IN THE END')          #THIS WILL ALWAYS EXECUTE.
```

```python
#MULTIPLE EXCEPT:-
try:
    #num=int(input('ENTER THE NUMBER:-'))
    #print(num)
    dict1={'name':'devanshu'}
    #print(dict1['age'])             #HERE WE DON'T HAVE KEY==AGE.
    print(12/2)
except ValueError as s:
    print(s)
except KeyError as v:
    print(v)
except ZeroDivisionError as z:
    print(z)
except exception as e:
    print(e)
else:
    print('IN ELSE')               #if no except is executed then else will be executed.
finally:
    print('PROGRAM OVER !! THANKS ')
```
