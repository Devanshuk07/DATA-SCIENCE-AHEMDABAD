```python
# 13.Write a Python program to sort a dictionary (ascending /descending) by value:-

a={4:'devanshu',
   5:'vijay',
   3:'ruhi',
   1:'mohit',
   2:'zara'
  }
dit1=sorted(a.items())                      # this is for int ascending. Add reverse=True for descending.
dit2=sorted(a.items(),key=lambda x:x[1])    # this is for NAMES ASCENDING.
dit3=sorted(a.items(),key=lambda x:x[1], reverse=True)
print(dit1,'\n',dit2,'\n',dit3)
```

```python
'''17.Write a python program using function to find the sum of odd series andevenseries
Odd series: 12/ 1! +32/ 3! + 52/ 5!+……n
Even series: 22/ 2! + 42/ 4! + 62/ 6!+……n'''
a = int(input('enter the start number:-'))
b = int(input('enter the end number:-'))
c = input('enter even or odd:').lower()

if c == 'odd':
    x = 1   # for odd numbers (1,3,5...)
    
    for i in range(a, b+1, 20):
        fact = 1
        
        # factorial of x
        for j in range(1, x+1):      #THIS IS FOR FACTORIAL SO DON'T ADD LAST STEP=2.FOR ODD NUMBERS.
            fact=fact*j
        
        print(f'{i}/{fact}={i/fact:.1f} +',end='  ')
        
        x += 2   # next odd number      
#EVERY TIME IT GOES TO 1ST LOOP FOR I IN RANGE TO CHECK IS NUMBER IN RANGE OR NOT THEN AND THEN COMES INNER LOOP AND PRINT FACTORIALS.
if c=='even':
    x=2  # for even numbers(2,4,6,8..)
    for i in range(a,b+1,20):
        fact=1
        for j in range(2,x+1):
            fact=fact*j

        print(f'{i}/{fact}={i/fact:.1f} +',end='  ')
        x=x+2
```

```python
#18. Python Program to Find Factorial of Number Using Recursion:-
a=int(input('ENTER THE NUMBER FOR WHOSE YOU WANT FACTORIAL:-'))
count=1
for i in range(1,a+1):
    count=count*i
print(count)
```

```python
# USING RECURSION:-
def factorial(n):
    if n == 0 or n == 1:   # base case
        return 1
    else:
        return n * factorial(n-1)   # recursive call

a = int(input("ENTER NUMBER: "))
print("Factorial =", factorial(a))
```
