```python
'''CONCATINATION OF ARRAY:-(ONLY WORK TO JUST JOIN TWO STRING TOGETHER)
ADDITION OF ARRAY:-'''
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
c=np.array([[9,10],[11,12]])
print(np.concatenate((a,b,c)),'\n')        # ARRAY MUST HAVE SAME SHAPE TO WORK CONCATENATE.
print('AXIS 0:- \n',np.concatenate((a,b,c),axis=0),'\n')       # AXIS=0 JOINS ROWS.
print('AXIS-1:-\n',np.concatenate((a,b,c),axis=1))       # EXIS=1 MEANS JOIN COLUMNS.
```

```python
# STAKING FUNCTION:-(HSTACK,DSTACK)
#HSTACK:-INCREASE COLUMNS
arr1=np.array([[1,2,3],
            [4,5,6]])
arr2=np.array([[7,8,9],
            [10,11,12]])
print('HSTACK:-',np.hstack((arr1,arr2)),'\n')   # INCREASE HORIZENTAL(COLUMN). LIKE AXIS-1
print('VSTACK :-',np.vstack((arr1,arr2)),'\n')   #INCREASE VERTICAL(ROWS). LIKE AXIS-0
print('DSTACK :-',np.dstack((arr1,arr2)))   # INCREASE depth.
'''hstack → horizontal → left-right
vstack → vertical → up-down
dstack → front-back (depth)'''
```

```python
'''COPY VS VIEW:-
View → shares original data.VIEW DOESN'T CREATE A NEW MEMORY.IF YOU CHANGE THE VIEW ORIGINAL ARRAY WILL ALSO CHANGE.
Copy → creates completely new data.
```

```python
import numpy as np

arr = np.array([1,2,3,4])
v=arr.view()
print(v)
v[0]=100   # added 100 in place of 1
print(v)   # now view has been changed
print(arr)  # due to this original has also been changed.
```

```python
import numpy as np

arr = np.array([1,2,3,4])
c=arr.copy()
c[0]=100   # added 100 in place of 1
print(c)   # chnaged copy of arr
print(arr)  # original array will not be affected.
```

```python
'''Create: arr = np.array([1,2,3,4]) 
Create: 
view_arr using view() 
copy_arr using copy() 
Modify arr[0] = 999 
Print all three arrays. Explain what happened.'''
import numpy as np

arr=np.array([1,2,3,4])    # this is original array
print(arr)
view_arr=arr.view()     # we created view
view_arr[0]=999         # we changed in view
copy_arr=arr.copy()     # we created copy
copy_arr[0]=777         # we changed in copy
print(view_arr)        # printing view
print(copy_arr)        #printing copy
print(arr)           # original array will be changed bcz view modify original array also.copy do not modify original array.
```

```python
'''Given: arr = np.array([[1,2,3], [4,5,6]]) 
Apply: 
flatten() ravel() 
Modify first element of both results. Observe difference'''
import numpy as np
arr = np.array([[1,2,3], [4,5,6]])

a=arr.ravel()   # ravel changed the original array like view
a[0]=100
print(a)
b=arr.flatten()   # flatten do changes in copy of original array and not the original array itself.like a copy.
b[0]=100
print(b)
print(arr)
```

```python
'''Create array: arr = [5, 10, 15, 20, 25, 30] 
• Extract elements greater than 10 AND less than 30. 
• Find their cumulative sum.'''

import numpy as np

arr=np.array([5, 10, 15, 20, 25, 30])
ext=arr[(10<arr) & (arr<30)]
print("ALL ELEMENTS GREATER THAN 10 AND LESS THAN 30 :---",ext)
print(np.cumsum(ext))                # cumsum is running total.


'''
| For normal Python | For NumPy |
| ----------------- | --------- |
| `and`             | `&`       |
| `or`              | `\|`      |'''
```

```python
'''Generate 1000 random numbers from normal distribution.
Compute mean and standard deviation.  
Clip values between -1 and 1. 
Count how many values were clipped.'''
import numpy as np

arr=np.random.randn(1000)
print(arr)
print(np.mean(arr))
print(np.std(arr))
b=np.clip(arr,-1,1)    # cliping means 
print(b)
count=np.sum(arr!=b)   # only check which value changed after clipping.
print(count)
```

```python
'''Create array: arr = [10, 20, np.nan, 40, 50, np.nan]
Compute mean ignoring NaN. 
Replace NaN values with O. 
Compute new mean. '''
arr=np.array([10, 20, np.nan, 40, 50, np.nan])
print(np.nanmean(arr))    # nanmean will ignore nan values.
a=np.where(np.isnan(arr),0,arr)   #nan is never equal to anyone not even itself so never where in condition arr==np.nan use this np.isnan(arr)
print(a)
print(np.mean(a))
```
