```python
# CREATING DATABASE IN SQL THROUGH PYTHON.

from sqlalchemy import create_engine,text

# create_engine creates a connection between python and mysql.

engine=create_engine("mysql+pymysql://root:@localhost")
with engine.connect() as conn:
    print("Connected Successfully")
    conn.execute(
        text("CREATE DATABASE IF NOT EXISTS analytics_db"))    # conn.execute() means:"Send this SQL query to the database and execute it."
    conn.commit()   #  conn.commit() means:"Save the changes permanently in the database."
    print('DATABASE CREATED SUCCESSFULLY !!')
```

```python
# engine.dispose()  all the connection with data base will be disconnected.
```

```python
# CREATING TABLE IN SQL:--

engine=create_engine("mysql+pymysql://root:@localhost/analytics_db")
with engine.connect() as conn:
    print('connected successfully')
    conn.execute(text('''
    CREATE TABLE IF NOT EXISTS rating(
    user_id INT,
    restaurant_id INT,
    rating INT
    )
    '''))
    conn.commit()
    print('table created !!')
```

```python
# CREATING COMPOSITE PRIMARY KEY:--
engine=create_engine("mysql+pymysql://root:@localhost/analytics_db")
with engine.connect() as conn:
    print('connected successfully')
    conn.execute(text('''
    CREATE TABLE IF NOT EXISTS rating(
    user_id INT,
    restaurant_id INT,
    rating INT,

    PRIMARY KEY(user_id,restaurant_id)
    )
    '''))
    conn.commit()
    print('primary key created !!')

'''giving here two primary keys because we want that uer_id to be unique also restaurant_id to be unique even though
any user can give ratings to multiple restaurant so'''
```

```python
# INSERTING DATA TO TABLE:--

engine=create_engine("mysql+pymysql://root:@localhost/analytics_db")
with engine.connect() as conn:
    print('connected successfully')
    conn.execute(text("""
    INSERT INTO rating
    VALUES
    (1,101,5),
    (2,101,4),
    (1,102,4),
    (2,102,3),
    (1,103,1),
    (2,103,5)
    """))
    conn.commit()
    print('INSERTED SUCCESSFULLY !!')
```

```python
# FETCHING DATA FROM SQL:--

engine=create_engine("mysql+pymysql://root:@localhost/analytics_db")
with engine.connect() as conn:
    print('connected successfully')
    
    result=conn.execute(text("SELECT * FROM rating"))
    for row in result:
        print(row)
```

```python
engine=create_engine('mysql+pymysql://root:@localhost/analytics_db')
with engine.connect() as conn:
    conn.execute(text('''
    ALTER TABLE LOCATION
    ADD CONSTRAINT FK_RESTAURANT
    FOREIGN KEY (RESTAURANT_ID)
    REFERENCES RESTAURANTS(RESTAURANT_ID)
    '''))
    conn.commit()
    print('ADDED FOREIGN KEY...!!')
```

```python
# changed the data-type of restaurant_id in location table:--

with engine.connect() as conn:
    conn.execute(text("""
        ALTER TABLE location
        MODIFY COLUMN restaurant_id INT
    """))
    
    conn.commit()
    print('done')
```

```python
# deleting from restaurants:--
with engine.connect() as conn:
    conn.execute(text('DELETE FROM RESTAURANTS WHERE RESTAURANT_ID=1'))
    conn.commit()
    print('deleted...!!')
```

```python
with engine.connect() as conn:
    conn.execute(text("""
        DROP TABLE location
    """))
    
    conn.commit()
    print('DROPPED...!!')
```

```python
# AGAIN CREATING TABLE LOCATION:--ALSO GIVING FORIGN KEY TO IT..

engine=create_engine('mysql+pymysql://root:@localhost/analytics_db')
with engine.connect() as conn:
    print('connection successfully...!')
    conn.execute(text('''
    CREATE TABLE IF NOT EXISTS LOCATION(
    location_id INT PRIMARY KEY,
    restaurant_id INT,
    city VARCHAR(50),
    FOREIGN KEY(RESTAURANT_ID)
    REFERENCES RESTAURANTS(RESTAURANT_ID)
    ON DELETE CASCADE
    )
    '''))
    conn.commit()
    print('table created successfully...!!')
```

```python
# Write a Python validation script using Pandas to clean the Zomato dataset,handling "NEW" or "-" values in the rating column before relational ingestion.
# CREATING A NEW TABLE ZOMATO AND THEN IN IT NEW COLUMN NAME RATING THEN I WILL ASSIGN VALUES LIKE NEW AND - TO IT THEN I WILL CLEAN IT.
from sqlalchemy import create_engine,text

engine=create_engine('mysql+pymysql://root:@localhost/analytics_db')
with engine.connect() as conn:
    print('connected successfully')
    conn.execute(text("""
    CREATE TABLE IF NOT EXISTS ZOMATO(
    RESTAURANT_ID INT PRIMARY KEY,
    RATING VARCHAR(50)
    )
    """))
    conn.execute(text("""
    INSERT INTO ZOMATO
    VALUES
    (101,4.1),
    (102,'NEW'),
    (103,3.8),
    (104,'-'),
    (105,4.4)
    """))
    conn.commit()
    print('table created successfully....')
```
